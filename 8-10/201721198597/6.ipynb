```python
#练习2 写函数，返回一个list中的最小值
```

```python
def my_min():
    min=1
    for i in numbers:
        if min>i:
            min=i
    return min
numbers=[1,11,22,33,44,55,66,77,88,99]
print('list中的最小值为:',my_min())
```

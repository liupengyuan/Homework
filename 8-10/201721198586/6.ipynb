```python
#2.写函数，返回一个list中所有数字的和
import random
def rand_list_sum():
    number_of_list= int(random.randint(0,7))
    print("input %d numbers : " %number_of_list)
    sum_of_number=0
    numbers=[]
    for i in range (number_of_list):
        number = float(input('input your list number: '))
        numbers.append(number)
        i+=1
    for j in numbers:
        sum_of_number+=j
    return sum_of_number
sum_final = rand_list_sum()
print("Sum of list is %f" %sum_final)
```

```python
#3.写函数，返回一个list中的最小值
import random

def get_min(number_of_list):
    numbers=[]
    min_number=1000
    for i in range (number_of_list):
        number = random.randint(1,10)
        print("the %d number is : %d " %(i,number))
        numbers.append(number)
    for j in numbers:
        if j < min_number:
            min_number = j
    return min_number
size = int(input("input list size : "))
min_number = get_min(size)
print("Min number is %d" %min_number)
```

```python
### 写函数，可求两个向量的夹角余弦值，向量可放在list中。主程序调用该函数。
def cosVector(x,y):
    if(len(x)!=len(y)):
        print("input error!")
        return;
    result1=0.0;
    result2=0.0;
    result3=0.0;
    for i in range(len(x)):
        result1+=x[i]*y[i]   #sum(X*Y)
        result2+=x[i]**2     #sum(X*X)
        result3+=y[i]**2     #sum(Y*Y)
    print("cos = "+str(result1/((result2*result3)**0.5)))
cosVector([2,1],[1,2])
cosVector([2,7,6],[1,6,8])
```

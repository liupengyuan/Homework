```python
# 五言诗游戏

import random
import re

def win():
    print("恭喜你，你赢了")
    return
    
def lose():
    print("哇哦，你输了")
    return
    
def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table

def idiom_robot(file_name,level):
    with open(file_name) as fh:
        text = fh.read()
    idioms = text.split()
    for item in idioms:
        ret_match=re.match("^\d+.*",item)
        if(ret_match):
            idioms.remove(item)
#    print(idioms)
    idiom = random.choice(idioms)
    prom_length = len(idiom)
#    print(str(prom_length)+"/"+idiom)
    chs = get_ch_table(text.replace('\n', ''))

    guess_ch_table = [ch for ch in idiom]
    while len(guess_ch_table) < (prom_length +level*2):
        ch = random.choice(chs)
        if ch not in guess_ch_table:
            guess_ch_table.append(ch)
    
    random.shuffle(guess_ch_table)
    
    index=0
    while(index < len(guess_ch_table)):
        if(index+2 < len(guess_ch_table)):
            print(guess_ch_table[index], guess_ch_table[index+1],guess_ch_table[index+2])
        elif(index+1 < len(guess_ch_table)):
            print(guess_ch_table[index], guess_ch_table[index+1])
        else:
            print(guess_ch_table[index])
        index = index +3
         
    return idiom

def main():
    filename = r'D:\temp\五言诗库.txt'
    score = 0
    level = 1
    print("第一关，当前误导字为2个")
    while score >= 0:
        real_idiom = idiom_robot(filename,level)
        answer_idiom = input('当前关数为'+str(level)+'请输入猜测诗句，回车结束，直接回车表示退出游戏：')
        
        if answer_idiom == real_idiom:
            print('答对了，加十分')
            score += 10
            print('你当前的分数是：', score)
            if score == 100 or score == 200 or score ==300:
                if(level<3):
                    level+=1
                    print("恭喜你来到第二关，当前误导字为4个")
                    
                else:
                    win()
                    return
        elif answer_idiom == '':
            print('退出游戏。')
            print('你最后的分数是：', score)
            return
        else:
            score -= 10
            print('答错了，减十分')
            if(score <100):
                level =1
                print("你掉到了第一关T_T")
            elif(score<200):
                level =2
                print("你掉到了第二关T_T")
            print('诗句其实是：', real_idiom)
            print('你当前的分数是：', score)
    else:
        lose()
        return

#if __name__ == '__main()__':
main()
#idiom_robot(r'D:\temp\五言诗库.txt')
```

```python
#点字成诗游戏
import itertools
import re

def find_poem_sentence(poems, characters):
    same_character_number = 0
    for poem in poems:
        for ch in poem:
            if ch in characters:
                same_character_number += 1
        if same_character_number == 7:
            return poem
    return "抱歉您的库存不足T_T"

fh = open(r'D:\temp\五言诗库.txt')
poems = fh.read().split()

new_poems=[]
for item in poems:
    
    ret_match=re.match("^\d+.*",item)
    if(ret_match):
        poems.remove(item)
    else:
        items = item.split("，")
        for tem in items:
            tem =tem.rstrip("。|？|！")
            new_poems.append(tem)
        
fh.close()
#print(new_poems)

text = '天上的明月照耀着森林中的寒松'
print(text+"-对应的诗句为-"+find_poem_sentence(new_poems,text))
text = '儿郎灯和竹编的马车奔来'
print(text+"-对应的诗句为-"+find_poem_sentence(new_poems,text))
text = '寒冷中的梅花是未来的芬芳'
print(text+"-对应的诗句为-"+find_poem_sentence(new_poems,text))
```

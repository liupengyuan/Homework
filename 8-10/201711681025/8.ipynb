```python
#完整诗词游戏

import itertools
text =str( input('请输入您猜测的诗句，以回车结束'))
guesses = itertools.permutations(text,6) 

fh=open(r'd:\temp\五言诗库.txt')
poems=fh.read().split()
fh.close()

for guess in guesses:
    if guess in poems:
        print('答案是：', guess)
    else:
        print('you lose!')

```

```python
#点字成诗机器人
def find_poem_sentence(poems,characters):
    same_character_number=0
    for poem in poems:
        for ch in poem:
            if ch in characters:
                same_character_number+=1
        if same_character_number ==7:
            return poem
        else:
            print('没有找到您想要的诗句')
        return None
    

fh=open(r'd:\temp\唐诗三百首.txt')
poems=fh.read()
fh.close()

characters=[]
characters=input('请输入您想出现在诗句中的字，以回车结束')
find_poem_sentence(poems,characters)

```

```python
#随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。


import random

a=[random.randint(0,100000) for i in range(10000)]

b=set(a)

c=sorted(b)

for d in c:
    print(d,':',a.count(d))
```

```python
#对本任务中的语料.txt文件，随机抽取其5001-10000行存为test1.txt文件，写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。

with open(r'd:\temp\test1.txt', 'w') as fw,open(r'd:\temp\语料.txt') as fr:
    line_number1 = 0
    for line1 in fr:
        if line_number1 >= 5000 and line_number1 <10000:
            fw.write(line1)
            line_number1 += 1
        elif line_number1 <5000:
            line_number1 += 1           
        else:
            break  
            
with open(r'd:\temp\test.txt', 'w') as fy,open(r'd:\temp\语料.txt') as fr:
    line_number2 = 0   
    for line2 in fr:
        if line_number2 < 5000:
            fy.write(line2)
            line_number2 += 1
        else:
            break
filename1=r'd:\temp\test.txt'
word_table=[]
along_table1=[]
along_table2=[]
with open(filename1) as n:
    text1 = n.read()
    words1 = [word1.split('/')[0] for word1 in text1.split()]
    
filename2=r'd:\temp\test1.txt'
with open(filename2) as m:
    text2 = m.read()
    words2 = [word2.split('/')[0] for word2 in text2.split()]

for i in words1:
    if i in words2:
        word_table.append(i)
    else:
        along_table1.append(i)
for i in words2:
    if i not in words1:
        along_table2.append(i)
print('test1和test的共用字为(前10个)')
for i in range(10):
    print(word_table[i])
print()

print('test的独用字为（前50个）')
for i in range(50):
    print(along_table1[i])
print()

print('test1的独用字为（前50个）')
for i in range(50):
    print(along_table2[i])
print()
```

```python
# 练习2 随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。
#并按照整数大小排序输出整数及出现次数。

import random
from collections import Counter

numbers = []

for i in range(10001):
    num = random.randint(1,100000)
    numbers.append(num)
count_num=Counter(numbers)
#print(count_num)

sorted(count_num.items(),key=lambda x:x[0],reverse=False)

```

```python
#统计共用字和独用字，分别统计2个文件的字表，求合集，以及不同的部分。

def get_word(file):
    word_set=[]
    with open(file) as f:
        for line in f:
            words=[word.split('/')[0] for word in line.split()]
            charect=''.join(words)
            for  hz in charect:
                if hz not in word_set:
                    word_set.append(hz)
    return word_set
            
def count_same_word(table1,table2):
    same_words=[]
    for word in table1:
        if word in table2:
            same_words.append(word)
    return same_words

def count_diff_word(table1,table2):
    diff_words=[]
    for word in table1:
        if word not in table2:
            diff_words.append(word)
    return diff_words

file1 = r'd:\Carol\Python\test1.txt'
table1=get_word(file1)
#print(table[:100])
file2 = r'd:\Carol\Python\test.txt'
table2=get_word(file2)
result_same=count_same_word(table1,table2)
result_diff=count_diff_word(table1,table2)
print('前50个合用字是：',result_same[:50])
print('前50个独用字是：',result_diff[:50])
```

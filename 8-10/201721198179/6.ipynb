```python
# 练习2 写函数，返回list中所有数字的和
# -*- coding:utf-8 -*-

def func_total(n):
    i = 0
    total = 0
    numbers = [] 
    while i < n :
        number=int(input("pls input a number:"))
        numbers.append(number)
        i += 1
        
    for number in numbers:
        total += number
    return total


n = int(input("input the number of figures:"))
print("数字和为：",func_total(n))
```

```python
# 练习3 写函数，返回一个list中的最小值
# -*- coding:utf-8 -*-

def func_min(n):
    i = 0
    numbers = []
    while i < n:
        number = int(input("input a number:"))
        numbers.append(number)
        i += 1

    for number in numbers:
        i = 0
        num_min = numbers[i]
        while i < n :
            if numbers[i] <= num_min :
                num_min = numbers[i]
            i += 1
            
    return num_min
        
n = int(input("input the number of figures:"))
print("最小值为：",func_min(n))        
```

```python
# 练习4 写函数，返回某个元素/对象在一个list中的位置，如果不在，则返回-1
# --*-- coding:utf-8 --*--

numbers=[]
num_list=int(input('pls input the size of number:'))
for i in range(num_list):
    num=int(input('input a number:'))
    numbers.append(num)
    i += 1
    
query=int(input("want to query:"))
if query in numbers:
    print("你输入的数字%s在list的%d位置" %(query,(numbers.index(query)+1)))
else:
    print("-1")
        

    
```

```python
# 练习5 写函数，可求两个向量的夹角余弦值，向量可放在list中。主程序调用该函数
# -*- coding:utf-8 -*-

def cosVector(x,y):
    if(len(x)!=len(y)):
        print('error input,x and y is not in the same space')
        return;
    result1=0.0;
    result2=0.0;
    result3=0.0;
    for i in range(len(x)):
        result1+=x[i]*y[i]   #sum(X*Y)
        result2+=x[i]**2     #sum(X*X)
        result3+=y[i]**2     #sum(Y*Y)
    #print(result1)
    #print(result2)
    #print(result3)
    print("result is "+str(result1/((result2*result3)**0.5))) #结果显示
       
cosVector([2,1],[1,1])
```

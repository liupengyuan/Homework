```python
#随机生成整数
import random
num_list=[]#store the random 10000 integers
count_hash={}
max=100000
while (len(num_list)<10000):
	num=random.uniform(0,max)
	num_list.append(int(num))

for i in num_list:
	i_times=num_list.count(i)
	count_hash[i]=i_times

num_list.sort()

for num in num_list:
	print("这个数是 %d,他出现了 %d 次."%(num,count_hash[num]))

print("我们一共有 %d 整数"%(len(num_list)))
```

```python
#写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。

#抽取5001-10000行
import random

with open(r'd:\lulu\data\test1.txt','w') as fw ,open(r'd:\lulu\data\test.txt','w') as fw2 ,open(r'd:\lulu\data\语料.txt') as fr:
    line_num=0
    for line in fr:
        line_num += 1
        if line_num >5000 and line_num <10001:
            fw.write(line)
        elif line_num <5001:
            fw2.write(line)
        else:
            break
#统计共用字和独用字，分别统计2个文件的字表，求合集，以及不同的部分。

def get_word(file):
    word_set=[]
    with open(file) as f:
        for line in f:
            words=[word.split('/')[0] for word in line.split()]
            charect=''.join(words)
            for  hz in charect:
                if hz not in word_set:
                    word_set.append(hz)
    return word_set
            
def count_same_word(table1,table2):
    same_words=[]
    for word in table1:
        if word in table2:
            same_words.append(word)
    return same_words

def count_diff_word(table1,table2):
    diff_words=[]
    for word in table1:
        if word not in table2:
            diff_words.append(word)
    return diff_words

file1 = r'd:\lulu\data\test1.txt'
table1=get_word(file1)
#print(table[:100])
file2 = r'd:\lulu\data\test.txt'
table2=get_word(file2)
result_same=count_same_word(table1,table2)
result_diff=count_diff_word(table1,table2)
print('前50个合用字是：',result_same[:50])
print('前50个独用字是：',result_diff[:50])
```

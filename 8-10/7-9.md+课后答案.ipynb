```python
#练习二：写函数，根据给定符号和行数，打印相应直角三角形，等腰三角形及其他形式的三角形。

def right_angled_triangle(s):
    print('直角三角形')
    for i in range(7):
        print(s*i)
        
def acute_angled_triangle(s):
    print('锐角三角形')
    for i in range(7):
        print(' '*(12 - i) + s*(3*i + 1))
        
def obtuse_angled_triangle(s):
    print('钝角三角形')
    for i in range(7):
        print(' '*(i+1) + s*(2*i + 1))
        
def isosceles_triangle(s):
    print('等腰三角形')
    for i in range(7):
        print(' '*(7 - i) + s*(2*i + 1))
        
s = input('请输入一个符号')
right_angled_triangle(s)
acute_angled_triangle(s)
obtuse_angled_triangle(s)
isosceles_triangle(s)
```

```python
#对本任务中的语料.txt文件，抽取其5001-10000行存为test1.txt文件，写函数，可得到其与本任务中test.txt文件的共用字以及独用字
def get_word_table(filename):    #给定文本文件，统计其词表
    word_table = []
    with open(filename) as f:
        text = f.read()
    words = [word.split('/')[0] for word in text.split()]
    for word in words:
        if word not in word_table:
            word_table.append(word)
    return word_table

def public_private(filename1,filename2):   #打印共用字和独用字
    word_list1 = get_word_table(filename1)
    word_list2 = get_word_table(filename2)
    public = []
    private = []
    
    for item1 in word_list1:
        flag = 0
        for item2 in word_list2:
            if item1 == item2:
                public.append(item1)
                flag = flag + 1
                break 
        if flag == 0:
            private.append(item1)
            
    print('共用字：',public)
    print('--------------------------------------------------------------------------------------------------------')
    print('独用字：',private)
        
filename1 = r'd:\test1.txt'
filename2 = r'd:\test.txt'
public_private(filename1,filename2)
```

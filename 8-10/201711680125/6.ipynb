```python
#6md.练习2 写函数，返回一个list中所有数字的和
numbers = []
n = int(input('请输入一个整数，表示将要输入的数字个数，回车结束。'))

for i in range(n):
    number = int(input('请输入一个数字，回车结束。'))
    numbers.append(number)

    
print(sum(numbers))
```

```python
#6md.练习3 写函数，返回一个list中的最小值
numbers = []
n = int(input('请输入一个整数，表示将要输入的数字个数，回车结束。'))

for i in range(n):
    number = int(input('请输入一个数字，回车结束。'))
    numbers.append(number)
    
print(min(numbers))
```

```python
#6md.练习4 写函数，返回某个元素/对象在一个list中的位置，如果不在，则返回-1.
numbers = []
n = int(input('请输入一个整数，表示将要输入的数字个数，回车结束。'))

for i in range(n):
    number = int(input('请输入一个数字，回车结束。'))
    numbers.append(number)
    
o = int(input('请输入一个整数，回车结束。'))
if o in numbers:
    print(numbers.index(o))
else :
    print(-1)
```

```python
#6md.练习5 写函数，可求两个向量的夹角余弦值，向量可放在list中。

m=int(input('请输入第一个向量的横坐标，回车结束。'))
n=int(input('请输入第一个向量的纵坐标，回车结束。'))
r=int(input('请输入第二个向量的横坐标，回车结束。'))
t=int(input('请输入第二个向量的纵坐标，回车结束。'))
    
c=(m*r)+(n*t)
o=((m**2+n**2)**(1/2))*((r**2+t**2)**(1/2))
cos_angle=c/o
print(cos_angle)
```

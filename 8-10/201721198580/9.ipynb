2、随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。

```python
import random
from collections import Counter

nums_freq_dict={}
for i in range(0,10000):
    num = random.randint(0, 100000)
    if num in nums_freq_dict: 
        nums_freq_dict[num] += 1
    else:   
        nums_freq_dict[num] = 1

num_list=sorted(nums_freq_dict.items(),key=lambda e:e[0],reverse=False)
for term in num_list:
    print(term)
```

3、对本任务中的语料.txt文件，随机抽取其5001-10000行存为test1.txt文件，写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。

```python
import random

with open(r'f:\eclipse\file\test.txt', 'w') as fw, open(r'd:\temp\语料.txt') as fr:
    line_number = 0
    for line in fr:
        if line_number < 5000:
            fw.write(line)
            line_number += 1
        else:
            break
            

num_list = range(5001,10001)
line_list = random.sample(num_list, 2000)

with open(r'f:\eclipse\file\test1.txt', 'w') as fw, open(r'd:\temp\语料.txt') as fr:
    line_number = 0
    for line in fr:
        if line_number < 10000:
            if line_number in line_list: 
                fw.write(line)       
            line_number += 1
        else:
            break
```

```python
def get_words(file1,file2):
    test=[]
    test1=[]
    common = []
    independent = []
    with open(file1) as f1:
        for line in f1:
            words = [word.split('/')[0] for word in line.split()]
            for word in words:
                test.append(word)
        
    with open(file2) as f2:
        for line in f2:
            words = [word.split('/')[0] for word in line.split()]
            for word in words:
                test1.append(word)
                    
    for word in test:
        for word1 in test1:
            if word == word1:
                common.append(word)
            else:
                independent.append(word)
                
    print("共用字为：\n")
    print(common) 
    print("独用字为：\n")
    print(independent)
            
file=r'f:\eclipse\file\test.txt'
file1=r'f:\eclipse\file\test1.txt'
get_words(file,file1)
```

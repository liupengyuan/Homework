写函数，返回一个list中所有数字的和

```python
numbers = []
number = int(input('输入数字的个数:'))
for i in range(number):
    num = int (input('输入数字：'))
    numbers.append(num)
print('所有数字的和是：', sum(numbers))
```

写函数，返回一个list中的最小值

```python
numbers = []
number = int(input('输入数字的个数:'))
for i in range(number):
    num = int (input('输入数字：'))
    numbers.append(num)
print('所有数字中最小的一个数是：', min(numbers))
```

```python
def distribute(n):
    dis = []
    num = 0
    for i in range(n):
        for j in range(n):
            for k in range(n):
                if (i + j + k == n and i <=j and i <= k and j <=k):
                    num += 1
                    dis_one = (i + 5, j + 5, k + 5)
                    dis.append(dis_one)
    print(num) 
    print(dis) 
n = 100 - 5 * 3
print (distribute(n))
```

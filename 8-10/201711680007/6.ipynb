```python
#•写函数，返回一个list中所有数字的和
def listn_total():
    numbers=[]
    n=int(input("请输入一个整数表示list中数字个数"))
    i = 0
    total=0
    while i < n:
        number = int(input('请输入一个数字，回车结束。'))
        numbers.append(number)
        i += 1
        total=total+number
    print(numbers)
    print("该list中所有数字的和是",total)
    
    
    
#主程序
listn_total()
```

```python
#•写函数，返回一个list中的最小值
def little():
    
    numbers=[]
    n=int(input("请输入一个整数表示list中数字个数"))
    min_number = int(input('请输入一个整数，回车结束'))
    i = 0
    while i < n:
        m=int(input('请输入一个数字'))
        numbers.append(m)
        i += 1
        if m > min_number:
            min_number = min_number
        elif m < min_number:
            min_number = m
    print(numbers)
        
    print("该list中最小数字是",min_number)
    
    
#主程序    
little()
```

```python
#•写函数，返回某个元素/对象在一个list中的位置，如果不在，则返回-1.
words=["apple","banana",'watermelon','orange','coconut']

print('该单词list是',words)
x=0
n=10000
while x < n:
    word=input('请输入一个水果类英文单词')
    
    if word in words:
        i=words.index(word)
        print('该单词在list中的位置是',(i+1))
    if word not in words:
        print("-1")
   
```

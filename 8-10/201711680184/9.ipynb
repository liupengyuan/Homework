```python
#2、随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
import random

def sort(seq):
    
    seq.sort()
    
    int_freq_pairs = []
    for num in seq:
        for item in int_freq_pairs:
            if num == item[0]:
                item[1]+=1
                break
        else:
            int_freq_pairs.append([num,1])
        
    return int_freq_pairs

seq = [random.randint(0,100001) for i in range(10000)]
print(sort(seq))
```

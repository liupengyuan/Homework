```python
import random


class Poem():
    def __init__(self, text, number):
        self.text = text
        self.number = number
        self.writer = ''
        self.title = ''
        self.content = ''
        self.lines = []

    def get_details(self):
        i = self.text.index('\n\n')
        j = self.text.index('：')
        self.writer = self.text[3:j]
        self.title = self.text[j+1:i]
        self.content = self.text[i+2:]
        t = self.content.replace('\n','')

        i = 0
        j = 0
        while i < len(t) + 2:
            i += 1
            if i % 6 == 5:
                self.lines.append(t[j:i])
                j = i+1

def split_poems():
    f1 = open(r'E:\Courseware\Python\五言诗库.txt')
    f1_text = f1.read()
    f1.close()

    t = f1_text
    i, j = 0, 0
    num = ['1', '2', '3', '4', '5', '6', '7', '8', '9', '0']
    poems = []
    while i < len(t):
        if t[i] in num and t[i+1] in num and t[i+2] in num and i != 0:
            poems.append(t[j:i-2])
            j = i
        i += 1
    poems.append(t[j:j+63])
    return(poems)

def get_character(t):

    characters = []
    for poem in t.values():
        for line in poem.lines:
            for character in line:
                if character not in characters:
                    characters.append(character)

    return characters

def generate_poem(t):
    temp = []
    for value in t.values():
        temp.append(value)
    poem = random.choice(temp)
    return poem

def generate_guess_table(line, characters):
    guess_ch_table = [ch for ch in line]
    while len(guess_ch_table) < 8:
        ch = random.choice(characters)
        if ch not in guess_ch_table:
            guess_ch_table.append(ch)

    random.shuffle(guess_ch_table)

    for i in range(0, 6, 4):
        print(guess_ch_table[i], guess_ch_table[i + 1], guess_ch_table[i + 2], guess_ch_table[i + 3])

def build_lib(t):
    poems = split_poems()
    for poem in poems:
        num = int(poem[0:3])
        t[num] = Poem(poem, num)
        t[num].get_details()
    return t



def welcome():
    print(
        '''
           ======欢迎来到猜数游戏=======

                     /  \     ,    ,
           _._     _ |oo| _  / \__/ \

          _||||   ((/ () \))   /   \

          |||||/|  ( ==== )    |oo|    
           \____/  _`\  /'_    /   \

           /   /.-' /\<>/\ `\.( () )_._      
           |    `  /  \/  \  /`'--'////)
            \__,-'`|  |.  |\/ |/\/\|"\"` 
                   |  |.  | \___/\___/  
                   |  |.  |   |    |

          ======欢迎来到猜数游戏=======
        '''
    )

def win():
    print(
        '''
           ======恭喜你，你赢了=======


                ."".    ."",
                |  |   /  /
                |  |  /  /
                |  | /  /
                |  |/  ;-._ 
                }  ` _/  / ;
                |  /` ) /  /
                | /  /_/\_/\
                |/  /      |
                (  ' \ '-  |
                 \    `.  /
                  |      |
                  |      |

          ======恭喜你，你赢了=======
        '''
    )

def lose():
    print(
        '''
           ======YOU LOSE=======




                   .-"      "-.
                  /            \
                 |              |
                 |,  .-.  .-.  ,|
                 | )(__/  \__)( |
                 |/     /\     \|
       (@_       (_     ^^     _)
  _     ) \_______\__|IIIIII|__/__________________________
 (_)@8@8{}<________|-\IIIIII/-|___________________________>
        )_/        \          /
       (@           `--------`



          ======YOU LOSE=======
        '''
    )

def game_over():
    print(
        '''
           ======GAME OVER=======

             _________ 
            / ======= \ 
           / __________\ 
          | ___________ | 
          | | -       | | 
          | |         | | 
          | |_________| |________________ 
          \=____________/                ) 
          / """"""""""" \               / 
         / ::::::::::::: \          =D-' 
        (_________________) 


          ======GAME OVER=======
        '''
    )

def menu():
    print('=====游戏菜单=====')
    print('1. 游戏说明')
    print('2. 开始游戏')
    print('3. 退出游戏')
    print('4. 制作团队')
    print('=====游戏菜单=====\n')
    choose = int(input('您的选择是：'))
    print('')
    if choose == 1:
        print('这是一个猜诗的游戏：\n\t\t您讲在屏幕上看到八个字\n\t\t这八个字中包含有五言律诗中的一句\n\t\t请将这句五言律诗写下来\n\t\t本游戏一共包含10个五言律诗\n\n\t\t您有50初始积分\n\t\t若您回答正确，则加10分\n\t\t若回答错误，则扣十分\n\t\t若游戏结束时，您还有积分，则获胜，否则失败')
        menu()
    elif choose == 2:
        game()
    elif choose == 3:
        game_over()
    elif choose == 4:
        print('''
        制作团队：
            总设计： 谢松成
            程序员： 谢松成
        特别鸣谢：
        
               谢松成    
        
        ''')
        menu()

def game():
    t = {}
    build_lib(t)
    poem = generate_poem(t)
    characters = get_character(t)

    player_points = 50
    guessed_poem = []
    i = 0
    while player_points > 0 and i < 11:
        i += 1
        while poem in guessed_poem:
            poem = generate_poem(t)
        else:
            guessed_poem.append(poem)
            line = random.choice(poem.lines)
            generate_guess_table(line, characters)
            poem_guess = input('\n请输入您的答案。\n')

            if poem_guess == line:
                player_points += 10
                print('你猜对了，这首诗出自' + poem.writer + '的《' + poem.title + '》。')
                print('你现在有' + str(player_points) + '分。\n')
            else:
                player_points -= 10
                print('“' + line + '”')
                print('这首诗出自' + poem.writer + '的《' + poem.title + '》。')
                print('你还有' + str(player_points) + '分。\n')
    if player_points > 0:
        win()
        print('您最后的分数是：' + str(player_points) + '。')
        game_over()
    else:
        lose()
        game_over()

def main():
    welcome()
    menu()

if __name__ == '__main__':
    main()
```

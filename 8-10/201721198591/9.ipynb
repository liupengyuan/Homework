```python
#2、随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
import random

i = 0
freq=1
num_table = []
num_freq_dict = {}
while i < 10000:
    num_table.append(random.randint(0,100000))
    i = i + 1
num_table.sort()
for num in num_table:
    if num in num_freq_dict:
        num_freq_dict[num] += 1
    else:
        num_freq_dict[num] = 1
    
print(num_freq_dict)
```

```python
#3、对本任务中的语料.txt文件，随机抽取其5001-10000行存为test1.txt文件，写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）

with open(r'D:\python\语料\语料.txt') as f,open(r'D:\python\test1.txt','w') as fh:
    line_num = 5000
    for line in f:
        if  line_num <= 10001:
            fh.write(line)
            line_num += 1
        else:
            break

# test.txt 和 test1.txt的词表生成
test_word_table=set()
test1_word_table=set()
with open(r'D:\python\test.txt') as fw,open(r'D:\python\test1.txt') as fh:
    for line in fw:
        test_word_table |= {word.split('/')[0] for word in line.split()}
    for line in fh:
        test1_word_table |= {word.split('/')[0] for word in line.split()} 

same_word_table =  test_word_table & test1_word_table
different_word_table = test_word_table - test1_word_table
print('同用字表：',same_word_table)  
print('独用字表：',different_word_table)
```

```python
#210-10
import random

i = 0
freq=1
num_table = []
num_freq_dict = {}
while i < 10000:
    num_table.append(random.randint(0,100000))
    i = i + 1
num_table.sort()
for num in num_table:
    if num in num_freq_dict:
        num_freq_dict[num] += 1
    else:
        num_freq_dict[num] = 1
    
print(num_freq_dict)
```

```python
#3.2写函数，可得到其与本任务中`test.txt`文件的共用字以及独用字（相关概念自行百度）。
def get_vocabulary(filename):
    word_table = set()
    with open(filename,encoding='utf-8') as f:
        for line in f:
            word_table |= {word.split('/')[0] for word in line.split()}             
    return word_table

#测试用

filename = r'/Users/weilinxuan/Project/Python/LiuPengYuan/test.txt'
set0 = get_vocabulary(filename)
filename = r'/Users/weilinxuan/Project/Python/LiuPengYuan/test1.txt'
set1 = get_vocabulary(filename)
#####注：为了方便展示 只取前100个，若要看完整结果，移除下述代码中的[:100]即可
print("共用字：")
print(list(set0 & set1)[:100])
print("\ntest的独用字：")
print(list(set0 - set1)[:100])
print("\ntest1的独用字：")
print(list(set1 - set0)[:100])
```

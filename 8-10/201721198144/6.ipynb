```python
#P1： 写函数，返回一个list中所有数字的和
num_list = []
n = int(input("请输入您要输入的数字的个数： "))

i = 0
while i < n:
	num = int(input("请输入一个数字： "))
	num_list.append(num)
	i += 1

def total():
	total = 0
	for num in num_list:
		total += num
	return total

print ("他们的和是：",total())
```

```python
#P3： 写函数，返回某个元素/对象在一个list中的位置，
#如果不在，则返回-1.

lis = []
n = int(input("请输入您要输入的对象的个数："))

i = 0
while i <n:
	val = input("请输入一个对象：")
	lis.append(val)
	i +=1

def look_val(lis):
	look_val = input("请输入您要查找的对象：")
	if look_val in lis:
		result = (lis.index(look_val)+1)
	else:
		result = "-1"

	return result

print("您查询的对象在：",look_val(lis))
```

```python
#P4：写函数，可求两个向量的夹角余弦值，向量可放在list中。主程序调用该函数。
def cos_vec(x,y):
	result1=0.0
	result2=0.0
	result3=0.0

	for i in range(len(x)):
		result1 += int(x[i]) * int(y[i])
		result2 += int(x[i])**2
		result3 += int(y[i])**2

	result = result1/((result2*result3)**0.5)
	return result

x=list(input("pls enter a vector:"))
y=list(input("pls enter a vector:"))
print(x,y,"的夹角余弦值是：", cos_vec(x,y))
```

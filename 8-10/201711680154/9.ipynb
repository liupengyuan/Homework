```python
# 2、随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
import random
my_list = []
for i in range(10000):
    a = random.randint(0,100000)
    my_list.append(a)
# print(my_list)
words_freq_dict = {}
for word in my_list:
    if word in words_freq_dict: #如果词典中有键为word，则该word的value+1
        words_freq_dict[word] += 1
    else:                       #否则，词典中加入键word，其value为1
        words_freq_dict[word] = 1
# print(words_freq_dict)
print (sorted(words_freq_dict.items(), key=lambda v: v[0]))
```

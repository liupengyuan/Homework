```python
import random
from collections import Counter

def main():
    numbers = [random.randint(0, 100000) for i in range(10000)]
    nums = Counter()
    nums.update(numbers)
    nums_list = [number for number in nums.keys()]
    nums_list.sort()
    for number in nums_list:
        print(str(number) + ':' + str(nums[number]))

if __name__ == '__main__':
    main()
```

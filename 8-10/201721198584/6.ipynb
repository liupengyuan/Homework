```python
#练习4：写函数，返回某个元素/对象在一个list中的位置，如果不在，则返回-1.
def SearchList(list,elem):
    for i, j in enumerate(list):
        if elem == j :
            return (i)
    return (-1)

numbers = [1, 2, 3, 4, 5, 6, 7]

print(SearchList(numbers,5), '\n')
print(SearchList(numbers,9), '\n')
```

```python
#写函数，可求两个向量的夹角余弦值，向量可放在list中。主程序调用该函数。
import math
def GetCos(a,b):
    mulSum=0;
    aSqrtSum=0;
    bSqrtSum=0;
    for i in range(2):
        mulSum += (a[i] * b[i])
        aSqrtSum += (a[i] ** 2)
        bSqrtSum += (b[i] ** 2)
    return  (mulSum/(math.sqrt(aSqrtSum * bSqrtSum)))

if __name__ == "__main__":
    a = [ 0, 1 ]
    b = [ 1, 1 ]
    print ( '向量',a,'和向量',b,'的夹角余弦为：',GetCos(a,b))
        
```

```python
import random
from collections import defaultdict
def count_n(nums):
    num_freq_dict = defaultdict(int)
    for num in nums:
        num_freq_dict[num] += 1    
    return num_freq_dict 
def main():
    nums=[random.randint(0,10000) for x in range(10000)]
    nums.sort(reverse=True) 
    print(count_n(nums))
if __name__ == '__main__':
    main()
```

```python
#3、对本任务中的语料.txt文件，随机抽取其5001-10000行存为test1.txt文件，
#写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。

def GetSet(filename):
    word_table = set()
    with open(filename) as f:
        for line in f:
            word_table |= {word.split('/')[0] for word in line.split()}  
    return word_table

def main():
    f1 = r'D:\Post Graduate\class\python\text.txt'
    f2 = r'D:\Post Graduate\class\python\text1.txt'
    set0=GetSet(f1)
    set1=GetSet(f2)
    print('共用字集合：')
    print(set0|set1)
    print('test.txt 独用字集合：')
    print(set0-set1)
    print('test1.txt 独用字集合：')
    print(set1-set0)


if __name__ == '__main__':
    main()
```

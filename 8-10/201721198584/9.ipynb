```python
import random
from collections import defaultdict
def count_n(nums):
    num_freq_dict = defaultdict(int)
    for num in nums:
        num_freq_dict[num] += 1    
    return num_freq_dict 
def main():
    nums=[random.randint(0,10000) for x in range(10000)]
    nums.sort(reverse=True) 
    print(count_n(nums))
if __name__ == '__main__':
    main()
```

```python
import random
from collections import Counter

def main():
    numbers = [random.randint(0, 100000) for i in range(10000)]
    nums = Counter()
    nums.update(numbers)
    nums_list = [number for number in nums.keys()]
    nums_list.sort()
    for number in nums_list:
        print(str(number) + ':' + str(nums[number]))


if __name__ == '__main__':
    main()
```

```python
from collections import Counter

def get_text1(filename,save_filename):
    with open(filename) as f:
        text1 = open(save_filename,'w')
        i = 0
        for line in f:
            i += 1
            if i > 5000 and i < 10001:
                text1.writelines(line)
            if i > 10000:
                break
        text1.close()
        f.close()

def main():
    get_text1(r'D:\语料.txt',r'D:\text.txt')
    words_freq_dict = Counter()
    words_common = []
    words_special = []
    with open(r'D:\text.txt') as f:
        for line in f:
            words_freq_dict.update([word.split('/')[0] for word in line.split()])
            words_list1 = [word for word in words_freq_dict.keys()]
    with open(r'D:\test.txt') as f:
        for line in f:
            words_freq_dict.update([word.split('/')[0] for word in line.split()])
            words_list2 = [word for word in words_freq_dict.keys()]
    for word in words_list2:
        if word in words_list1:
            words_common.append(word)
        else:
            words_special.append(word)

    print(len(words_common),len(words_special))
    print(words_common)
    print(words_special)


if __name__ == '__main__':
    main()

```

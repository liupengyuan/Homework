```python
#练习2 随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
import random

n = 0
freq=1
num_table = []
num_freq_dict = {}
while n < 10000:
    num_table.append(random.randint(0,100000))
    n = n + 1
num_table.sort()
for num in num_table:
    if num in num_freq_dict:
        num_freq_dict[num] += 1
    else:
        num_freq_dict[num] = 1
    
print(num_freq_dict)
```

```python
with open(r'E:\python\test.txt', 'w') as fw, open(r'E:\python\语料.txt') as fr:
    line_number = 0
    for line in fr:
        line_number += 1
        if line_number < 5000:
            fw.write(line)
        else:
            break

```

```python
filename = r'E:\python\语料.txt'

def random_file(filename):
    i = 0
    fh = open(r'E:\python\test1.txt','w')
    with open(filename) as f:
        for line in f:
            i = i + 1
            if i > 5000 and i < 10001:
                fh.write(line)
    fh.close()
    return r'E:\python\test1.txt'

random_file(filename)
```

```python
test_word_table=set()
test1_word_table=set()
with open(r'E:\python\test.txt') as fw,open(r'E:\python\test1.txt') as fh:
    for line in fw:
        test_word_table |= {word.split('/')[0] for word in line.split()}
    for line in fh:
        test1_word_table |= {word.split('/')[0] for word in line.split()} 

gongyong_word_table =  test_word_table & test1_word_table
duyong_word_table = test_word_table - test1_word_table

print('共用字表：')
print(gongyong_word_table) 
print('独用字表：')
print(duyong_word_table)
```

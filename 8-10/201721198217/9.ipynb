```python
#2 10-10
import random

n = 0
freq=1
num_table = []
num_freq_dict = {}
while n < 10000:
    num_table.append(random.randint(0,100000))
    n = n + 1
num_table.sort()
for num in num_table:
    if num in num_freq_dict:
        num_freq_dict[num] += 1
    else:
        num_freq_dict[num] = 1
    
print(num_freq_dict)
```

```python
#练习2
def compute(list1):
    total=0
    for i in list1:
        total+=i
    return total
print(compute([1,2,3]))
```

```python
#练习3
def compute(list2):
    min_list=list2[0]
    for i in list2:
        if i<min_list:
            min_list=i
    return min_list
print(compute([3,1,2]))
        
    
```

```python
#练习4
def compute(number,list1):
    if number not in list1:
        return -1
    else:
        return list1.index(number)
    
list1=[]    
m=int(input('输入列表元素的个数'))
i=0
while i<m:
    i+=1
    n=int(input('输入列表的一个组成元素'))
    list1.append(n)
print(list1)
       
number=int(input('输入要调查的元素'))  

print(compute(number,list1))
    
    
    
    
```

```python
#练习5
import math
def compute(m,n):
    if len(m)!=2 or len(n)!=2:
        print('向量错误')
        return False
    mn=m[0]*n[0]+m[1]*n[1]
    return mn/math.sqrt((m[0]**2+m[1]**2)*(n[0]**2+n[1]**2))
print(compute(m=[13,20],n=[18,20]))
    
    
```

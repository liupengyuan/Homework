```python
#2
import random
from collections import Counter
import operator

def num_freq(numbers):
    num_freq_dict=Counter()
    num_freq_dict.update(numbers)
    return sorted(num_freq_dict.items(), key=operator.itemgetter(0)) 
 
numbers=[random.randint(0,100000) for _ in range(10000)]
print(num_freq(numbers))
```

```python
#3

def words_freq(filename):
    words_freq_table = set()
    
    with open(filename) as f:
        for line in f:
            words_freq_table|={word.split('/')[0] for word in line.split()}
    return words_freq_table

filename1=r'd:\temp\test.txt'
table1=words_freq(filename1)
filename2=r'd:\temp\test1.txt'
table2=words_freq(filename2)
print('',table1&table2)
print('',(table1-table2)|(table2-table1))

```

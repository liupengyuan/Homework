```python
#210-10

import random
from operator import itemgetter
def random_list():
    length=10000
    random_list=[]
    for i in range(0,length):
        random_list.append(random.randint(0,10000))
    return random_list

#
def get_word_table(random_list):
    word_list=[]
    for num in random_list:
        if num not in word_list:
            word_list.append(num)
    return word_list

#
def count_number(word_list,random_list):
    count_list=[]
    for num in word_list:
        countnum=random_list.count(num)
        count_list.append([num,countnum])
    return count_list

#
def sortlist(count_list):
    count_list.sort(key=itemgetter(1))
    return count_list
  
randomlist=[]
wordlist=[]
countlist=[]

randomlist=random_list()  
wordlist=get_word_table(randomlist)
countlist=count_number(wordlist,randomlist)
countlist=sortlist(countlist)
print(countlist)
```

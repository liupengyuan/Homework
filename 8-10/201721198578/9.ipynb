```python
#2、随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。

import random
from operator import itemgetter
def random_list():
    length=10000
    random_list=[]
    for i in range(0,length):
        random_list.append(random.randint(0,10000))
    return random_list

#传入随机数数组
def get_word_table(random_list):
    word_list=[]
    for num in random_list:
        if num not in word_list:
            word_list.append(num)
    return word_list

#传入不同数字的数组
def count_number(word_list,random_list):
    count_list=[]
    for num in word_list:
        countnum=random_list.count(num)
        count_list.append([num,countnum])
    return count_list

#排序
def sortlist(count_list):
    count_list.sort(key=itemgetter(1))
    return count_list
  
randomlist=[]
wordlist=[]
countlist=[]

randomlist=random_list()  
wordlist=get_word_table(randomlist)
countlist=count_number(wordlist,randomlist)
countlist=sortlist(countlist)
print(countlist)
```

```python
#3、对本任务中的语料.txt文件，随机抽取其5001-10000行存为test1.txt文件，写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。

from operator import itemgetter

with open(r'd:\temp\test.txt', 'w') as fw, open(r'd:\temp\语料.txt','r') as fr:
    line_number = 0
    for line in fr:
        if line_number > 5000 and line_number < 10000:
            fw.write(line)
        line_number += 1
        if line_number >10001:
            break

def count_words_freq(filename):
    word_freq_pairs = []
    
    with open(filename) as f:
        text = f.read() 
    words = [word.split('/')[0] for word in text.split()]
    for word in words:
        for item in word_freq_pairs:
            if word == item[0]:
                item[1] += 1
                break
        else:
            word_freq_pairs.append([word, 1])
    return word_freq_pairs

filename = r'd:\temp\test.txt'
table = count_words_freq(filename)
table.sort(key=itemgetter(1))
print(table) 
```

```python
#随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
import random

numbers = []

for i in range(10000):
    number = random.randint(0,100000)
    numbers.append(number)

numbers_frequency = []

for number in numbers:
    for item in numbers_frequency:
        if number ==item[0]:
            item[1] = item[1] + 1
            break
    else:
        numbers_frequency.append([number,1])
numbers_frequency.sort()

length = len(numbers_frequency)
j = 0
for j in range(length):
    print(numbers_frequency[j])
```

```python
#对本任务中的语料.txt文件，抽取其5001-10000行存为test1.txt文件，写函数，可得到其与本任务中test.txt文件的共用字以及独用字
def get_word_table(filename):    #给定文本文件，统计其词表
    word_table = []
    with open(filename) as f:
        text = f.read()
    words = [word.split('/')[0] for word in text.split()]
    for word in words:
        if word not in word_table:
            word_table.append(word)
    return word_table

def public_private(filename1,filename2):   #打印共用字和独用字
    word_list1 = get_word_table(filename1)
    word_list2 = get_word_table(filename2)
    public = []
    private = []
    
    for item1 in word_list1:
        flag = 0
        for item2 in word_list2:
            if item1 == item2:
                public.append(item1)
                flag = flag + 1
                break 
        if flag == 0:
            private.append(item1)
            
    print('共用字：',public)
    print('--------------------------------------------------------------------------------------------------------')
    print('独用字：',private)
        
filename1 = r'd:\test1.txt'
filename2 = r'd:\test.txt'
public_private(filename1,filename2)
```

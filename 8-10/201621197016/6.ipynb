```python
#写函数，返回一个list中所有数字的和
numbers=[]
n=int(input('请输入一个整数，表示数列数字个数，并以回车结束：'))

i=0
while i<n:
      number=int(input('请输入数字，并以回车结束：'))
      numbers.append(number)
      i +=1
total=0
for number in numbers:
    total +=number
print('numbers：',total)
```

```python
#写函数，返回一个list中的最小值
numbers=[]
n=int(input('请输入一个整数，表示数列数字个数，并以回车结束：'))

i=0
while i<n:
      number=int(input('请输入数字，并以回车结束：'))
      numbers.append(number)
      i +=1

print('numbers序列最小值：',min(numbers),'\n')
```

```python
#写函数，返回某个元素/对象在一个list中的位置，如果不在，则返回-1.
numbers=[]
n=int(input('请输入一个整数，表示数列数字个数，并以回车结束：'))
i=0
while i<n:
      number=int(input('请输入数字，并以回车结束：'))
      numbers.append(number)
      i +=1

def place(m):
    if m in numbers:
        return numbers.index(m)
    else:
        return -1

m=int(input('请输入一个对象的值：'))
print(m,'在numbers序列中的位置是',place(m),'，是第',place(m)+1,'个对象。')
```

```python
#写函数，可求两个向量的夹角余弦值，向量可放在list中。主程序调用该函数。
m=int(input('请输入第一个向量的横坐标，回车结束。'))
n=int(input('请输入第一个向量的纵坐标，回车结束。'))
x=int(input('请输入第二个向量的横坐标，回车结束。'))
y=int(input('请输入第二个向量的纵坐标，回车结束。'))
    
cos=((m*x)+(n*y))/(((m**2+n**2)**(1/2))*((x**2+y**2)**(1/2)))
print(cos)
```

```python
#挑战性习题：python语言老师为了激励学生学python，
#自费买了100个完全相同的Macbook Pro，分给三个班级，每个班级至少分5个，
#用穷举法计算共有多少种分法？
order=0
for i in range(5,91):
    for j in range(5,91):
        for k in range(5,91):
            if i+j+k==100:
                order+=1
                print('第',order,'种分法：',i,j,k)
            else:
                continue
print('一共有',order,'种分法。')
```

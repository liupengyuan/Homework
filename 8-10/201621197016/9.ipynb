```python
import random

numbers = {}
for i in range(0,10000):
    number = random.randint(0,100000)
    if number in numbers: 
        numbers[number] += 1
    else: 
        numbers[number] = 1

list=sorted(numbers,reverse = True)
k=0
for ele in list:
    k=k+1
    if k%9==0:
        print('\n')
        k=0
    print(ele,':',numbers[ele],end='  ')
```

```python
from collections import Counter

def random_file1(filename):
    i = 0
    fh = open(r'c:\moqiming\test_yuliao.txt','w')
    with open(filename) as f:
        for line in f:
            i = i + 1
            if i > 5000:
                fh.close()
                return r'c:\moqiming\test_yuliao.txt'
            fh.write(line)
            
def random_file2(filename):
    i = 0
    fh = open(r'c:\moqiming\test_yuliao.txt','w')
    with open(filename) as f:
        for line in f:
            i = i + 1
            if i > 5000 and i < 10001:
                fh.write(line)
    fh.close()
    return r'c:\moqiming\test_yuliao.txt'
    
def compare(filename1,filename2):
    dict1 = Counter()
    dict2 = Counter()
    same = {}
    differ = {}
    
    with open(filename1) as f1:
        for line in f1:
            dict1.update([word.split('/')[0] for word in line.split()])
    with open(filename2) as f2:
        for line in f2:
            dict2.update([word.split('/')[0] for word in line.split()])
    for word in dict2:
        if word in dict1:
            same[word] = 1
        else:
            differ[word] = 1

                
    return same,differ

def main():
    filename = r'c:\moqiming\test_yuliao.txt'
    filename1 = random_file1(filename)
    filename2 = random_file2(filename)
    same,differ = compare(filename1,filename2)
    print('这两个文件的共用字为：')
    for key in same:            
        print(key,end='  ')
    print('\n-------------------------------------------------------------------------------------------------------')
    print('这两个文件的独用字为：')
    for key in differ:            
        print(key,end='  ')
   
if __name__ == '__main__':
    main()
```

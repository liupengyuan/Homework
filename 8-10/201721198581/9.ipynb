# 2、随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。

```python
import random

def count_numbers_freq_dict():
    numbers_freq_dict = {}
    
    for i in range(10000):
        number = random.randint(0,100000)
        if number in  numbers_freq_dict:
            numbers_freq_dict[number] += 1
        else:
            numbers_freq_dict[number] = 1
    
    return numbers_freq_dict

#测试用

table = count_numbers_freq_dict()
a = sorted(table)
for i, item in enumerate(a):
    print(item, table[item])
```

# 3、对本任务中的语料.txt文件，随机抽取其5001-10000行存为test1.txt文件，
# 写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。

```python
with open(r'f:\课件\A\python\test.txt', 'w') as fw1, open(r'f:\课件\A\python\test_re_gbk.txt') as fr,open(r'f:\课件\A\python\test1.txt', 'w') as fw2:
    line_number = 0
    word_table1 = set()
    word_table2 = set()
    
    for line in fr:
        if line_number < 200:
            word_table1 |= {word.split('/')[0] for word in line.split()}
            fw1.write(line)
            line_number += 1
        elif line_number >= 200 and line_number < 400:
            word_table2 |= {word.split('/')[0] for word in line.split()}
            fw2.write(line)
            line_number += 1
        else:
            break

ch_table = set()        
for wordbase in word_table1:
    ch_table |= {ch for ch in wordbase}

ch_share =[]
ch_single = []
for word in word_table2:
    for chara in word:
        if chara in ch_table:
            ch_share.append(chara)
        else:
            ch_single.append(chara)

print('共用字：\n')
for Zishare in ch_share:
    print(Zishare, end='|')
print('\n')    
print('独用字：\n')
for Zisingle in ch_single:
    print(Zisingle, end='、')
```

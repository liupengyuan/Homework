```python
# 随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
import random
num_list = [random.randint(0,100000) for i in range(10000)]
num_dict = {}
for num in num_list:
    if num in num_dict:
        num_dict[num] += 1
    else:
        num_dict[num] = 1
print({k:num_dict[k] for k in sorted(num_dict)})
```

```python
# 3、对本任务中的语料.txt文件，随机抽取其5001-10000行存为test1.txt文件，写函数，
# 可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。
f = open('./语料.txt','r',encoding='gbk')
f_test1 = open('./test1.txt','w+')
f_test2 = open('./test2.txt','w+')
num = 1
for line in f:
    if num <= 50:
        f_test1.write(line)
        num += 1
    if num > 50 and num <= 100:
        f_test2.write(line)
        num += 1
    if num > 100:
        break
f_test1.seek(0)
f_test2.seek(0)
vocb_test1 = set()
vocb_test2 = set()
for line in f_test1:
    line = set(word_lable.split('/')[0] for word_lable in line.split(' '))
    vocb_test1 |= line
for line in f_test2:
    line = set(word_lable.split('/')[0] for word_lable in line.split(' '))
    vocb_test2 |= line
intersec = vocb_test1 & vocb_test2  #交集
test1_single = vocb_test1 - intersec
test2_single = vocb_test2 - intersec
print('共用字：',intersec)
print('test1中独用字：',test1_single)
print('test2中独用字：',test2_single)
f.close()
f_test1.close()
f_test2.close()
```

```python
n=int(input('请输入一个正整数，以回车结束。'))

i=1
total=1

while i<n:
    i=i+1
    total=total*i
    
print(total)
```

```python
#仿照实践1，写出由用户指定整数个数，并由用户输入多个整数，并求和的代码。
n=int(input('请输入你想输入数字的个数，以回车结束'))
print('你好',n)

i=0
total=0

while i<n:
    m=int(input('请输入一个整数，以回车结束'))
    i=i+1
    total=total+m
    
print(total)
```

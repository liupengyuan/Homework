```python
#练习 2：写程序，可由键盘读入两个整数m与n(n不等于0)，询问用户意图，如果要求和则计算从m到n的和输出，如果要乘积则计算从m到n的积并输出，如果要求余数则计算m除以n的余数的值并输出，否则则计算m整除n的值并输('输入计算式以#结束')
print('输入计算式以#结束')
p=input('输入计算符')
while p!='#':
    m=int(input('输入第一个整数'))
    n=int(input('输入第二个整数'))
    if p=='+':
        print(m+n)
    elif p=='*':
          print(m*n)
    elif p=='%':
          print(m%n)
    else:
        print(m/n)
    p=input('输入计算符')
```

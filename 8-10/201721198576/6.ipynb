```python
#写函数，返回一个list中所有数字的和
import random
def sum_(a):
    s=0
    for i in a:
        s=s+i
    return s
a=[]
n=int(input('输入list长度：'))
i=0
while(i<n):
    ele=random.randint(0,100)
    a.append(ele)
    i=i+1
print('list为：',a)
print('list中所有数字和为：',sum_(a))
```

```python
#写函数，返回一个list中的最小值
def min_(a):
    m=a[0]
    for i in a[1:]:
        if(i<m):
            m=i
    return m
a=[]
n=int(input('输入list长度：'))
i=0
while(i<n):
    ele=random.randint(0,100)
    a.append(ele)
    i=i+1
print('list为：',a)
print('list中最小值为：',min_(a))
```

```python
#写函数，返回某个元素/对象在一个list中的位置，如果不在，则返回-1.
def find(ele,a):
    if(ele in a):
        return a.index(ele)
    else:
        return -1
a=[]
n=int(input('输入list长度：'))
i=0
while(i<n):
    ele=random.randint(0,100)
    a.append(ele)
    i=i+1
print('list为：',a)
m=int(input('请输入一个元素用来查找：'))
print(m,'在列表中位置：',find(m,a))
```

```python
#写函数，可求两个向量的夹角余弦值，向量可放在list中。主程序调用该函数。
def cos_(m,n):
    if(len(m)!=2 or len(n)!=2):
        print('向量不合法!')
        return False
    mn=m[0]*n[0]+m[1]*n[1]
    len_m=(m[0]**2+m[1]**2)**0.5
    len_n=(n[0]**2+n[1]**2)**0.5
    return (mn/(len_m*len_n))
    
a=[]
i=0
while(i<2):
    x=random.randint(1,20)
    y=random.randint(1,20)
    vector=(x,y)
    a.append(vector)
    i=i+1
print('向量为：',a)    
print('余弦值为：',cos_(a[0],a[1]))
```

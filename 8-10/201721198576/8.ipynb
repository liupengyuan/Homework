```python
#五言诗游戏   成语游戏  
#游戏过程：
#      显示菜单（1游戏说明，2成语游戏，3诗词游戏，4退出游戏）
#      1，2,3结束后都会显示是否继续，
#      继续游戏返回主菜单，否则Game Over
import random
def win():
    print(
        '''
           ======恭喜你，你赢了=======
        
    
                ."".    ."",
                |  |   /  /
                |  |  /  /
                |  | /  /
                |  |/  ;-._ 
                }  ` _/  / ;
                |  /` ) /  /
                | /  /_/\_/\
                |/  /      |
                (  ' \ '-  |
                 \    `.  /
                  |      |
                  |      |
          
          ======恭喜你，你赢了=======
        '''
    )
    return
    
def lose():
    print(
        '''
           ======YOU LOSE=======
        
    
                

                   .-"      "-.
                  /            \
                 |              |
                 |,  .-.  .-.  ,|
                 | )(__/  \__)( |
                 |/     /\     \|
       (@_       (_     ^^     _)
  _     ) \_______\__|IIIIII|__/__________________________
 (_)@8@8{}<________|-\IIIIII/-|___________________________>
        )_/        \          /
       (@           `--------`
       
       
       
          ======YOU LOSE=======
        '''
    )
    return
    

def menu():
    print('------------------菜单-----------------')
    print('   1:游戏说明')
    print('   2:成语游戏')
    print('   3:诗词游戏')
    print('   4:退出游戏')
    i=input('请选择：')
    return int(i)
def If_continue():
    print('是否继续游戏？(1 or 0)')
    i=input('请选择：')
    i=int(i)
    if(i==1):
        main()
    else:
        print('Game Over!')   
def Idiom_Game():
    def get_ch_table(idioms):
        ch_table = []
        for idiom in idioms:
            for ch in idiom:
                if ch not in ch_table:
                    ch_table.append(ch)
        return ch_table

    def idiom_robot(idioms):
    #给出2*3个汉字，真正的成语    
        idiom = random.choice(idioms)
        chs = get_ch_table(idioms)
        guess_ch_table = [ch for ch in idiom]
        while len(guess_ch_table) < 6:
            ch = random.choice(chs)
            if ch not in guess_ch_table:
                guess_ch_table.append(ch)
        random.shuffle(guess_ch_table)
        for i in range(0,6,2):
            print(guess_ch_table[i], guess_ch_table[i+1])
        return idiom
    print('----------------点字成语-----------------')
    f=open(r'D:\temp\idioms.txt','r+',encoding='utf-8')
    idioms=[]
    for i in f.readlines():
        i=i.encode('utf-8').decode('utf-8-sig')
        i=i.replace('\n','')
        i=i.replace('、','')
        if(len(i)==4):
            idioms.append(i)
    f.close()
    score = 10
    i=0
    while score >= 0:
        print('-----第',i+1,'轮游戏-----')
        real_idiom = idiom_robot(idioms)
        answer_idiom = input('请输入猜测成语(#退出本游戏)：')
        if answer_idiom == real_idiom:
            print('答对了，加十分')
            score += 10
            print('你当前的分数是：', score)
            if score == 100:
                win()
                break
        elif answer_idiom == '#':
            print('退出游戏。')
            print('你最后的分数是：', score)
            
            break
        else:
            score -= 10
            print('答错了，减十分')
            print('成语其实是：', real_idiom)
            print('你当前的分数是：', score)
            if(score<0):
                lose()
                break
        i=i+1
def Poetry_Game():
    def get_ch_table(poetries):
        ch_table = []
        for p in poetries:
            for line in p[1:]:
                for ch in line:
                    if(ch not in ch_table):
                        ch_table.append(ch)
        return ch_table
    def poetry_robot(poetries):
        #给出3*3汉字，真正五言诗句
        poetry=random.choice(poetries)
        remind=poetry[0]
        real=random.choice(poetry[1:])
        chs = get_ch_table(poetries)
        guess_ch_table = [ch for ch in real]
        while(len(guess_ch_table)<9):
            ch = random.choice(chs)
            if ch not in guess_ch_table:
                guess_ch_table.append(ch)
        random.shuffle(guess_ch_table)
        for i in range(0,9,3):
            print(guess_ch_table[i], guess_ch_table[i+1],guess_ch_table[i+2])
        return real,remind
    print('------------------点字成诗-----------------')
    f=open(r'D:\temp\poetry.txt','r+',encoding='utf-8')
    poetries=[]
    poetry=[]
    for line in f.readlines():
        line=line.encode('utf-8').decode('utf-8-sig')
        line=line.strip()
        if(any(char.isdigit() for char in line)):
            if(poetry!=[]):
                poetries.append(poetry)
            poetry=[]
            poetry.append(line)
        else: 
            if(line!=''):
                if(poetry!=[]):
                    line1=line[:5]
                    line2=line[6:11]
                    poetry.append(line1)
                    poetry.append(line2)
    f.close()
    score=10
    i=0
    while(score>=0):
        print('-----第',i+1,'轮游戏-----')
        real,remind=poetry_robot(poetries)
        print('提示：',remind)
        answer=input('请输入猜测诗句(#退出本游戏)：')
        if answer == real:
            print('答对了，加十分')
            score += 10
            print('你当前的分数是：', score)
            if score == 100:
                win()
                break
                
        elif answer == '#':
            print('退出游戏。')
            print('你最后的分数是：', score)
            break
        else:
            score -= 10
            print('答错了，减十分')
            print('成语其实是：', real)
            print('你当前的分数是：', score)
            if(score<0):
                lose()
                break
        i=i+1
def main():
    i=menu()
    if (i==1):
        print('------------------游戏说明-----------------')
        print('  初始分值10分，每答对一次加10分，答错扣10分，')
        print('  分值小于0    失败!')
        print('  分值达到100  成功')
        print('  输入#       退出当前游戏')
        If_continue()
    elif(i==2):
        Idiom_Game()
        If_continue()
    elif(i==3):
        Poetry_Game()
        If_continue()
    else:
        print('Game Over!')

if __name__ == '__main__':
    main()
```

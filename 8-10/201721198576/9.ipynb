```python
#2:随机生成1万个整数，范围在0-10万之间，
#求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
import random
def creat_random():
    i=0
    numbers=[]
    while(i<10000):
        numbers.append(random.randint(0,100000))
        i=i+1
    return numbers
def Count_(numbers):
    nums_dict={}
    for n in numbers:
        if(n in nums_dict):
            nums_dict[n]=nums_dict[n]+1
        else:
            nums_dict[n]=1
    return nums_dict
def main():
    Dict_=Count_(creat_random())
    for n in sorted(Dict_.keys()):
        print(n,'出现次数',Dict_[n])
if __name__ =='__main__':
    main()
```

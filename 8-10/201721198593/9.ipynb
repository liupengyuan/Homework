## 随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。

```python
import random

def sort_simple_selection(seq):
    for i in range(len(seq)-1):
        min = i
        for j in range(i+1, len(seq)):
            if seq[j] < seq[min]:
                min = j
        seq[i], seq[min] = seq[min], seq[i]

a = [random.randint(0, 100000) for i in range(10000)]
num_freq_pairs = []
sort_simple_selection(a)
for i in a:
    number = a.count(i)
    num_freq_pairs.append([i, number])
for d in num_freq_pairs:
    print(d)
```

### 对本任务中的语料.txt文件，随机抽取其5001-10000行存为test1.txt文件，写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。

```python
with open(r'F:\Working Tool\NO.1\Python\Anaconda jupyter workspace\Python课程练习\11-29 homework\test.txt', 'w') as fw, open(r'F:\source\语料\语料.txt') as fr,open(r'F:\Working Tool\NO.1\Python\Anaconda jupyter workspace\Python课程练习\11-29 homework\test1.txt', 'w') as fv:
    line_number = 0
    for line in fr:
        if line_number < 5000:
            fw.write(line)
            line_number += 1
        elif line_number < 10000 and line_number >= 5000:
            fv.write(line)
            line_number += 1
        else:
            break
            
def get_word_table(filename):
    word_table = []
    with open(filename) as f:
        text = f.read()
    words = [word.split('/')[0] for word in text.split()]
    for word in words:
        if word not in word_table:
            word_table.append(word)
    return word_table

def get_d_g(t1,t2):
    num_g_freq_pairs = []
    num_d_freq_pairs = []
    for i in t2:
        if i not in t1:
            num_g_freq_pairs.append([i])
        else:
            num_d_freq_pairs.append([i])
    print('共用字有：')
    for d in num_g_freq_pairs:
        print(d)
    print('独用字有：')
    for f in num_g_freq_pairs:
        print(f)

filename1 = r'F:\Working Tool\NO.1\Python\Anaconda jupyter workspace\Python课程练习\11-29 homework\test.txt'
filename2 = r'F:\Working Tool\NO.1\Python\Anaconda jupyter workspace\Python课程练习\11-29 homework\test1.txt'
table1 = get_word_table(filename1)
table2 = get_word_table(filename2)
get_d_g(table1,table2)
```

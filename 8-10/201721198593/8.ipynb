#### 下面是点字成诗机器人

```python
import re

def menu():
    print('\n=====游戏菜单=====')
    print('1. 游戏说明')
    print('2. 开始游戏')
    print('3. 退出游戏')
    print('4. 制作团队')
    print('=====游戏菜单=====')

def show_instruction():
    print('\n=====本游戏为点字成诗游戏=====')
    print('1.输入\'2\'开始游戏')
    print('2.输入任意超过五个你所想输入的汉字')
    print('3.机器人会根据你所给出的汉字输出五言诗词！')
    print('====让我们赶快开始游戏吧！====')
    
def find_poem_sentence(poems, characters):
    same_character_number = 0
    for poem in poems:
        for ch in poem:
            if ch in characters:
                same_character_number += 1
        if same_character_number == 5:
            return poem
        same_character_number = 0
    return None

def inputwork():
    fh = open(r'五言诗库.txt')
    poems = re.split('。\n|，',fh.read())
    fh.close()
    return poems

def play():
    characters = list(input('请输入至少五个汉字:  \n'))
    poems = inputwork()
    result = find_poem_sentence(poems, characters)
    print('机器人给出的结果是：  \n')
    print(result)
    
def main():
    while True:
        menu()
        choice = int(input('\n请输入你的选择'))
        if choice == 1:
            show_instruction()
        elif choice == 2:
            play()
        elif choice == 3:
            game_over()
            break
        else:
            show_team()
            
def show_team():
    print('\n=====本游戏由郭铭昊倾情奉献！===')
    
main()
```

#### 下面是五言诗游戏

```python
import re
import random

def menu():
    print('\n=====游戏菜单=====')
    print('1. 游戏说明')
    print('2. 开始游戏')
    print('3. 退出游戏')
    print('4. 制作团队')
    print('=====游戏菜单=====')

def show_instruction():
    print('\n=====本游戏为五言诗游戏=====')
    print('1.输入\'2\'开始游戏')
    print('2.根据系统给你的字写出你认为对的诗句')
    print('3.机器人会根据你所给出的诗句来判断对错')
    print('====让我们赶快开始游戏吧！====')
    
def find_poem_sentence(poem1, characters):
    #转成字符串格式
    characters = ''.join(characters)
    for poem in poem1:
        if characters == poem:
            return "回答正确："+ poem
    return "对不起，回答错误"

def inputwork():
    fh = open(r'五言诗库.txt')
    poems = re.split('。\n|，',fh.read())
    fh.close()
    return poems

def play():
    word="";
    poems = inputwork()
    #随机选一首诗
    poem1 = random.sample(poems, 4)
    #下面这句是答案提示
    print(poem1)
    for poem in poem1:
        word = word+poem
    words = list(word)
    word_new = random.sample(words,18)
    
    print('下面是系统给出的18个字：\n')
    for ch in word_new: 
        print(ch)
    
    characters = list(input('\n请输入你认为正确的诗句:  \n'))
    result = find_poem_sentence(poem1, characters)
    print('机器人给出的结果是：  \n')
    print(result)
    
def main():
    while True:
        menu()
        choice = int(input('\n请输入你的选择'))
        if choice == 1:
            show_instruction()
        elif choice == 2:
            play()
        elif choice == 3:
            game_over()
            break
        else:
            show_team()
            
def show_team():
    print('\n=====本游戏由郭铭昊倾情奉献！===')
    
main()
```

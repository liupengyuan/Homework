```python
#2、随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
import numpy as np
from collections import Counter
from operator import itemgetter

nums=np.random.randint(0,100000,size=10000)

c = Counter(nums)
a=c.items()
b=sorted(a,key=itemgetter(0),reverse=True)

print(b)
```

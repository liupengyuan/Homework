```python
# 9-2：随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数

import random

list = []
numDict = {}

while len(list) < 10001:
    num = random.randint(0,100000)
    list.append(num)
print('随机列表：',list)  

lst = sorted(list)
for number in lst:
    count = 0
    count = lst.count(number)
    numDict[number] = count
print('整数出现次数：',numDict)
```

3、对本任务中的语料.txt文件，随机抽取其5001-10000行存为test1.txt文件，写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。

```python
with open(r'f:\temp\test.txt','w',encoding='utf-8') as fw1,open(r'f:\temp\test1.txt','w',encoding='utf-8') as fw2,open(r'f:\temp\语料.txt','r') as fr:
    line_number = 0
    for line in fr:
        if line_number <= 5000:
            fw1.write(line)
            line_number += 1
        elif line_number>5000 and line_number<10000:
            fw2.write(line)
            line_number += 1
        else:
            break
            
def getWord(filename):
    wordTable=[]
    with open(filename,'r',encoding='utf-8') as f:
        for line in f:
            table=[word.split('/')[0] for word in line.split()]
            char=''.join(table)
            for item in char:
                if item not in wordTable:
                    wordTable.append(item)
    return wordTable

def sharedWord(table1,table2):
    sharedWordList = []
    for word in table1:
        if word in table2:
            sharedWordList.append(word)
    return sharedWordList    


def indepenWord(table1,table2):
    indepenWordList = []
    for word in table1:
        if word not in table2:
            indepenWordList.append(word)
    return indepenWordList    

def main():
    filename1 = r'f:\temp\test.txt'
    filename2 = r'f:\temp\test1.txt'
    table1 = getWord(filename1)
    table2 = getWord(filename2)
    indepenWordList1 = indepenWord(table1,table2)
    indepenWordList2 = indepenWord(table2,table1)
    print(indepenWordList1)
    print(indepenWordList2)
    sharedWordList = sharedWord(table2,table1)
    print(sharedWordList)
if __name__ == '__main__':
    main()
    
```

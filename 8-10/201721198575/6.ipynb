# 2.写函数，返回一个list中所有数字的和

```python
def total(param):
    total = 0
    for number in param:
        total += number
    return total

print('和为', total([1, 2, 3, 4, 5, 6, 7]))
```

# 3.写函数，返回一个list中的最小值

```python
def minNum(list):
    minNum = list[0]
    for number in list:
        if number < minNum:
            minNum = number
    return minNum
    
print('最小值为', minNum([1, 2, 3, 4, 5, 6, 7]))
```

# 5.写函数，可求两个向量的夹角余弦值，向量可放在list中。主程序调用该函数。

```python
import math
def cos(a,b):
    total_a = 0
    total_b = 0
    total = 0
    i = j = 0
    for x in a:
        total_a = total_a + x * x
        for y in b:
            total_b = total_b + y * y
            if i == j:
                total = total + x * y
            j = j + 1
        i = i + 1
    return total/( math.sqrt(total_a)*math.sqrt(total_b) )
            

vector=[[2,3,5],[3,3,3]]
print('两个向量的余弦值为：',cos(vector[0],vector[1]))
```

# 挑战性习题：python语言老师为了激励学生学python，自费买了100个完全相同的Macbook Pro，分给三个班级，每个班级至少分5个，用穷举法计算共有多少种分法？

```python
total = 100-5-5-5
count = 0
for i in range(total + 1):
    for j in range(total - i + 1):
        for k in range(total - i - j + 1):
            if i + j + k == 85:
                count = count + 1
            else:
                continue
print('一共',count,'种分法')
```

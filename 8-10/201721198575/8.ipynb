# 题目：五言诗游戏和点字成诗机器人。
* 要求：给出“五言诗库”（大约100首），仿照教程中成语游戏的模式完成五言诗游戏和点字成诗的机器人。有能力的同学根据“唐诗三百首”诗库进行游戏设计。

```python
# 点字成语游戏完整代码

import random

def win():
    print('恭喜你成功了')
    return
    
def lose():
    print('很遗憾，你失败了，再接再励，下次加油')
    return
    
def menu():
    print('========游戏菜单=========')
    print('   1:游戏说明')
    print('   2:成语游戏')
    print('   3:诗词游戏')
    print('   4:退出游戏')
    i=input('请选择：')
    return int(i)
def If_continue():
    print('是否继续游戏？(是 or 否)')
    i=input('请选择：')
    if(i=='是'):
        main()
    else:
        print('Game Over!')   
def Idiom_Game():
    def get_hz_table(idioms):
        hz_table = []
        for idiom in idioms:
            for hz in idiom:
                if hz not in hz_table:
                    hz_table.append(hz)
        return hz_table

    def idiom_robot(idioms):
    #给出2*4个汉字   
        idiom = random.choice(idioms)
        hzs = get_hz_table(idioms)
        guess_hz_table = [hz for hz in idiom]
        while len(guess_hz_table) < 8:
            hz = random.choice(hzs)
            if hz not in guess_hz_table:
                guess_hz_table.append(hz)
        random.shuffle(guess_hz_table)
        for i in range(0,8,2):
            print(guess_hz_table[i], guess_hz_table[i+1])
        return idiom
    print('========点字成语========')
    f=open(r'c:\LEISHUTING\class\Python\idioms.txt','r+',encoding='utf-8')
    idioms=[]
    for i in f.readlines():
        i=i.encode('utf-8').decode('utf-8-sig')
        i=i.replace('\n','')
        i=i.replace('、','')
        if(len(i)==4):
            idioms.append(i)
    f.close()
    score = 10
    i=0
    while score >= 0:
        print('========第',i+1,'轮游戏========')
        real_idiom = idiom_robot(idioms)
        answer_idiom = input('请输入猜测成语(q退出本游戏)：')
        if answer_idiom == real_idiom:
            print('答对了，加十分')
            score += 10
            print('你当前的分数是：', score)
            if score == 100:
                win()
                break
        elif answer_idiom == 'q':
            print('退出游戏。')
            print('你最后的分数是：', score)
            
            break
        else:
            score -= 10
            print('答错了，减十分')
            print('成语其实是：', real_idiom)
            print('你当前的分数是：', score)
            if(score<0):
                lose()
                break
        i=i+1
def Poetry_Game():
    def get_hz_table(poetries):
        hz_table = []
        for p in poetries:
            for line in p[1:]:
                for ch in line:
                    if(ch not in hz_table):
                        hz_table.append(hz)
        return hz_table
    def poetry_robot(poetries):
        #给出3*3汉字
        poetry=random.choice(poetries)
        remind=poetry[0]
        real=random.choice(poetry[1:])
        hzs = get_hz_table(poetries)
        guess_hz_table = [hz for hz in real]
        while(len(guess_hz_table)<9):
            hz = random.choice(hzs)
            if hz not in guess_hz_table:
                guess_hz_table.append(hz)
        random.shuffle(guess_hz_table)
        for i in range(0,9,3):
            print(guess_hz_table[i], guess_hz_table[i+1],guess_hz_table[i+2])
        return real,remind
    print('========点字成诗========')
    f=open(r'c:\LEISHUTING\class\Python\poetry.txt','r+',encoding='utf-8')
    poetries=[]
    poetry=[]
    for line in f.readlines():
        line=line.encode('utf-8').decode('utf-8-sig')
        line=line.strip()#s.strip(rm)rm为空则删除空\n\r\t' '
        if(line!=''):
            if(poetry!=[]):
                poetries.append(poetry)
            poetry=[]
            line1=line[:5]
            line2=line[6:11]
            poetry.append(line1)
            poetry.append(line2)
    f.close()
    score=10
    i=0
    while(score>=0):
        print('========第',i+1,'轮游戏========')
        real,remind=poetry_robot(poetries)
        print('提示：',remind)
        answer=input('请输入猜测诗句(q退出本游戏)：')
        if answer == real:
            print('答对了，加十分')
            score += 10
            print('你当前的分数是：', score)
            if score == 100:
                win()
                break
                
        elif answer == 'q':
            print('退出游戏。')
            print('你最后的分数是：', score)
            break
        else:
            score -= 10
            print('答错了，减十分')
            print('诗句其实是：', real)
            print('你当前的分数是：', score)
            if(score<0):
                lose()
                break
        i=i+1
def main():
    i=menu()
    if (i==1):
        print('========游戏说明========')
        print('  初始分值10分，每答对一次加10分，答错扣10分，')
        print('  分值小于0    失败!')
        print('  分值达到100  成功')
        print('  输入q        退出当前游戏')
        If_continue()
    elif(i==2):
        Idiom_Game()
        If_continue()
    elif(i==3):
        Poetry_Game()
        If_continue()
    else:
        print('Game Over!')

if __name__ == '__main__':
    main()
```

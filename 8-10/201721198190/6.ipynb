#写函数，返回一个list中所有数字的和++


```python
def sum_num(n):
    number=[]
    i=0
    total=0
    while i<n:
        num=int(input('请输入数字：'))
        number.append(num)
        total += number[i] 
        print(number)
        i += 1
    return total
   
n=int(input('请输入数字个数：'))
print('所有数字的和：',sum_num(n))
```

#写函数，返回一个list中的最小值

```python
def min_num(n):
    number=[]
    i=0
    while i<n:
        num=int(input('请输入数字：'))
        number.append(num)
        print(number)  
        i +=1
    m=min(number)
    return m
n=int(input('请输入数字个数：'))
print('最小值：',min_num(n))
```

#写函数，返回某个元素/对象在一个list中的位置，如果不在，则返回-1.

```python
def get_index(object,list):
    list=[1,2,3,4]
    for i in range(len(list)): 
        if object == list[i]:
            return i
        if object not in list:
            return -1
object=int(input('请输入数字：'))
print('对象的位置：',get_index(object,list))
        
```


#写函数，可求两个向量的夹角余弦值，向量可放在list中。主程序调用该函数。

```python
import math
def cos_function(v1,v2):
    mul = 0
    for m in v1:
        for n in v2:
           mul += m*n
    v1_a = math.sqrt(sum([m ** 2 for m in v1]))
    v2_a = math.sqrt(sum([m ** 2 for m in v2]))
    return mul / (v1_a * v2_a)
vector1 = [1,2,4]
vector2 = [3,4,5]
print("余弦值为：",cos_function(vector1,vector2))
```

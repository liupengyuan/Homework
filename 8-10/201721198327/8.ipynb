```python
#五言诗游戏

import random

def win():
    print(
        '''
           ======恭喜你，你赢了=======
        
    
                ."".    ."",
                |  |   /  /
                |  |  /  /
                |  | /  /
                |  |/  ;-._ 
                }  ` _/  / ;
                |  /` ) /  /
                | /  /_/\_/\
                |/  /      |
                (  ' \ '-  |
                 \    `.  /
                  |      |
                  |      |
          
          ======恭喜你，你赢了=======
        '''
    )
    
def lose():

    print(
        '''
           ======YOU LOSE=======
        
    
                

                   .-"      "-.
                  /            \
                 |              |
                 |,  .-.  .-.  ,|
                 | )(__/  \__)( |
                 |/     /\     \|
       (@_       (_     ^^     _)
  _     ) \_______\__|IIIIII|__/__________________________
 (_)@8@8{}<________|-\IIIIII/-|___________________________>
        )_/        \          /
       (@           `--------`
       
       
       
          ======YOU LOSE=======
        '''
    )
    
def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table

def idiom_robot(file_name):
    with open(file_name) as fh:
        text = fh.read()
    idioms = text.split()
    idiom = random.choice(idioms)
    chs = get_ch_table(text.replace('\n', ''))

    guess_ch_table = [ch for ch in idiom]
    while len(guess_ch_table) < 9:
        ch = random.choice(chs)
        if ch not in guess_ch_table:
            guess_ch_table.append(ch)
    
    random.shuffle(guess_ch_table)
    
    for i in range(0,8,3):
        print(guess_ch_table[i], guess_ch_table[i+1], guess_ch_table[i+2])
    
    return idiom

def main():
    filename = r'Z:\2硕士\Python统计分析及可视化\资料\出题库.txt'
    print('欢迎来到五言诗答题大赛，您现在有40分的基础分，答对一道题得10分，答错一道题扣10分，不答题直接退出。下面开始答题！')
    score = 40
    while score >= 0:
        real_idiom = idiom_robot(filename)
        answer_idiom = input('请输入猜测诗句，回车结束，直接回车结束答题：')
        
        if answer_idiom == real_idiom:
            print('答对了，加十分')
            score += 10
            print('你当前的分数是：', score)
            
            if score == 100:
                win()

        elif answer_idiom == '':
            print('退出游戏。')
            print('你最后的分数是：', score)
            break

        else:
            score -= 10
            print('答错了，减十分')
            print('诗句其实是：', real_idiom)
            print('你当前的分数是：', score)
    else:
        lose()

main()
```

```python
#点字成诗机器人

import itertools
characters = input('请输入一个让机器人猜测的包含一个打乱的五言诗句的句子：')
guesses = itertools.permutations(characters, 5)

fh = open(r'Z:\2硕士\Python统计分析及可视化\资料\出题库.txt')
poems = fh.read().split()
fh.close()

for guess in guesses:
    if ''.join(guess) in poems:
        print('答案是：' ,guess)
```

```python
#练习二 随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
import random

n = 0
freq=1
num_table = []
num_freq_dict = {}
while n < 10000:
    num_table.append(random.randint(0,100000))
    n = n + 1
num_table.sort()
for num in num_table:
    if num in num_freq_dict:
        num_freq_dict[num] += 1
    else:
        num_freq_dict[num] = 1
    
print(num_freq_dict)
```

```python
#练习三 对本任务中的语料.txt文件，随机抽取其5001-10000行存为test1.txt文件，写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。
with open(r'd:\python\test1.txt', 'w') as fw, open(r'd:\python\语料.txt') as fr:
    line_number = 0
    for line in fr:
        if line_number in range(5001,10001):
            fw.write(line)
            line_number += 1
        else:
            break
with open(r'd:\python\test.txt', 'w') as a, open(r'd:\python\语料.txt') as b:
    line_number = 0
    for line in b:
        if line_number < 5000:
            a.write(line)
            line_number += 1
        else:
            break           

def get_word_table(filename):    
    word_table = []
    with open(filename) as f:
        text = f.read()
    words = [word.split('/')[0] for word in text.split()]
    for word in words:
        if word not in word_table:
            word_table.append(word)
    return word_table

def public_private(filename1,filename2):  
    word_list1 = get_word_table(filename1)
    word_list2 = get_word_table(filename2)
    public = []
    private = []
    
    for item1 in word_list1:
        flag = 0
        for item2 in word_list2:
            if item1 == item2:
                public.append(item1)
                flag = flag + 1
                break 
        if flag == 0:
            private.append(item1)
            
    print('共用字：',public)
    print('--------------------------------------------------------------------------------------------------------')
    print('独用字：',private)
        
filename1 = r'd:\python\test1.txt'
filename2 = r'd:\python\test.txt'
public_private(filename1,filename2)
```

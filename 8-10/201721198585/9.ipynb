# 2、随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。

```python
import random
from collections import Counter

nums_freq_dict={}
for i in range(0,10000):
    num = random.randint(0, 100001)
    if num in nums_freq_dict: 
        nums_freq_dict[num] += 1
    else:   
        nums_freq_dict[num] = 1

num_list=sorted(nums_freq_dict.items(),key=lambda e:e[1],reverse=True)
for term in num_list:
    print(term)
```

# 3、对本任务中的语料.txt文件，随机抽取其5001-10000行存为test1.txt文件，写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。

```python
import random

with open(r'd:\temp\test.txt', 'w') as fw, open(r'd:\temp\语料.txt') as fr:
    line_number = 0
    for line in fr:
        if line_number < 5000:
            fw.write(line)
            line_number += 1
        else:
            break
            

num_list = range(5001,10001)
line_list = random.sample(num_list, 2000)

with open(r'd:\temp\test1.txt', 'w') as fw, open(r'd:\temp\语料.txt') as fr:
    line_number = 0
    for line in fr:
        if line_number < 10000:
            if line_number in line_list: 
                fw.write(line)       
            line_number += 1
        else:
            break
```

```python
def char_dict(file1,file2):
    test_set=set()
    test1_set=set()
    with open(file1) as f1:
        for line in f1:
            words = [word.split('/')[0] for word in line.split()]
            for word in words:
                for char_h in word: 
                    test_set.add(char_h)
        
    with open(file2) as f2:
        for line in f2:
            words = [word.split('/')[0] for word in line.split()]
            for word in words:
                for char_h in word: 
                    test1_set.add(char_h)
                    
    common = test_set & test1_set
    solo_test = test_set - common
    solo_test1 = test1_set - common
    
    with open(r'd:\temp\common.txt', 'w') as fc:
        for item in common:
            fc.write(item)
            
    with open(r'd:\temp\solo_test.txt', 'w') as f1:
        for item in solo_test:
            f1.write(item)

    with open(r'd:\temp\solo_test1.txt', 'w') as f2:
        for item in solo_test1:
            f2.write(item)
            
file=r'd:\temp\test.txt'
file1=r'd:\temp\test1.txt'
char_dict(file,file1)
```

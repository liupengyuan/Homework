```python
import random

def menu():
    print('''
=====游戏菜单=====
1. 游戏说明
2. 诗词游戏
3. 退出游戏
4. 制作团队
=====游戏菜单=====''')

def show_instruction():
    print('''
          游戏说明 
诗词游戏：
根据所给汉字串，猜出其中隐藏的诗句。
       ''')

def show_team():
    print('copyright © 2009-2014 MLY。All Rights Reserved.')

def win():
    print('You Win!')
    
def lose():
    print('Game Over!')
    
def idiom_list(file_name):
    with open(file_name) as fh:
        text = fh.read()
    idioms_tmp = text.split('\n')
    fh.close()
    return idioms_tmp

def get_ch_table(file_name):
    with open(file_name) as fh:
        text = fh.read()
    text = text.replace(r'\n', '')
    text = text.replace(r'，', '')
    text = text.replace(r'。', '')
    ch_table = []
    for ch in text:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table

def poem_list(file_name):
    with open(file_name) as fh:
        text = fh.read()
    text = text.replace(r'，', '\n')
    text = text.replace(r'。', '')
    poems_tmp = text.split('\n')
    fh.close()
    return poems_tmp

def robot_poem(poems,characters):
    for poem in poems:
        same_character_number = 0
        for ch in poem:
            if ch in characters:
                same_character_number += 1
        if same_character_number == 5:
            print('机器人的答案：',poem)
            return poem
    return None

def poem_game():
    file_name = r'd:\temp\五言诗库.txt'
    poemlist = poem_list(file_name)
    poems = poem_list(file_name)
    chs = get_ch_table(file_name)
    guessed = []
    score = 10
    while score >= 0:
        real_poem = random.choice(poems)
        while real_poem in guessed:
            real_poem = random.choice(poems)
        
        guess_ch_table = [ch for ch in real_poem]
        while len(guess_ch_table) < 9:
            ch = random.choice(chs)
            if ch not in guess_ch_table:
                guess_ch_table.append(ch)
    
        random.shuffle(guess_ch_table)
        for i in range(0,8,3):
            print(guess_ch_table[i], guess_ch_table[i+1],guess_ch_table[i+2])        
        
        print('''
        请输入猜测诗句，回车结束。请机器人答题请按1。
        ''')
        answer_poem = input()
        if answer_poem=='1':
            answer_poem = robot_poem(poemlist,guess_ch_table)
            
        if answer_poem == real_poem:
            print('答对了，加十分')
            score += 10
            guessed.append(real_poem)
            poems.remove(real_poem)
            print('你当前的分数是：', score)
            if score == 50:
                win()
                return
        elif answer_poem == '':
            print('退出游戏。')
            print('你最后的分数是：', score)
            return
        else:
            score -= 10
            guessed.append(real_poem)
            print('答错了，减十分')
            print('诗句是：', real_poem)
            print('你当前的分数是：', score)
    else:
        lose()
        return

def lose():
    print('Game Over!')

def win():
    print('You Win!')

    
def main():
    while True:
        menu()
        choice = int(input('你的选择'))
        if choice == 1:
            show_instruction()
        elif choice == 2:
            print('''
            游戏开始
            
            ''')
            poem_game()
        elif choice == 3:
            print('退出游戏! Bye!')
            return
        elif choice == 4:
            show_team()
        else:
            print('请选择：')
            
main()
```

```python
#练习二：随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。

import random
n = random.sample(range(0, 100000), 10000)




def count_freq(filename):
    freq_pairs = []
    with open(filename) as f:
        text = f.read()
        words = [word.split('/')[0] for word in text.split()]
        for word in words:
                for item in freq_pairs:
                    if word == item[0]:
                        item[1] += 1
                        break
                else:
                    freq_pairs.append([word, 1])
                    
    return freq_pairs





filename = (r'D:\temp\randint.txt') 
table = count_freq(filename)
sorted(table[:100000])

    
    
```

```python
#写函数，返回一个list中所有数字的和
numbers=[]
n=int(input("请输入1个整数表示序列中元素的个数，回车以结束。"))
i=0
while i < n:
    number=int(input("请输入一个整数,回车以结束。"))
    numbers.append(number)
    i=i+1
print(numbers)

total=0
i=0
while i < n:
    total=total+numbers[i]
    i=i+1
print(total)
```

```python
#写函数，可求两个向量的夹角余弦值，向量可放在list中。主程序调用该函数
def cosVector(x,y):
    if(len(x)!=len(y)):
        print('error input,x and y is not in the same space')
        return
    result1=0
    result2=0
    result3=0
    for i in range(len(x)):
        result1+=x[i]*y[i]
        result2+=x[i]**2     
        result3+=y[i]**2
    print("result is "+str(result1/((result2*result3)**0.5))) 
    
cosVector([3,1],[2,1])
```

```python
#写函数，返回一个list中的最小值
numbers=[]
n=int(input("请输入1个整数表示序列中元素的个数，回车以结束。"))
i=0
while i < n:
    number=int(input("请输入一个整数,回车以结束。"))
    numbers.append(number)
    i=i+1
print(numbers)

def min_numbers():
    min=numbers[0]
    for i in numbers:
        if i<= min:
            min=i
    return min
print(min_numbers())
```

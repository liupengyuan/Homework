```python
# -*- coding: utf-8 -*-
# 李华勇 Python作业 
# 程序说明：本程序把五言诗游戏和机器人点字成诗合并到了一起。
#           五言诗游戏是机器出题，人输入回答，当分数小于0时退出游戏。
#           机器人点字成诗是机器出题，机器人猜测答案

import random


def init_poem_list(file_path):
    """
    参数：file_path是五言诗库.txt的保存位置
    返回：程序返回一个诗句列表
    """
    poem_list = []
    try:
        with open(file_path) as f:
            for line in f.readlines():
                line = line.replace('\n', '')
                line = line.replace('。', '')
                for poem_ in line.split('，'):
                    poem_list.append(poem_)
    except Exception as e:
        print("程序运行错误：找不到诗库文件！！")
        print("诗库文件必须和本程序在同一目录下！！否则程序无法运行。")
        print(e)
    else:
        return poem_list


def init_ch_table(poem_list):
    """
        参数：poem_list是五言诗句列表
        返回：程序返回一个字表
    """
    ch_table_ = []
    for poem_ in poem_list:
        for ch in poem_:
            if ch not in ch_table_:
                ch_table_.append(ch)
    return ch_table_


def get_guess_poem(poem_, ch_table_, length_):
    """
        参数：
        返回：程序返回一个乱序9字
        """
    guess_poem_ = [ch for ch in poem_]
    while len(guess_poem_) < length_:
        ch = random.choice(ch_table_)
        if ch not in guess_poem_:
            guess_poem_.append(ch)
    random.shuffle(guess_poem_)
    return guess_poem_


def poem_game():
    score = 10
    while score >= 0:
        poem = random.choice(poemList)
        guess_poem_list = get_guess_poem(poem, ch_table, 9)
        for i in range(0, 9, 3):
            print(guess_poem_list[i], guess_poem_list[i + 1], guess_poem_list[i + 2])
        answer = input('请输入猜测诗句，回车结束，直接回车表示退出游戏：')
        if answer == poem:
            print('答对了，加十分')
            score += 10
            print('你当前的分数是：', score)
            if score == 100:
                print("你赢了！！")
                break
        elif answer == '':
            print('退出游戏。')
            print('你最后的分数是：', score)
            break
        else:
            score -= 10
            print('答错了，减十分')
            print('诗句其实是：', poem)
            print('你当前的分数是：', score)
    else:
        print("你输了！！")


def find_poem_sentence(poems, characters, length_):
    for poem_ in poems:
        same_character_number = 0
        for ch in poem_:
            if ch in characters:
                same_character_number += 1
        if same_character_number == length_:
            return poem_
    return None


def poem_robot():
    # 出题：
    poem = random.choice(poemList)
    guess_poem_list = get_guess_poem(poem, ch_table, 9)
    print("题目是：")
    for i in range(0, 9, 3):
        print(guess_poem_list[i], guess_poem_list[i + 1], guess_poem_list[i + 2])
    print("机器人的猜测是：")
    result = find_poem_sentence(poemList, guess_poem_list, len(poem))
    if not result:
        print("抱歉，机器人找不到答案~~~")
        print("正确答案是：", poem)
    else:
        print(result)
        print("正确答案是：", poem)
        if result == poem:
            print("机器人答对了！！好厉害")
        else:
            print("啊啊啊，机器人也会出错？！")


# 以下是主程序
path = '五言诗库.txt'
poemList = init_poem_list(path)
ch_table = init_ch_table(poemList)
print(""""程序说明：本程序把五言诗游戏和机器人点字成诗合并到了一起。
              五言诗游戏是机器出题，人输入回答，当分数小于0时退出游戏。
              机器人点字成诗是机器出题，机器人猜测答案""")
while True:
    print("*" * 60)
    print("游戏选项：\n输入1：五言诗游戏\n输入2：机器人点字成诗\n输入其他按键：退出游戏")
    option = input("你的选择是：")
    if option == '1':
        # 五言诗游戏
        poem_game()
        continue
    if option == '2':
        # 机器人猜测诗句
        poem_robot()
        continue
    else:
        print("再见~")
        break
```

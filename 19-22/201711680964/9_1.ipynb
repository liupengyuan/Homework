```python
#生成列表A[1,2,3......100000]
#生成列表B[A的乱序]
#随机生成1000个1-110000之间的整数：
#用顺序查找算法在B中查找上述整数
#用二分查找算法在A中找上述整数
#并分别计算两者总时间
```

```python
import time,random,bisect
start_time = time.time()
A=[]
for a in range(1,100001):
    A.append(a)
B=A.copy()
random.shuffle(B)
N=random.sample(range(1,110001),k=1000)
#用顺序查找算法在B中查找上述整数
#并分别计算两者总时间
start_timeB = time.time()
N1=[]
for y in N:
    X=[]
    X.append(y)
    if y in B:
        m=B.index(y)
        X.append(m)
    else:
        X.append('none')
    N1.append(X)
    end_timeB=time.time()
print('''顺序查找：元素和位置分别为

{}





顺序查找经过{}秒




'''.format(N1,end_timeB-start_timeB))
#用二分查找算法在A中找上述整数
#并分别计算两者总时间
start_timeA = time.time()
N2=[]
for x in N:
    X=[]
    n=bisect.bisect_left(A, x)
    X.append(x)
    if n==0 and A[0]==x:
        X.append(n)
    elif n==len(A):
        X.append('none')
    else:
        X.append(n)
    N2.append(X)
    end_timeA=time.time()
print('''二分查找：元素和位置分别为

{}





二分查找经过{}秒




'''.format(N2,end_timeA-start_timeA))

```

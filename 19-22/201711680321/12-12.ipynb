```python
def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table

fh = open(r'd:\temp\idioms_correct.txt')
text = fh.read()
chs = get_ch_table(text.replace('\n', ''))
```

```python
text = text.replace('、', '')
idioms = text.split('\n')
```

```python
import random

for i in range(1,11):
    s=[]
    for j in range(10*i+1):
        idiom = random.choice(chs)
        print(idiom)
        s.append(idiom)
    print(j)
fh = open(r'c:\\temp\\'+str(i)+'.txt','w')
fh.write(''.join(s))
fh.close()
    
```

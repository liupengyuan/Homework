```python
from collections import defaultdict 
with open(r'd:\temp\test1.txt', 'w') as fw, open(r'd:\temp\语料.txt') as fr: 
    line_number = 5000
    for line in fr:
        if line_number<10001:
            fw.write(line) 
            line_number += 1 
        else: 
            break 
def pick_words(filename): 
    letter=set() 
    with open(filename) as f: 
        text = f.read() 
    words = [word.split('/')[0] for word in text.split()] 
    for word in words: 
        for ch in word: 
            if ch not in letter: 
                letter |= {ch} 
    return letter 
def compare_words(m,n): 
    a = m & n 
    b = m.symmetric_difference(n) 
    print('共用字为：',a) 
    print('*'*50,'分割线','*'*50) 
    print('独用字为：',b) 
    
letter1=pick_words(r'd:\temp\test1.txt') 
letter2=pick_words(r'd:\temp\test.txt' ) 
compare_words(letter1,letter2) 
```

```python
from collections import defaultdict 
with open(r'd:\temp\语料.txt') as fh: 
    text = fh.read() 
words = [word.split('/') for word in text.split()] 
table=[word[0] for word in words] 
words_freq_dict = defaultdict(int)  
for word in table: 
    words_freq_dict[word] += 1 
print(words_freq_dict) 
```

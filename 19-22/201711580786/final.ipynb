```python
#2
def Ackermann(m,n):
    if m==0:
        A=n+1
    elif m>0 and n==0:
        return Ackermann(m-1,1)
    elif m>0 and n>0:
        return Ackermann(m-1,Ackermann(m,n-1))
    return A
m=int(input())
n=int(input())
print(Ackermann(m,n))
```

```python
# 3(1)(2)
import random
a=int(input("下限"))
b=int(input("上限"))
n=int(input("题数"))
A=[]
B=[]
C=[]
score=0
for i in range(0,n):
    x=random.randint(a,b)
    y=random.randint(a,b)
    q=random.randint(1,4)
    if q==1:
        print(x,"+",y,"=")
        p=x+y
        A.append(p)
        B.append(5)
    if q==2:
        print(x,"-",y,"=")
        p=x-y
        A.append(p)
        B.append(5)
    if q==3:
        print(x,"*",y,"=")
        p=x*y
        A.append(p)
        B.append(10)
    if q==4:
        print(x,"/",y,"=")
        p=x/y
        A.append(p)
        B.append(10)
for i in range(0,n):
    c=int(input("答案"))
    C.append(c)
for i in range(0,n):
    if C[i]==A[i] and B[i]==5:
        score += 5
    if C[i]==A[i] and B[i]==10:
        score += 10
print(score)
```

```python
#3(3)
n=int(input("题数"))
score=0
for i in range(0,n):
    a=str(input("请输入算式"))
    for word in a:
        if word == "+":
            a=a.strip("=")
            a=a.split("+")
            x=int(a[0])
            y=int(a[1])
            z=x+y
            print(z)
            score+=5
        if word == "-":
            a=a.strip("=")
            a=a.split("-")
            x=int(a[0])
            y=int(a[1])
            z=x-y
            print(z)
            score+=5
        if word == "*":
            a=a.strip("=")
            a=a.split("*")
            x=int(a[0])
            y=int(a[1])
            z=x*y
            print(z)
            score+=10
        if word == "/":
            a=a.strip("=")
            a=a.split("/")
            x=int(a[0])
            y=int(a[1])
            z=x/y
            print(z)
            score+=10
print('score',score)
```

```python
#4(3)(4)
result={}
word_nr=set()
with open(r'D:\temp\test_re_gbk.txt') as f:
    for line in f:
        words=line.split()
        for word in words:
            word_pos=word.split('/')
            if word_pos[0] in result:
                for index,pos_freq in enumerate(result[word_pos[0]]):
                    if pos_freq[0] == "nr":
                        word_nr.add(word_pos[0])
                    if pos_freq[0] == word_pos[1]:
                        result[word_pos[0]][index][1] += 1
                        break
                else:
                    result[word_pos[0]].append([word_pos[1],1])
            else:
                result[word_pos[0]]=[]
                result[word_pos[0]].append([word_pos[1],1])
list(result.items())
print(word_nr)
```

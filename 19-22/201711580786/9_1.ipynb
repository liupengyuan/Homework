```python
import random,time,bisect
A=[a for a in range(1,100001)]
B=[]
C=[]
D=[]
for a in range(100000):
    b=random.choice(A)
    B.append(b)
for a in range(1000):
    c=random.randint(1,110000)
    C.append(c)
start_time = time.time()
for a in C:
    if a in B:
        D.append(a)
end_time = time.time()
time_inorder=end_time-start_time
print(D)
print(time_inorder)
D=[]
start_time = time.time()
for a in C:
    if a <=100000:
        a=bisect.bisect_left(A, a)
        D.append(a)
end_time = time.time()
time_bisect=end_time-start_time
print(D)
print(time_bisect)
```

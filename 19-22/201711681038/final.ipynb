```python
#2
def A(m,n):
    total=0
    if m == 0:
        total = n + 1
    elif m > 0 and n == 0 :
        m = m - 1
        n = 1
        total = A(m,n)
    elif m > 0 and n > 0 :
        a = m - 1
        b = n - 1
        n = A(m,b)
        total = A(a,n)
    return total

m=3
n=4
A(m,n)
```

```python
#3(a)
import random
def ti(n,a,b):
    i=0
    tishu = 0 #计算有几道可以加分的题目
    daan1=[] #题目的答案
    jian=[]
    defen=0
    cheng=[]   #可加分小题的题号
    while i < n:
        fuhao=['+','-','*','/']
        m=random.randint(a,b)
        y=random.randint(a,b)
        x=random.choice(fuhao)
        fu=fuhao.index(x)+1
        if fu == 1 :
            z=m+y
            if z % 1 == 0:
                daan1.append(z)
                i+=1
                print(i)
                print(m,x,n,'=',sep='')
        elif fu == 2 :
            z=m-y
           
            if z % 1 == 0:
                daan1.append(z)
                i+=1
                tishu+=1
                jian.append(i)
                print(i)
                print(m,x,n,'=',sep='')
        elif fu == 3 :
            z=m*y
            
            if z % 1 == 0:
                daan1.append(z)
                i+=1
                cheng.append(i)
                tishu+=1
                print(i)
                print(m,x,n,'=',sep='')
        elif fu == 4 :
            z=m/y
            if z % 1 == 0:
                daan1.append(z)
                i+=1
                print(i)
                print(m,x,n,'=',sep='')
        
    print('有',tishu,'道可以加分的题目')
    print('题目的答案是为了方便检验才输出，实际不用输出',daan1)
    #3(b)
    daan = input('请依次输入答案，用逗号分开').split(',')
    for i in range (1,n+1):
        if i in jian:
            if int(daan[i-1]) == daan1[i-1]:
                defen += 5
        if i in cheng:
            if int(daan[i-1]) == daan1[i-1]:
                defen += 10
    print('得分是',defen)
        

n=15
a=3
b=25
ti(n,a,b)


            
    
```

```python
#4(b)
def get_word_table(filename):
    word_table = []
    with open(filename) as f:
        text = f.read()
    words = [word.split('/')[0] for word in text.split()]
    for word in words:
        if word not in word_table:
            word_table.append(word)
    return word_table


filename = r'c:\temp\成语大全.txt'
table = get_word_table(filename)


def count_words_freq(filename, words):
    word_freq_pairs = []
    with open(filename) as f:
        text = f.read()
    for word in words:
        number = text.count(word)
        word_freq_pairs.append([word, number])
    return word_freq_pairs

words_freq = count_words_freq(filename, table)

#测试用print
print(sorted(words_freq[:20]))
```

```python
#练习 1：写函数，求n个随机整数均值的平方根，整数范围在m与k之间（n,m,k由用户输入）。
import random
def pfg(n,m,k):
    i=0
    total=0
    while i<n:
        a=random.randint(m,k)
        total+=a
        i=i+1
    b=total/n
    c=b**0.5
    return c
m=int(input('请输入一个整数： '))
n=int(input('请输入一个整数： '))
k=int(input('请输入一个整数： '))
print(pfg(n,m,k))
```

```python
#练习 2：写函数，共n个随机整数，整数范围在m与k之间，（n,m,k由用户输入）。求1：西格玛log(随机整数)，2：西格玛1/log(随机整数)
import random ,math
def xgm1(n,m,k):
    i=0
    total=0
    while i<n:
        a=random.randint(m,k)
        b=math.log(a)
        total+=b
        i=i+1
    return total
def xgm2(n,m,k):
    i=0
    total=0
    while i<n:
        a=random.randint(m,k)
        b=math.log(a)
        total+=1/b
        i=i+1
    return total
m=int(input('请输入一个整数： '))
n=int(input('请输入一个整数： '))
k=int(input('请输入一个整数： '))
print(xgm1(n,m,k))
print(xgm2(n,m,k))
```

```python
#练习 3：写函数，求s=a+aa+aaa+aaaa+aa...a的值，其中a是[1,9]之间的随机整数。例如2+22+222+2222+22222(此时共有5个数相加)，
#几个数相加由键盘输入。
def aaa(a,n):
    i=1
    s=0
    b=a
    while i<=n:
        s+=a
        a=a+b*(10**i)
        i=i+1
    return s
n=int(input('请输入一个整数： '))
a=int(input('请输入一个整数在[1,9]之间： '))
print(aaa(a,n))
```

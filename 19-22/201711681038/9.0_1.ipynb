```python
import random,time
a=[]
b=[]
i=1
while i<=100000:
    a.append(i)
    b.append(i)
    i+=1
random.shuffle(b)
c=[]
n=1
while n<=110000:
    c.append(n)
    n+=1
suiji=random.sample(c, k=1000)



#顺序法
start_time = time.time()
for number in suiji:
    if number in b:
        print('找到了','这个数是',number,'在第',b.index(number)+1,'位')
    else:
        print(None,'没有找到')
end_time = time.time()
                
print('用时',end_time - start_time,'秒')


#二分法
import bisect
start_time = time.time()
a = [i for i in range(1,100001)]

for item in suiji:
    print(bisect.bisect_left(a, item))

end_time = time.time()
print('用时',end_time - start_time,'秒')
```

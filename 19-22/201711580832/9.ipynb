```python
import bisect,time,random
def order_search(list1,list2):
    start_time=time.time()
    c=[]
    for a in list1:
        for i in range(len(list2)):
            if a==list2[i]:
                c.append(a)
    print('在列表B中的数为:',c)              
    end_time=time.time()
    print('顺序查找算法时间为：',end_time-start_time)

   
def bi_search(list1,list2):
    start_time=time.time()
    c=[]
    for a in list1:
        b=bisect.bisect_left(list2,a)
        c.append(b)
    print('在列表A中的数为：',c)
    end_time=time.time()
    print('二分查找算法时间为:',end_time-start_time)


B=[]
for i in range(1,100001):
    B.append(i)
A=B.copy()
random.shuffle(B)

C=[]
for i in range(1000):
    C.append(random.randint(1,110000))
print()



order_search(C,B)
bi_search(C,A)
```

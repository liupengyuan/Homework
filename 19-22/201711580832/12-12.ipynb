```python
fh = open(r'd:\temp\idioms.txt')
text = fh.read()
text = text.replace('、', '')
idioms = text.split('\n')
fh = open(r'd:\temp\idioms_correct.txt', 'w')
fh.writelines(idioms)
fh.close()

def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table


fh = open(r'd:\temp\idioms_correct.txt')
text = fh.read()
chs = get_ch_table(text)

import random
def mou(x):
    a=''
    while len(a)<x:
        b=random.choice(chs)
        a+=b
    print(a)
mou(10)
hcy = open(r'd:\temp\1.txt', 'w')
hcy.writelines(a)
hcy.close()

mou(20)
hcy = open(r'd:\temp\2.txt', 'w')
hcy.writelines(a)
hcy.close()

mou(30)
hcy = open(r'd:\temp\3.txt', 'w')
hcy.writelines(a)
hcy.close()

mou(30)
hcy = open(r'd:\temp\3.txt', 'w')
hcy.writelines(a)
hcy.close()

mou(40)
hcy = open(r'd:\temp\4.txt', 'w')
hcy.writelines(a)
hcy.close()

mou(50)
hcy = open(r'd:\temp\5.txt', 'w')
hcy.writelines(a)
hcy.close()

mou(60)
hcy = open(r'd:\temp\6.txt', 'w')
hcy.writelines(a)
hcy.close()

mou(70)
hcy = open(r'd:\temp\7.txt', 'w')
hcy.writelines(a)
hcy.close()

mou(80)
hcy = open(r'd:\temp\8.txt', 'w')
hcy.writelines(a)
hcy.close()

mou(90)
hcy = open(r'd:\temp\9.txt', 'w')
hcy.writelines(a)
hcy.close()

mou(100)
hcy = open(r'd:\temp\10.txt', 'w')
hcy.writelines(a)
hcy.close()
```

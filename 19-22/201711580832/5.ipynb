```python
#练习 1：写函数，求n个随机整数均值的平方根，整数范围在m与k之间（n,m,k由用户输入）。
def main():
    import random
    m=int(input('请输入起始值：'))
    k=int(input('请输入终止值：'))
    n=int(input('请输入随机数的个数：'))
    i=0
    total=0
    while i<n:
        i=i+1
        number=random.randint(m,k)
        total+=number
    mean=total/n
    import math
    a = math.sqrt(mean)
    print('结果为:',a)
main()
```

```python
#练习 2：写函数，共n个随机整数，整数范围在m与k之间，（n,m,k由用户输入）。求1：西格玛log(随机整数)，2：西格玛1/log(随机整数)
m=int(input('请输入整数范围起始值：'))
k=int(input('请输入整数范围终止值：'))
n=int(input('请输入随机数的个数：'))
def plu():
    total1=0
    total2=0
    i=0
    while i<n:
        import random
        import math
        number=random.randint(m,k)
        i+=1
        a=math.log(number, 10)
        if a!=0:
            b=1/a 
        total1=a+total1
        total2=b+total2
    print('西格玛log(随机整数)和为:',total1)
    print('西格玛1/log(随机整数)和为：',total2)
plu()    


```

```python
#练习 3：写函数，求s=a+aa+aaa+aaaa+aa...a的值，其中a是[1,9]之间的随机整数。例如2+22+222+2222+22222(此时共有5个数相加)，几个数相加由键盘输入。
def pul():
    import random
    a=random.randint(1,9)
    n=int(input('请输入一个1~9正整数：'))
    i=0
    total1=0
    total2=0
    while i<n:
        m=a*(10**i)
        i+=1
        total1+=m
        total2+=total1
    print(total2)
pul()

```

```python
#挑战性练习：仿照task5，将猜数游戏改成由用户随便选择一个整数，让计算机来猜测的猜数游戏，要求和task5中人猜测的方法类似，但是人机角色对换，由人来判断猜测是大、小还是相等，请写出完整的猜数游戏
import random, math
def win():
    print(
        '''
           ======电脑赢了=======
        
    
                ."".    ."",
                |  |   /  /
                |  |  /  /
                |  | /  /
                |  |/  ;-._ 
                }  ` _/  / ;
                |  /` ) /  /
                | /  /_/\_/\
                |/  /      |
                (  ' \ '-  |
                 \    `.  /
                  |      |
                  |      |
          
          ======电脑赢了=======
        '''
    )
    
def lose():
    print(
        '''
           ======COMPUTER LOSE=======
        
    
                

                   .-"      "-.
                  /            \
                 |              |
                 |,  .-.  .-.  ,|
                 | )(__/  \__)( |
                 |/     /\     \|
       (@_       (_     ^^     _)
  _     ) \_______\__|IIIIII|__/__________________________
 (_)@8@8{}<________|-\IIIIII/-|___________________________>
        )_/        \          /
       (@           `--------`
       
       
       
          ======COMPUTER LOSE=======
        '''
    )
    
def game_over():
    print(
        '''
           ======GAME OVER=======
        
             _________ 
            / ======= \ 
           / __________\ 
          | ___________ | 
          | | -       | | 
          | |         | | 
          | |_________| |________________ 
          \=____________/                ) 
          / """"""""""" \               / 
         / ::::::::::::: \          =D-' 
        (_________________) 

       
          ======GAME OVER=======
        '''
    )
    
def show_team():
    print('牟景')

def show_instruction():
    print('这是一个电脑猜字游戏！')

def menu():
    print('''=====游戏菜单=====
                1. 游戏说明
                2. 开始游戏
                3. 退出游戏
                4. 制作团队
             =====游戏菜单=====''')
def guess_game():
    n=int(input('请输入一个要电脑猜的数，以回车结束：'))
    m=int(input('请输入猜数上界'))
    max_times = math.ceil(math.log2(m))
    guess_times = 0
    a=0
    b=m
    guess = random.randint(a,b)
    while guess_times < max_times:
        if guess<n:
            print(guess)
            input('请判断大小：')
            a=guess
            guess=random.randint(a,b)
            guess_times += 1
            print('一共可以猜', max_times, '次')
            print('电脑已经猜了', guess_times, '次')
            
        
        elif guess>n:
            print(guess)
            input('请判断大小')
            b=guess
            guess=random.randint(a,b)
            guess_times += 1
            print('一共可以猜', max_times, '次')
            print('电脑已经猜了', guess_times, '次')
        
        
        
        if guess == n:
            win()
            print('神秘数字是：', guess)
            print('电脑比标准次数少', max_times-guess_times, '次')
            break
        
    else:
        print('神秘数字是：', n)
        lose()
        
    
    
   
def main():
    while True:
        menu()
        choice = int(input('请输入你的选择'))
        if choice == 1:
            show_instruction()
        elif choice == 2:
            guess_game()
        elif choice == 3:
            game_over()
            break
        else:
            show_team()

main()

```

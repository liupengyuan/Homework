```python
import random
def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table

# 主程序
fh = open(r'c:\temp\idioms.txt')
text = fh.read()
chs = get_ch_table(text.replace('\n', ''))

for i in range(1,11):
    for j in range(1,10*i+1):
        idiom=random.choice(chs)
        print(i,idiom)
        
```

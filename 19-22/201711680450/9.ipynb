```python
import random
A = []
a = 0
while a < 100000 :
    a += 1
    A.append(a)
    
B = []
i = 0
while i < 100000 :
    b = random.choice(A)
    B.append(b)
    A.remove(b)
    i += 1
```

```python
C = []
for m in range(1000):
    c = random.randint(1,110000)
    C.append(c) 

import time 

starttime = time.time()

for c in C :
    if c in B :
        print('该元素是：',c,'该元素在列表C中的位置是：',C.index(c),'该元素在列表B中的位置是：',B.index(c))
    else :
        print('该元素是：',c,'该元素在列表C中的位置是：',C.index(c),'它不在列表B中')

endtime = time.time()
print(endtime - starttime)
```

```python
C.sort()
def bi_search(number, numbers,low,high):
    mid = (low + high)//2
    if number > numbers[-1] :
        return -1
    elif number == numbers[mid] :
        return mid
    elif number > numbers[mid] :
        return bi_search(number, numbers, mid+1, high)
    else :
        return bi_search(number, numbers, low, mid-1)


start_time = time.time()

for n in range(1000):
    c = C[n]
    print(bi_search(c,A,1,100000))

end_time = time.time()

print(end_time - start.time)

    
```

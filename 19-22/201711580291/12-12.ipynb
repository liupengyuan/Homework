```python
def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table

fh = open(r'c:\temp\idioms.txt')
text = fh.read()
chs = get_ch_table(text.replace('\n', ''))


def lwl():
    for i in range(10,101,10):
        import random
        a=str(i/10)
        fh = open('c:/temp/'+a+',txt','w')
        for j in range(i):
            idiom=random.choice(chs)
            print(idiom)
            g=str(idiom)
            fh.write(g+'\n')
lwl()
            
    
            
            
    
    
```

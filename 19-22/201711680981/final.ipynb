```python
#2
def a(m,n):
    if m == 0:
        result = n + 1
    elif m > 0 and n == 0:
        result = a(m-1,1)
    elif m > 0 and n > 0:
        result = a(m-1,a(m,n-1))
    return result
a(3,4)
```

```python
#4
#b
from collections import Counter

def count_words_freq(filename, words):
    word_freq_pairs = Counter()
    with open(filename) as f:
        text = f.read()
    for word in words:
        number = text.count(word)
        word_freq_pairs.update([word, number])
    return word_freq_pairs

words_freq = count_words_freq(r'C:\Users\18813\Desktop\语料.txt', A)
table = Counter(words_freq)
print(table.most_common(20))
```

```python
#4
#c
from collections import Counter

def count_words_freq_dict(filename):
    words_freq_dict = Counter()
    
    with open(filename) as f:
        for line in f:
            words_freq_dict.update([word.split('/')[1] for word in line.split()])
    return words_freq_dict

def main():
    filename = r'C:\Users\18813\Desktop\语料.txt'
    table = count_words_freq_dict(filename)
    print(table.most_common(50))

if __name__ == '__main__':
    main()
```

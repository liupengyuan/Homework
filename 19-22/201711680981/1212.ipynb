```python
def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table

fh = open(r'C:\temp\成语大全.text.txt')
text = fh.read()
chs = get_ch_table(text.replace('\n', ''))

import random

cha = []
for i in range(10):
    idiom = random.choice(chs)
    cha.append(idiom)
print(cha)
fh = open(r'C:\temp\成语大全.text.txt', 'w')
fh.writelines(cha)
fh.close()

cha = []
for i in range(20):
    idiom = random.choice(chs)
    cha.append(idiom)
print(cha)
fh = open(r'C:\temp\成语大全.text.txt', 'w')
fh.writelines(cha)
fh.close()

cha = []
for i in range(30):
    idiom = random.choice(chs)
    cha.append(idiom)
print(cha)
fh = open(r'C:\temp\成语大全.text.txt', 'w')
fh.writelines(cha)
fh.close()

cha = []
for i in range(40):
    idiom = random.choice(chs)
    cha.append(idiom)
print(cha)
fh = open(r'C:\temp\成语大全.text.txt', 'w')
fh.writelines(cha)
fh.close()

cha = []
for i in range(50):
    idiom = random.choice(chs)
    cha.append(idiom)
print(cha)
fh = open(r'C:\temp\成语大全.text.txt', 'w')
fh.writelines(cha)
fh.close()

cha = []
for i in range(60):
    idiom = random.choice(chs)
    cha.append(idiom)
print(cha)
fh = open(r'C:\temp\成语大全.text.txt', 'w')
fh.writelines(cha)
fh.close()

cha = []
for i in range(70):
    idiom = random.choice(chs)
    cha.append(idiom)
print(cha)
fh = open(r'C:\temp\成语大全.text.txt', 'w')
fh.writelines(cha)
fh.close()

cha = []
for i in range(80):
    idiom = random.choice(chs)
    cha.append(idiom)
print(cha)
fh = open(r'C:\temp\成语大全.text.txt', 'w')
fh.writelines(cha)
fh.close()

cha = []
for i in range(90):
    idiom = random.choice(chs)
    cha.append(idiom)
print(cha)
fh = open(r'C:\temp\成语大全.text.txt', 'w')
fh.writelines(cha)
fh.close()

cha = []
for i in range(100):
    idiom = random.choice(chs)
    cha.append(idiom)
print(cha)
fh = open(r'C:\temp\成语大全.text.txt', 'w')
fh.writelines(cha)
fh.close()
```

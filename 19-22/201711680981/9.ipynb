```python
# 生成：列表A:【1，2，3，4，……，100000】,生成列表B：【A的乱序】。计算总时间
import random,time

start_time = time.time()
A = []
for i in range(1,100001):
    A.append(i)
print('列表A:',A)
end_time = time.time()
print('总时间为{:.2f}秒'.format(end_time - start_time))

start_time2 = time.time()
random.shuffle(A)
print('列表B:',A)
end_time2 = time.time()
print('总时间为{:.2f}秒'.format(end_time2 - start_time2))
```

```python
# 随机生成1000个1-110000之间的整数：用顺序查找算法在B中找上述整数；用二分法查找算法在A中找上述整数。计算总时间
import random,time,bisect

A = [i for i in range(1,100001)]
B = [i for i in range(1,100001)]
random.shuffle(A)
A,B = B,A

C = [i for i in range(1,110000)]

D = []
for i in range(1000):
    n = random.choice(C)
    if n not in D:
        D.append(n)
        
# 在B中找D
start_time = time.time()
word_position_pairs = []
for num in D:
    for i in range(100000):
        if B[i] == num:
            word_position_pairs.append([num,i])
print(word_position_pairs)
end_time = time.time()
print('总时间为{:.2f}秒'.format(end_time - start_time))
        
# 在A中找D
start_time2 = time.time()
word_position_pairs = []
for num in D:
    i = bisect.bisect_left(A,num)
    word_position_pairs.append([num,i])
print(word_position_pairs)
end_time2 = time.time()
print('总时间为{:.2f}秒'.format(end_time2 - start_time2))
```

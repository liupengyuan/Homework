```python
import random,time,bisect
    
def time1(B,C):
    n = 0
    start_time1 = time.time()
    for ch1 in C:
        if ch1 in B:
            print(ch1)
            continue
    end_time1 = time.time()
    print("顺序法的时间是:",end_time1-start_time1)
    
def time2(A,C):
    start_time2 = time.time()
    for ch2 in C:
        print(A[bisect.bisect_left(A, ch2)])
    end_time2 = time.time()
    print("二分法的时间是:",end_time2-start_time2)
    
B = [i for i in range(100001)]
random.shuffle(B)
A = [j for j in range(100001)]
C = random.sample(A,1000)
time1(B,C)
time2(A,C)
```

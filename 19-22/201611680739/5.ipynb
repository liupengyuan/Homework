```python
#求n个随机整数均值的平方根，整数范围在m与k之间（n,m,k由用户输入）。
import random

def root():
    n = int(input('请输入你要求的数的个数n'))
    m = int(input('请输入最小的数为m'))
    k = int(input('请输入最大的数为k'))
    data = 0
    i=1
    while i<=n:
        i+=1
        t= random.randint(m,k)
        data = data + t
        print(t)
    data =(data/n)**(0.5)
    return data
#主程序
print(root())
```

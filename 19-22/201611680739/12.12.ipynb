```python
import random

def my_save(idiom):
    outfile = open(r'C:\Users\ASUS\Downloads\idioms.txt','w')
    outfile.writelines(idiom)
    outfile.close()

fh = open(r'C:\Users\ASUS\Downloads\idioms.txt')
text = fh.read()
fh.close()

text = text.replace('\n', '')
text = text.replace('、', '')
for i in range(10):
    for i in range(i*10):
        idiom = random.choice(text)
        print(idiom)
        my_save(idiom)
```

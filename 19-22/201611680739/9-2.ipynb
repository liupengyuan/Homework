```python
#随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。

import random,bisect

list1 = []
word_freq_pairs = []
for i in range(10001):
    list1.append(random.randint(1,100000))
for word in list1:
    i = bisect.bisect_left(word_freq_pairs, [word])
    if i != len(word_freq_pairs) and word_freq_pairs[i][0] == word:
        word_freq_pairs[i][1] += 1
    else:
        word_freq_pairs.insert(i, [word,1])
print(word_freq_pairs)
```

```python
#对本任务中的语料.txt文件，抽取其5001-10000行存为test1.txt文件，写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。

def count_common_words():
    count_common_list = []
    with open(r'D:\temp\test.txt',) as test:
         text = test.read()
    words = [word.split('/')[0] for word in text.split()]
    with open(r'D:\temp\test1.txt') as test1:
        for line in test1:
            words1 = [word.split('/')[0] for word in line.split()]
            for word in words1:
                if word in words and word not in count_common_list:
                    count_common_list.append(word)
    return count_common_list

def count_uncommon_words(common_words):
    count_uncommon_list = []
    with open(r'D:\temp\test.txt',) as test:
         text = test.read()
    words = [word.split('/')[0] for word in text.split()]
    for word in common_words:
        if word not in words and word not in count_uncommon_list:
            count_uncommon_list.append(word)
    with open(r'D:\temp\test1.txt') as test1:
        for line in test1:
            words1 = [word.split('/')[0] for word in line.split()]
            for word1 in common_words:
                if word1 not in words1 and word not in count_uncommon_list:
                    count_uncommon_list.append(word1)
    return count_uncommon_list

def file_deal():
    with open(r'D:\temp\test1.txt', 'w') as fw, open(r'D:\temp\语料.txt') as fr:
        line_number = 0
        for line in fr:
            if line_number > 5000 and line_number <10001:
                fw.write(line)
                line_number += 1
            elif line_number <= 5000:
                line_number += 1
                continue
            else:
                break


file_deal()
print('文件处理完成')
print(count_common_words())
common_words = count_common_words()
print(count_uncommon_words(common_words))
```

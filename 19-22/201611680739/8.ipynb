```python
import random

def show_instruction():
    print('''游戏说明:
    ''')
    
def show_team():
    print("show_team")        
    
def game_over():
    print('''======GAME OVER=======''')
    

def idioms_deal(filename):
    fh = open(filename)
    text = fh.read()
    fh.close()
    text = text.split()
    idiom = random.choice(text)
    print("第一个成语是:",idiom)
    return idiom

def idiom_robot(filename,idiom,human_score,robot_score):
    fh = open(filename)
    text = fh.read()
    fh.close()
    text = text.split()
    
    while True:
        answer_phase = input("给出你的答案吧,人类:")
        if answer_phase[0]==idiom[3]:
            if answer_phase in text:
                human_score +=10
                print("答对了,你现在的分数是:",human_score)
                for ch in text:
                    if ch[0]==answer_phase[3]:
                        print("我的答案是",ch,)
                        robot_score=robot_score+10
                        idiom = ch
                        switch = 1
                        break
                    else:
                        switch = 0
                if switch == 0:
                    print("好吧,你赢了,你的分数是",human_score,"我的分数是",robot_score)
                    return
            else:
                print("你输了,你的分数是",human_score,"我的分数是",robot_score)
                return
        else:
            print("你输了,你的分数是",human_score,"我的分数是",robot_score)
            break
    

def menu():
    print('''=====游戏菜单=====
                1. 游戏玩法说明
                2. 进入成语游戏
                3. 退出游戏
                4. 游戏制作团队
                
             =====游戏菜单====='''
         )        
             
def phase_game():
    filename = r'c:\temp\成语大全.txt'
    human_score = 0
    robot_score = 0
    switch = 0
    idiom = idioms_deal(filename)
    idiom_robot(filename,idiom,human_score,robot_score)

# 主函数
def main():
    while True:
        menu()
        choice = int(input('请输入你的选择'))
        if choice == 1:
            show_instruction()
        elif choice == 2:
            phase_game()
        elif choice == 3:
            game_over()
            break
        else:
            show_team()


#主程序
if __name__ == '__main__':
    main()
```

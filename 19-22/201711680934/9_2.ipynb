```python
# 随机生成1万个整数，范围在0-10万之间，分别进行简单选择排序、快速排序（自行递归实现的）以及内置sort函数3种排序，打印出3种排序的运行时间。

import random,time

list = []
for i in range(10001):
    n = random.randint(0,100001)
    list.append(n)
    
def sort_simple_selection(A):
    start_time1 = time.time()
    for i in range(len(A)-1):
        min = i
        for j in range(i+1, len(A)):
            if A[j] < A[min]:
                min = j
        A[i], A[min] = A[min], A[i]
    end_time1 = time.time()
    print('简单排序用时为：', end_time1 - start_time1)
    
sort_simple_selection(list)
```

```python
import sys,random,time
sys.setrecursionlimit(10000)

list = []
for i in range(10001):
    n = random.randint(0,100001)
    list.append(n)

def quick_sort(seq):
    start_time2 = time.time()
    left_seq =  []
    right_seq = []
    p=seq[0]
    for number in seq[1:]:
        if number <= p:
            left_seq.append(number)
        else:
            right_seq.append(number)
    if left_seq:
        left_seq = quick_sort(left_seq)
    if right_seq:
        right_seq = quick_sort(right_seq)
    end_time2 = time.time()
    print('快速排序用时为：', end_time2 - start_time2)
    
quick_sort(list)
```

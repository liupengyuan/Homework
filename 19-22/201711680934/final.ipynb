```python
#2

def Ackermann(m,n):
    if m == 0:
        return n + 1
    elif m > 0 and n == 0:
        return Ackermann(m - 1,1)
    else:
        return Ackermann(m-1,Ackermann(m,n-1))
Ackermann(3,4)
```

```python
#3
#(a)、 (b)
import random
def math_test(n,a,b):
    signs = ['+','-','*','/']
    i = 1
    outcomes = []
    while i < n+1:
        sign = random.choice(signs)
        A = random.randint(a,b)
        B = random.randint(a,b)
        if sign == '/':
            if A%B == 0:
                print('Q',i,'：',A,sign,B)
                outcome = int(A/B)
                outcomes.append(outcome)
                i += 1
            else:
                i = i
        elif sign == '-':
            if A - B > 0:
                print('Q',i,'：',A,sign,B)
                outcome = A-B
                outcomes.append(outcome)
                i += 1
            else:
                i = i
        elif sign == '*':
            print('Q',i,'：',A,sign,B)
            i += 1
            outcome = A*B
            outcomes.append(outcome)
        else:
            print('Q',i,'：',A,sign,B)
            outcome = A+B
            outcomes.append(outcome)
            i += 1
    score = 0
    my_outcome = list(input('您的答案是'))
    print('正确答案是',outcomes)
    for i in range(1,n+1):
        if my_outcome[i] == outcomes[i]:
            score += 5
        else:
            score = score
math_test(15,3,25)
```

```python
#(b)
def count_words_freq(filename):
    word_freq_pairs = []

    with open(r'd:\temp\成语大全(含成语解释)_utf8.txt',encoding='utf8') as f:
        text = f.read()
    with open(filename) as fh:
        test = fh.read()
    idioms = [idiom.split('/')[0] for idiom in text.split()]
    words = [word.split('/')[0] for word in test.split()]
    for idiom in idioms:
        while idiom in words:
            for item in word_freq_pairs:
                if idiom == item[0]:
                    item[1] += 1
                    break
            else:
                word_freq_pairs.append([idiom, 1])
            
    return word_freq_pairs


filename = r'd:\temp\语料.txt'
table = count_words_freq(filename)
print(table[:20])
```

```python
# (c)
def count_firm_freq(filename):
    '''
    给定文本文件，建立词表并同时统计词频
    '''
    firm_freq_pairs = []
    
    with open(filename) as f:
        for line in f:
            words = [word.split('/')[1] for word in line.split()]
            for word in words:
                for item in firm_freq_pairs:
                    if word == item[0]:
                        item[1] += 1
                        break
                else:
                    firm_freq_pairs.append([word, 1])
                    
    return firm_freq_pairs


filename = r'd:\temp\语料.txt'
table = count_firm_freq(filename)
print(table)
```

```python
# 生成列表A：[1,2,3,4,......,100000]；生成列表B：[A的乱序]
# 随机生成1000个1-110000之间的整数：用顺序查找算法在B中找上述整数；用二分查找算法在A中找上述整数，并分别计算两者的时间。

import random
import bisect
import time

A = []
for i in range(1,100001):
    A.append(i)
print (A)

B = sorted(A)
random.shuffle(B)
print(B)

C = []
for i in range(1001):
    a = random.randint(1,110001)
    C.append(a)
    
start_time = time.time()
for m in C:
    if m in B:
        print(m)
end_time = time.time()
print('顺序查找算法用时为：',end_time - start_time,'秒')

start_time = time.time()
for n in C:
    print(bisect.bisect_left(A, n))
end_time = time.time()
print('二分查找算法用时为：',end_time - start_time,'秒')
```

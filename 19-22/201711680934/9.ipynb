```python
# A[1,2,3,4,......,100000]B[A]
# 10001-110000BA

import random
import bisect
import time

A = []
for i in range(1,100001):
    A.append(i)
print (A)

B = sorted(A)
random.shuffle(B)
print(B)

C = []
for i in range(1001):
    a = random.randint(1,110001)
    C.append(a)
    
start_time = time.time()
for m in C:
    if m in B:
        print(m)
end_time = time.time()
print('',end_time - start_time,'')

start_time = time.time()
for n in C:
    print(bisect.bisect_left(A, n))
end_time = time.time()
print('',end_time - start_time,'')
```

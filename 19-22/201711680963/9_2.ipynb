#### 随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。

```python
import random
from collections import defaultdict


wordfreq=defaultdict(int)
A=[random.randint(1,100000) for i in range(10000)]
B=sorted(A)

for word in B:
    if word in B:
        wordfreq[word]+=1
    else:
        wordfreq[word]=1
print(B)
print(wordfreq)
```

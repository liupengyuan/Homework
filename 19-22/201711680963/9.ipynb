```python
import random
import time

A=[]
i=0
for i in range(1,100000):
    A.append(i)
    i=i+1
print(A)

B=A
random.shuffle(B)
print(B)


c=random.randint(1,110000)
print(c)

#
a=0
for b in B:
    d=B[a]
    if d==c:
        endtime1=time.time()
        print(a)
        print(endtime1)
    else:
        a=a+1

#
import bisect
e=bisect.insort_left(A,c)
endtime2=time.time()
print(e)
print(endtime2)

```

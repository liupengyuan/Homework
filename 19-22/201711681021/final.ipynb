```python
# 3(b)
import random

def jia(a,b):
    x = random.randint(a,b)
    y = random.randint(a,b)
    print(x,'+',y,'= ?')
    trueanswer = x + y
    return trueanswer

def jian(a,b):
    x = random.randint(a,b)
    y = random.randint(a,b)
    print(x,'-',y,'= ?')
    trueanswer = x - y
    return trueanswer
    
def mul(a,b):
    x = random.randint(a,b)
    y = random.randint(a,b)
    print(x,'*',y,'= ?')
    trueanswer = x * y
    return trueanswer
    
def chu(a,b):    
    x = random.randint(a,b)
    y = random.randint(a,b)
    print(x,'//',y,'= ?')
    trueanswer = x // y
    return trueanswer
    
def question(n,a,b):
    score = 0
    i = 1
    trueanswers = []
    while i <= n:
        c = random.randint(1,4)
        if c == 1:
            jia(a,b)
            trueanswers.append(jia(a,b))
        elif c == 2:
            jian(a,b)
            if jian(a,b) > 0:
                continue
            else:
                pass
                jian(a,b)
            trueanswers.append(jian(a,b))
        elif c == 3:
            mul(a,b)
            trueanswers.append(mul(a,b))
        else:
            chu(a,b)
            trueanswers.append(chu(a,b))
            
        i += 1
    j = 0
    answer = [input()]
    while j <= n-1:
        
        if answer[j] != trueanswers[j]:
            score = score + 0
        else:
            score = score + 5
        j += 1
    print(score)
question(15,3,25)
```

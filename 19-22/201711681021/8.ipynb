```python
import random

def win():
      print(
        '''
           ======恭喜你，你赢了=======
        
    
                ."".    ."",
                |  |   /  /
                |  |  /  /
                |  | /  /
                |  |/  ;-._ 
                }  ` _/  / ;
                |  /` ) /  /
                | /  /_/\_/\
                |/  /      |
                (  ' \ '-  |
                 \    `.  /
                  |      |
                  |      |
          
          ======恭喜你，你赢了=======
        '''
    )
    
    
def lose():
        print(
        '''
           ======YOU LOSE=======
        
    
                

                   .-"      "-.
                  /            \
                 |              |
                 |,  .-.  .-.  ,|
                 | )(__/  \__)( |
                 |/     /\     \|
       (@_       (_     ^^     _)
  _     ) \_______\__|IIIIII|__/__________________________
 (_)@8@8{}<________|-\IIIIII/-|___________________________>
        )_/        \          /
       (@           `--------`
       
       
       
          ======YOU LOSE=======
        '''
    )
    
    
def get_ch_table(line):
    ch_table = []
    for ch in line:         
        ch_table.append(ch)
    return ch_table

def idiom_robot(file_name):
    with open(file_name) as fh:
        text = fh.read()
        text = text.replace('、', '')

    idioms = text.split()
    
    word = random.choice(idioms)
    print(word)
    x = get_ch_table(word)
    y = len(x)
    wordbefore_last = x[y-1]
    wordafter_first = wordbefore_last 
    for idiom in idioms:
        if idiom.startswith(wordafter_first):
            trueword = idiom
            return trueword
        
    

def process(z):
    with open(file_name) as fh:
        text = fh.read()
        text = text.replace('、', '')

    idioms = text.split()
    x = get_ch_table(z)
    y = len(x)
    wordbefore_last = x[y-1]
    wordafter_first = wordbefore_last 
    for idiom in idioms:
        if idiom.startswith(wordafter_first):
            print(idiom)
        else:
            print('无成语可接，电脑失败！')
            win()
            
    h = get_ch_table(idiom)
    f = len(h)
    wordbefore_last_1 = h[f-1]
    wordafter_first_1 = wordbefore_last_1
    for idiom in idioms:
        if idiom.startswith(wordafter_first_1):
            trueword_1 = idiom
    return trueword_1

def main():
    filename = r'c:\temp\成语大全.txt'
    score_1 = 0
    score_2 = 0
    n = 1
    while n > 0:
        real_idiom = idiom_robot(filename)
        answer_idiom = input('请输入成语（若输入空格则退出游戏）：')
        if answer_idiom == real_idiom:
            print('答对了，加十分')
            score_1 += 10
            print('你当前的分数是：', score_1)
            process(answer_idiom)
            score_2 += 10
            
        elif answer_idiom == ' ':
            print('退出游戏。')
            print('你最后的分数是：', score_1)
            return
        else:
            print('答错了，游戏结束')
            print('你最后的分数是：', score_1)
            lose()
            return


print('游戏开始')
main()
```

# 生成列表A：[1，2，3，4，……，100000].
# 生成列表B：[A的乱序].

# 随机生成1000个1-110000之间的整数：
# 用顺序查找算法在B中找上述整数
# 用二分查找算法在A中找上述整数

# 并分别计算两者总时间

```python
import random,time,sys
# 二分查找那个递归深度…………
A = []
a = 1
i = 100000
while a <= i:
    A.append(a)
    a += 1

B = A
random.shuffle(B)

guess_ch_table = []
while len(guess_ch_table) <=1000:
    number = random.randint(1,110000)
    guess_ch_table.append(number)
print(guess_ch_table) 

def count_words_freq(list_1,guess_1):
    word_freq_pairs = []
    for i in list_1:
        if i in guess_1:
            word_freq_pairs.append(i)
    total = time.time()
    print(total)
    return word_freq_pairs

def bi_search(list_2,guess_2,low ,high):
    real = []
    sys.setrecursionlimit(3000)
    mid = (low + high)//2
    for i in guess_2:
        number = i
        if number == list_2[mid]:
            real.append(mid)
        elif number > list_2[mid]:
            return bi_search(list_2,guess_2, mid+1, high)
        else:
            return bi_search(list_2,guess_2, low, mid-1)
        total = time.time()
        print(total)
        return real
a = A[0]
b = A[len(A)-1]
count_words_freq(B,guess_ch_table)
print('顺序查找：',count_words_freq(B,guess_ch_table))
bi_search(A,guess_ch_table,a ,b)
print('二分查找：',bi_search(A,guess_ch_table,a ,b))
```

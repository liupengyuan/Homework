```python
#成语接龙游戏。初始计算机得分为0、人类得分为0，每接对一个成语加十分。接龙错误则诺戏结束。最后给出人类得分，和计算机得分。
import random, math

def game_over():
    print('游戏结束啦！')
def show_team():
    print('zxf')
def show_instruction():
    print('由计算机先出成语，初始计算机得分为0、人类得分为0，每接对一个成语加十分。接龙错误则诺戏结束。最后给出人类得分，和计算机得分。')
def menu():
    print('''          游戏菜单
          1. 游戏说明
          2. 开始游戏
          3. 退出游戏
          4. 制作团队''')

fh = open(r'c:\temp\成语大全.txt')
text = fh.read()
fh.close()
idioms = text.split()
def cyjl():
    idiom1 = random.choice(idioms)
    print('成语是',idiom1)
    n=0
    score_p=0
    score_c=0
    while n>=0:    
        answer_idiom = input('请输入成语.')
        n+=1
        if answer_idiom[0] ==idiom1[3]:
            print('答对了，加十分')
            score_p += 10
            for idiom in idioms:
                i=1
                if idiom[0] == answer_idiom[-1]:
                    score_c += 10
                    idiom1 = idiom
                    print(idiom1)
                    break
                else:
                    i += 1
                    if i == len(idioms):
                        game_over()
                        print('你的分数是',score_p)
                        print('我的分数是',score_c)  
                        break
        else:
            game_over()
            print('你的分数是',score_p)
            print('我的分数是',score_c)
            break
    return
def main():
    while True:
        menu()
        choice = int(input('请输入你的选择'))
        if choice == 1:
            show_instruction()
        elif choice == 2:
            cyjl()
        elif choice == 3:
            break
        else:
            show_team()


if __name__ == '__main__':
    main()


```

```python
#挑战性练习：仿照task5，将猜数游戏改成由用户随便选择一个整数，让计算机来猜测的猜数游戏，要求和task5中人猜测的方法类似，但是人机角色对换，由人来判断猜测是大、小还是相等，请写出完整的猜数游戏。
import random, math

def win():
    print('哈哈٩(๑❛ᴗ❛๑)۶ ，我赢了！')
def lose():
    print('/(ㄒoㄒ)/~~，我输了 ！！')
def game_over():
    print('ヾ(✿ﾟ▽ﾟ)ノ，游戏结束！')
def show_team():
    print('小风')
def show_instruction():
    print('由用户随便选择一个整数，让计算机来猜测的猜数游戏，由人来判断猜测是大、小还是相等。')

def menu():
    print('''
               游戏菜单(◕ᴗ◕✿)
                1. 游戏说明
                2. 开始游戏
                3. 退出游戏
                4. 制作团队
             (｡◕ˇ∀ˇ◕)游戏菜单''')
            
def guess_game():
    n = int(input('请输入一个大于0的整数作为你给出的范围上线，回车结束。'))
    m = int(input('请输入一个大于0小于n的整数作为你给出的范围下线，回车结束。'))
    k = int(input('请在你给出的范围挑一个数，回车结束。'))
    max_times = math.ceil(math.log(k, 2))
    guess_times = 0
    
    while guess_times < max_times:
        a=random.randint(m,n)
        guess_times += 1
        print('我猜测你输入的数字是',a)
        print('一共可以猜', max_times, '次')
        print('我已经猜了', guess_times, '次')
        i =int(input('请比较我猜测的数与你给的数，如果我猜对了请输入0，猜大了请输入1，猜小了请输入2，回车结束。'))
        if i<1:
            win()
            print('你给的数字是：', a)
            print('我比标准次数少', max_times-guess_times, '次')
            break
        elif 0<i<2:
            n = a-1
            print('哇，我猜大了！')
        else:
            m = a+1
            print('哇，我猜小了！')
    else:
        lose()

def main():
    while True:
        menu()
        choice = int(input('请输入你的选择'))
        if choice == 1:
            show_instruction()
        elif choice == 2:
            guess_game()
        elif choice == 3:
            game_over()
            break
        else:
            show_team()

if __name__ == '__main__':
    main()
```

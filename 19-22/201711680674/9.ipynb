```python
#生成列表A：[1,2,3,4,......100000]，生成列表B：[A的乱序]，
#随机生成1000个1-110000之间的整数，
#用顺序查找算法在B中找上述整数，用二分查找算法在A中找上述整数
#并分别计算两者总时间
```

```python
import random
import time
import bisect

A=[]
a=1
while a<100001:
    A.append(a)
    a += 1
B=[]
b=1
while b<100001:
    B.append(b)
    b += 1
random.shuffle(B)
c_list = range(0,110001)
C=random.sample(c_list,1000)

b1=[]
start_time = time.time()
for number1 in C:
    if number1 in B :
        b1.append(number1)
    else:
        pass
end_time = time.time()
print(b1)        
print( end_time - start_time)
print('\n')

a1=[]
start_time = time.time()
for number in C :
    if number <= 100000:
        number=bisect.bisect_left(A,number)
        a1.append(number)
    else:
        pass
end_time = time.time()
print(a1)
print( end_time - start_time)
```

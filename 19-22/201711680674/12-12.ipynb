```python
import random
def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table

fh = open(r'c:\temp\idioms.txt')
text = fh.read()
chs = get_ch_table(text.replace('\n', ''))

print(len(chs))
print('-----------------------------------------------------------------------')

c=[]
i=0
a=1
while a<11:
    while len(c)<(a*10):    
        ch= random.choice(chs)
        c.append(ch)
        b=[r'c:\temp\i','i','.txt']
        b[0]=b[0][:-1]
        b[1]=str(a)
        b=' '.join(b)
        fh=open(b, 'w')
        fh.writelines(c)
        fh.close()
    a+=1
    print(c)
    c=[]
    print('-----------------------------------------------------------------------------------------------------------------------------------')

```

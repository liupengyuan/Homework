```python
A=[]
i=0
while i<100000:
    i=i+1
    A.append(i)
B=[]
i=0
while i<100000:
    i=i+1
    B.append(i)
import random,bisect,time

random.shuffle(B)

c=[]
i=0
while i<1000:
    i=i+1
    c.append(random.randint(1,110000))

d = []
start_time = time.time()
for word_1 in c:
    if word_1 in B:
        d.append(word_1)
    else:
        pass
end_time = time.time()
print(d)
print("所用时间为",end_time-start_time)

e=[]
start_time = time.time()
for word_1 in c:
    if word_1 in A:
        e.append(bisect.bisect_left(A, word_1)+1)
    else:
        pass
end_time = time.time()
print(e)
print("所用时间为",end_time-start_time)
print("可查找到的元素的个数为",len(e),len(d))


    

    
    




            
            
            
    

            

    



    
```

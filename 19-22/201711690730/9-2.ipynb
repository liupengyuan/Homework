```python
#、随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
import random
c=[]
i=0
while i<10000:
    i=i+1
    c.append(random.randint(0,100000))
b=[]
a=set(c)
a=list(a)
def sort_simple_selection(seq):
    for i in range(len(seq)-1):
        min = i
        for j in range(i+1, len(seq)):
            if seq[j] < seq[min]:
                min = j
        seq[i], seq[min] = seq[min], seq[i]


sort_simple_selection(a)


for item in a:
    number=c.count(item)
    b.append([item,number])
print(b)

    




    


```

# •写函数，返回某个元素/对象在一个list中的位置，如果不在，则返回-1.

```python
def ret(number,a):
    if a in number:
        b=number.index(a)
        return b
    else:
        return -1
    
number=[1,2,3,4,5,6,7,89]
a=int(input())
ret(number,a)
    
```

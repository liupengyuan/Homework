# 3、对本任务中的语料.txt文件，抽取其5001-10000行存为test1.txt文件，写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。

```python
def word(filename):
    lnum=0
    list1 = []
    with open(filename,'r') as fd:
        for line in fd:
            lnum += 1
            if lnum >=5001 and lnum<=10000:
                list1.append(line)
    fd.close()
    
    fh = open(r'd:\temp\test1.txt', 'w')
    fh.writelines(list1)
    fh.close()

    fh = open(r'd:\temp\test.txt')
    list2 = fh.read()
    fh.close()
    
    common = []
    only = []
    for word in list1:
        if word in list2:
            common.append(word)
        else:
            only.append(word)
    return common,only

filename = r'd:\temp\语料.txt'

word(filename)
```

# 4、挑战性任务：统计本任务中的语料.txt文件，得到每一个词对应不同词性的频次，可表示为：{词:[['n',1000], ['v', 98]...],...}，即在dict对象中，key为词，value为列表，输出词频最高的50个条目。

```python
from collections import Counter

def count_words_freq_dict(filename):
    words_freq_dict = Counter()
    words = []
    with open(filename) as f:
        for line in f:
            words.update([word.split('/')[0] for word in line.split()])
            for word in words:
                words_freq_dict.update([word.split('/')[1] for word in line.split()])
    return words_freq_dict

def main():
    filename = r'd:\temp\语料m.txt'
    table = count_words_freq_dict(filename)
    a = table.most_common(50)
    print(a)

main()
```

```python
import random , time , bisect

a = [i for i in range(100000)]
b = a
random.shuffle(b)

list = [random.randint(0,100000) for i in range(1000)]

start_b = time.time()
num_b = []
for num in list:
    n = list.index(num)
    num_b.append(n)
end_b = time.time()
time_b = end_b - start_b 
print('用顺序法查找算法在B中找出整数的时间；',time_b)
print(num_b)

print('-'*20)
start_a = time.time()
num_a = []
for i in list:
    m = bisect.bisect_left(range(1000000), i)
    num_a.append(m)
end_a = time.time()
time_a = end_a - start_a
print('用二分查找算法在A中找出整数的时间；',time_a)
print(num_a)
```

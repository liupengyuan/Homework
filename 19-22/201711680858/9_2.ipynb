```python
#3、挑战性任务：统计本任务中的语料.txt文件，得到每一个词对应不同词性的频次，可表示为：{词:[['n',1000], ['v', 98]...],...}，即在dict对象中
#key为词，value为列表，输出词频最高的50个条目。
def count_words_freq(filename):
    from operator import itemgetter
    word_freq_pairs = []
    with open(filename) as f:
        for line in f:
            words = [ word for word in line.split()]
            for word in words:
                for item in word_freq_pairs:
                    if word == item[0]:
                        item[1] += 1
                        break
                else:
                    word_freq_pairs.append([word, 1])
        print(sorted(word_freq_pairs,key=itemgetter(1),reverse=False))

filename = r'd:\temp\语料.txt'
count_words_freq(filename)
```

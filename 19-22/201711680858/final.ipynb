```python
#3(c)
def add(x, y):
 
   return x + y
 
def subtract(x, y):
    return x - y
 
def multiply(x, y):
    return x * y
 
def divide(x, y):
    return x / y
print("选择运算：")
print("1、相加")
print("2、相减")
print("3、相乘")
print("4、相除")
n=0
score=0
while n<=10:
    choice = input("输入你的选择(1/2/3/4):")
    num1 = int(input("输入第一个数字: "))
    num2 = int(input("输入第二个数字: "))
    n=n+1
    if choice == '1':
        print(num1,"+",num2,"=", add(num1,num2))
        score=score+5
    elif choice == '2':
        score=score+5
        print(num1,"-",num2,"=", subtract(num1,num2))
    elif choice == '3':
        score=score+10
        print(num1,"*",num2,"=", multiply(num1,num2))
    elif choice == '4':
        score=score+10
        print(num1,"/",num2,"=", divide(num1,num2))
    else:
        print("非法输入")
print(score)
```

```python
#4(c)
def count_words_freq(filename):
    from operator import itemgetter
    word_freq_pairs = []
    with open(filename) as f:
        for line in f:
            words = [ word for word in line.split()]
            for word in words:
                for item in word_freq_pairs:
                    if word == item[0]:
                        item[1] += 1
                        break
                else:
                    word_freq_pairs.append([word, 1])
        print(sorted(word_freq_pairs,key=itemgetter(1),reverse=False))
filename = r'd:\temp\成语.txt'
count_words_freq(filename)
```

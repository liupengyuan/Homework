```python
#用顺序查找法算法在B中找上述整数
import datetime
start = datetime.datetime.now()
def search(list,item):
    pos=0
    found=False
    while pos!=len(list) and not found:
        if list[pos]==item:
            found=True
            print(list[pos])
        else:
            pos=pos+1
    return found
import random
x = [random.randint(1,110000) for i in range(1000)]
A=[x for x in range(1,100001)]
random.shuffle(A)
B=A
for i in x:
    search(B,i)
end = datetime.datetime.now()
print ('程序运行时间为:',end-start)
```

```python
#用二分法查找算法在A中找上述整数
import datetime
start = datetime.datetime.now()
def binary_search_bisect(lst,n):
    from bisect import bisect_left
    m=bisect_left(lst,n)
    if m!=len(lst)and lst[m]==n:
        print(lst[m])
A=[x for x in range(1,100001)]
x = [random.randint(1,110000) for i in range(1000)]
for i in x:
    binary_search_bisect(A,i)
end = datetime.datetime.now()
print ('程序运行时间为:',end-start)
```

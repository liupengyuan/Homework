```python
#练习 1：仿照求$ \sum_{i=1}^mi + \sum_{i=1}^ni  + \sum_{i=1}^ki$的完整代码，写程序，可求m!+n!+k!

def mul(end):
    i = 0
    total_n = 1
    
    while i < end:
        i += 1
        total_n *= i
        
    return total_n

n = int(input('请输入第一个整数，以回车结束。'))
m = int(input('请输入第二个整数，以回车结束。'))
k = int(input('请输入第三个整数，以回车结束。'))

print('最终的和是：', mul(n) + mul(m) + mul(k))
```

```python
#练习 2：写函数可返回1 - 1/3 + 1/5 - 1/7...的前n项的和。在主程序中，分别令n=1000及100000，打印4倍该函数的和。

from fractions import Fraction

def total(n):
    i = 1
    total = 1
    while i < n:
        i += 1
        if i % 2 == 0:
            total -= Fraction(1,2*i-1)
        elif i % 2 == 1:        
            total += Fraction(1,2*i-1)
    return total


print(total(1000)*4)
print(total(10000)*4)
```

```python
#挑战性练习：写程序，可以求从整数m到整数n累加的和，间隔为k，求和部分需用函数实现，主程序中由用户输入m，n，k调用函数验证正确性。

def total(m,n,k):
    
    if m > n:
        a = n
        n = m
        m = a
    total = 0
    i = m 
    while i <= n:
        total += i
        i += k
        
    return total

m = int(input('请输入第一个整数：'))
n = int(input('请输入第二个整数：'))
k = int(input('请输入间隔：'))

print(total(m,n,k))
```

```python
#随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。

import random
nums = []
for i in range(10000):
    num = random.randrange(0, 100001)
    nums.append(num)
    
nums_freq_dict = {}
for num in nums:
    if num in nums_freq_dict:
        nums_freq_dict[num] += 1
    else:
        nums_freq_dict[num] = 1

sorted_dict = sorted(nums_freq_dict.items())
print(sorted_dict)
```

```python
#对本任务中的语料.txt文件，抽取其5001-10000行存为test1.txt文件，写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。


with open(r'd:\temp\test1.txt','w') as fw,\
open(r'd:\temp\语料.txt') as fr:
    line_num = 0
    for line in fr:
        if line_num < 5000:
            line_num += 1
        elif 5000 <= line_num < 10000:
            fw.write(line)
            line_num += 1
        else:
            break
            

def ch_table(filename):
    words_table = set()
    ch_table = set()
    with open(filename) as f:
        for line in f:
            words_table |= {word.split('/')[0] for word in line.split()} 
            for word in words_table:
                ch_table |= {ch for ch in word}
        return ch_table
                    
ch_test_table = ch_table(r'd:\temp\test.txt')
ch_test1_table = ch_table(r'd:\temp\test1.txt')

def intersection(a,b):
    return a & b

def difference(a,b):
    return a - b
    
print('共用字：', intersection(ch_test_table, ch_test1_table))
print('test1独用字：', difference(ch_test1_table, ch_test_table))
```

```python
#用1M的语料跑了十多分钟，就不尝试语料.txt了… 应该是我方法有问题，但是尽力了QAQ

#挑战性任务：统计本任务中的语料.txt文件，得到每一个词对应不同词性的频次，可表示为：{词:[('n',1000), ('v', 98)...],...}，即在dict对象中，key为词，value为列表，输出词频最高的50个条目。
big_list = [] #[[word,property],[...,..]...]
word_table = set() #（word, word, ...）

with open(r'd:/temp/test.txt') as f:
    for line in f:
        words = [word for word in line.split()]
        word_table |= {word.split('/')[0] for word in line.split()}
        for item in words:
            word = []
            word.append(item.split('/')[0])
            word.append(item.split('/')[1])
            big_list.append(word)
        
word_list = list(word_table) #word_table 转list
big_dict = {} #所有词按给定格式生成的词典


for i in range(0, len(word_list)):
    pro_count_dict = {}
    for j in range(0, len(big_list)):
        if word_list[i] == big_list[j][0]:
            if big_list[j][1] in pro_count_dict:
                pro_count_dict[big_list[j][1]] += 1
            else:
                pro_count_dict[big_list[j][1]] = 1
    properties = (list(pro_count_dict.items()))
    big_dict[word_list[i]] = properties

#选出前50 词频最高
words_freq_dict = {}
with open(r'd:\temp\test.txt') as ff:
    for line in ff:
        words = [word.split('/')[0] for word in line.split()]
        for word in words:
            if word in words_freq_dict:
                words_freq_dict[word] += 1
            else:
                words_freq_dict[word] = 1
sorted_dict = sorted(words_freq_dict.items(), key = lambda item:item[1], reverse = True)
sorted_list = sorted_dict[:50]
small_words_list = []#词频最高的50词
for item in sorted_list:
    list_item = list(item)
    small_words_list.append(list_item)


small_dict = {}

for item in small_words_list:
    for key in big_dict.keys():
        if item[0] == key:
            small_dict[key] = big_dict[key]
print(small_dict)
```

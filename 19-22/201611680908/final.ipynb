```python
#1.
def pic(sym,num):
    for i in range(1, num + 1):
        print(' ' * (num - i) + ' ' * (num - i) + (' ' + sym)*(i-1) + ' ' + sym*5 + (' ' + sym)*(i-1) )
        
pic('*', 5)
```

```python
#2.
def A(m, n):
    if m == 0:
        return n + 1
    elif m > 0 and n == 0:
        return A(m - 1, 1)
    elif m > 0 and n > 0:
        return A(m - 1, A(m, n - 1))

A(3, 4) 
```

```python
#3.(1)
import random
def ques(n ,a, b):
    i = 0
    while i < n:
        x = random.randrange(a, b+1)
        y = random.randrange(a, b+1)
        sym = random.choice(['+', '-', '*', '/'])
        ans = eval(str(x) + sym + str(y))
        if ans % 1 == 0 and ans >= 0: 
            ans = []
            ans.append(ans)
            print(x, sym, y, '=')
            i += 1
    
ques(15,3,25)

#3.(2)
user_ans = int(input('输入答案，逗号隔开'))
sheet = list(user_ans)
print(sheet)

```

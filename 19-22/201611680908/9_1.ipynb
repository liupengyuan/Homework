```python
import random, copy, time, bisect
A = []
B = []
numbers = []
for i in range(1,100001):
    A.append(i)
B = copy.deepcopy(A)
random.shuffle(B)
print(A)
print(B)

print('\n'*3)

#随机生成1000个1-110000之间的整数
for i in range(1000):
    num = random.randrange(1, 110001)
    numbers.append(num)
print(numbers)

#顺序查找 B 
start_time = time.time()
for num in numbers:
    for item in B:
        if num == item:
            print(num, '在B中对应索引：', B.index(item))
            break
    else:
        print(num, '在B中不存在')
end_time = time.time()
print('顺序查找用时{:.2f}秒'.format(end_time - start_time))

print('\n'*3)

#二分查找 A
start = time.time()
for num in numbers:
    if bisect.bisect_left(A, num) != len(A):
        print(num, '在A 中对应索引：', bisect.bisect_left(A, num))
    else:
        print(num, '在A中不存在')
end = time.time()
print('二分查找用时{:.2f}秒'.format(end - start))
```

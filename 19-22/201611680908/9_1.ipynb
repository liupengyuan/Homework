```python
import random, copy, time, bisect
A = []
B = []
numbers = []
for i in range(1,100001):
    A.append(i)
B = copy.deepcopy(A)
random.shuffle(B)
print(A)
print(B)

print('\n'*3)

#10001-110000
for i in range(1000):
    num = random.randrange(1, 110001)
    numbers.append(num)
print(numbers)

# B 
start_time = time.time()
for num in numbers:
    for item in B:
        if num == item:
            print(num, 'B', B.index(item))
            break
    else:
        print(num, 'B')
end_time = time.time()
print('{:.2f}'.format(end_time - start_time))

print('\n'*3)

# A
start = time.time()
for num in numbers:
    if bisect.bisect_left(A, num) != len(A):
        print(num, 'A ', bisect.bisect_left(A, num))
    else:
        print(num, 'A')
end = time.time()
print('{:.2f}'.format(end - start))
```

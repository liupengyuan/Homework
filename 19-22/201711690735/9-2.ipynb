```python
#随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
```

```python
import random
nums = []

for i in range(10001):
    j = random.randint(0,100000)
    nums.append(j)
num_freq_pairs = []

for num in nums:
    for item in num_freq_pairs:
        if num == item[0]:
            item[1] += 1
            break
    else:
        num_freq_pairs.append([num,1])
        
num_freq_pairs.sort()
for k in range(len(num_freq_pairs)):
    print(num_freq_pairs[k])
        
```

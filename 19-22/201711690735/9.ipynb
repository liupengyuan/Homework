```python
import random,bisect,time

A = []
B = []
C = []
for i in range(1,100001):
    A.append(i)
    j=random.randint(1,100000)
    B.append(j)
    
start_time_A=time.time()
for k in range(1000):
    s=random.randint(1,110000)
    C.append(s)
    print('在A中:',s,bisect.bisect_left(A,s))
end_time_A=time.time()

start_time_B=time.time()
for s in C:
    if s in B:
        print('在B中：',s,B.index(s))
    else:
        print('在B中：',s,-1)
end_time_B=time.time()  

time_A=end_time_A-start_time_A
time_B=end_time_B-start_time_B
print('A用时：',time_A,'秒')
print('B用时：',time_B,'秒')

    
    
```

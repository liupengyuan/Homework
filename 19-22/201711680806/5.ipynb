```python
练习 2：写函数，共n个随机整数，整数范围在m与k之间，（n,m,k由用户输入）。求1：西格玛log(随机整数)，2：西格玛1/log(随机整数)
```

```python
import random,math

def my_number():
    number = random.randint(m,k)
    return number


def my_math():
    x = my_number()
    c = math.log(x)
    return c

n=int(input('请输入一个正整数，以回车结束。'))
m=int(input('请输入一个正整数，以回车结束。'))
k=int(input('请输入一个正整数，以回车结束。'))

if m>k:
    a = k
    k = m
    m = a
    
i = 0
total_1 = 0
total_2 = 0
while i < n:
    i += 1
    b = my_math()
    d = 1/b
    total_1 += b
    total_2 += d
    
print(total_1,total_2)
```

```python
挑战性练习：仿照task5，将猜数游戏改成由用户随便选择一个整数，让计算机来猜测的猜数游戏，要求和task5中人猜测的方法类似，但是人机角色对换，由人来判断猜测是大、小还是相等，请写出完整的猜数游戏。
```

```python
import random, math


def win():
     print(
        '''
           ======恭喜你，你赢了=======
        
    
                ."".    ."",
                |  |   /  /
                |  |  /  /
                |  | /  /
                |  |/  ;-._ 
                }  ` _/  / ;
                |  /` ) /  /
                | /  /_/\_/\
                |/  /      |
                (  ' \ '-  |
                 \    `.  /
                  |      |
                  |      |
          
          ======恭喜你，你赢了=======
        '''
    )

def lose():
     print(
        '''
           ======YOU LOSE=======
        
    
                

                   .-"      "-.
                  /            \
                 |              |
                 |,  .-.  .-.  ,|
                 | )(__/  \__)( |
                 |/     /\     \|
       (@_       (_     ^^     _)
  _     ) \_______\__|IIIIII|__/__________________________
 (_)@8@8{}<________|-\IIIIII/-|___________________________>
        )_/        \          /
       (@           `--------`
       
       
       
          ======YOU LOSE=======
        '''
    )


def game_over():
    print(
        '''
           ======GAME OVER=======
        
             _________ 
            / ======= \ 
           / __________\ 
          | ___________ | 
          | | -       | | 
          | |         | | 
          | |_________| |________________ 
          \=____________/                ) 
          / """"""""""" \               / 
         / ::::::::::::: \          =D-' 
        (_________________) 

       
          ======GAME OVER=======
        '''
    )

def show_team():
    print('''             本游戏由666团队制作。
              想知666团队都有谁？
              请持续关注我们的游戏，你就会发现我们的小秘密哦!''')

def show_instruction():
    print('''            玩家先输入一个正整数，并将这个正整数作为神秘数字，由计算机来猜测玩家输入的数字是什么。
             由玩家输入神秘数字的上界，计算机将在这个范围之内进行猜测。
             计算机每输入一个猜测值，由玩家告诉计算机，猜大了，小了还是对了。
             若计算机在限制次数内猜中，则玩家失败，否则则玩家成功。''')
   
def menu():
    print('''             =====游戏菜单=====
                1. 游戏说明
                2. 开始游戏
                3. 退出游戏
                4. 制作团队
             =====游戏菜单=====''')
             
def guess_game():
    number = int(input('请输入神秘整数。'))
    n = int(input('请输入一个大于0的整数，作为神秘整数的上界，回车结束。'))
    max_times = math.ceil(math.log(n, 2))
    guess_times = 0
    a = 1
    b = n
    
    while guess_times < max_times:
        guess = random.randint(a,b)
        guess_times += 1
        print('计算机猜测神秘整数是',guess)
        print('计算机一共可以猜', max_times, '次')
        print('计算机已经猜了', guess_times, '次')
        
        if guess == number:
            lose()
            print('计算机比标准次数少', max_times-guess_times, '次')
            break
        else:
            print('''                     如果计算机猜小了，请输入数字1.
                     如果计算机猜大了，请输入数字2.''')
            x = int(input('请输入1或2，以回车结束。'))
            if x == 1:
                a = guess
                b = n
            elif x == 2:
                a = a
                b = guess
            else :
                print('输入无效。')
               
    else:
        print('您真棒！计算机猜测不到您设置的神秘数字。')
        win()


def main():
    while True:
        menu()
        choice = int(input('请输入你的选择'))
        if choice == 1:
            show_instruction()
        elif choice == 2:
            guess_game()
        elif choice == 3:
            game_over()
            break
        else:
            show_team()


if __name__ == '__main__':
    main()
```

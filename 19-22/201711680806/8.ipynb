```python
import random 

def game_over():
    print(
        '''
           ======GAME OVER=======
        
             _________ 
            / ======= \ 
           / __________\ 
          | ___________ | 
          | | -       | | 
          | |         | | 
          | |_________| |________________ 
          \=____________/                ) 
          / """"""""""" \               / 
         / ::::::::::::: \          =D-' 
        (_________________) 

       
          ======GAME OVER=======
        '''
    )

def show_team():
    print('''            本游戏由666团队制作。
              想知666团队都有谁？
              请持续关注我们的游戏，你就会发现我们的小秘密哦!''')

def show_instruction():
    print('''           成语接龙游戏''')
   
def menu():
    print('''             =====游戏菜单=====
                1. 游戏说明
                2. 开始游戏
                3. 退出游戏
                4. 制作团队
             =====游戏菜单=====''')


def guess_game():
    with open(r'C:\temp\成语大全.txt') as fh:          
        texts = fh.read()
        text = ''
        text = texts.replace('、',' ')
        idioms = text.split()                   
    
    com_score = 0           
    my_score = 0
    
    idiom1 = random.choice(idioms)           
    print(idiom1)
    
    j = 0
    while j >= 0:
        idiom2 = input()
        j += 1
        if idiom2 in idioms and idiom2[0] == idiom1[-1]:
            my_score += 10               
            for idiom in idioms:
                i = 1
                if idiom[0] == idiom2[-1]:
                    com_score += 10
                    idiom1 = idiom
                    print(idiom1)
                    break
                else:
                    i += 1
                    if i == len(idioms):
                        game_over()
                        print('你的分数是',my_score)
                        print('计算机的分数是',com_score)  
                        break
        else :
            com_score += 10
            game_over()
            print('你的分数是',my_score)
            print('计算机的分数是',com_score)     
            break
    return

    
def main():
    while True:
        menu()
        choice = int(input('请输入你的选择'))
        if choice == 1:
            show_instruction()
        elif choice == 2:
            guess_game()
        elif choice == 3:
            game_over()
            break
        else:
            show_team()


if __name__ == '__main__':
    main()     
```

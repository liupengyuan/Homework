```python
import random,bisect,time

A = [i for i in range(100001)]
B = [j for j in range(100001)]
random.shuffle(B)

C = [k for k in range(100001)]
words = []
words = random.sample(C,1000)

start_time_a = time.time()
for word in words:    
    i = bisect.bisect_left(A,word) 
    print(A[i],'在序列A中')
end_time_a = time.time()
print('二分法总共耗时{:.2f}秒'.format( end_time_a - start_time_a))


start_time_b = time.time()
for word in words:  
    if word in B:
        print(word,'在序列B中')    
end_time_b = time.time()
print('顺序法总共耗时{:.2f}秒'.format( end_time_b - start_time_b))       
```

```python
4、挑战性任务：统计本任务中的语料.txt文件，得到每一个词对应不同词性的频次，可表示为：{词:[['n',1000], ['v', 98]...],...}，
   即在dict对象中，key为词，value为列表，输出词频最高的50个条目。
```

```python
from collections import Counter

def cixing(n):
    with open(r'c:\temp\test.txt') as fh:
        text = fh.read()
        word_freq_dict = {}
    for word in text:
        words = [word.split('/')[1] for word in text.split()]
    for w in words:
        if w == n:
            if word in word_freq_dict:
                word_freq_dict[word] += 1
            else:
                word_freq_dict[word] = 1
    return word_freq_dict

with open(r'c:\temp\test.txt') as fh:
    text = fh.read()
words1 = [word.split('/')[0] for word in text.split()]
a = []
for i in words1:
    b = cixing(i)
    a.append(a,b)
c = dict(a)
print(c)
```

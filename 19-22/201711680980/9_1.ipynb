```python
A=[]
B=[]

import random
for a in range(1,100001):
    A.append(a)

B=A
random.shuffle(B)
C=random.sample(B,k=1000)
    
    
#二分查找算法
import time
import bisect
n=0
start_time1=time.time()
while n<1000:
    print(bisect.bisect_left(A, C[n]))
    n+=1
if n==1000:
    end_time1=time.time()
timeA=(end_time1-start_time1)

#顺序查找算法
m=0
item=0
start_time2=time.time()
while m<1000:
    for b in B:
        if C[m]==b:
            print(B.index(b))                    
    m+=1
end_time2=time.time()
print(item)
timeB=(end_time2-start_time2)


print('利用二分法，一共用了',timeA,'s')
print('利用顺序法，一共用了',timeB,'s')
```

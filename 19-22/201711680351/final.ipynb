```python
#2
def a(m,n):
    
    if m == 0:
        return n+1
    elif m>0 and n==0:
        m = m-1
        n = 1
        return a(m,n)
    elif m>0 and n>0:
        m = m-1
        n = a(m,n-1)
        return a(m,n)

def main():
    m=3
    n=4
    print(a(m,n))
    
if __name__ == '__main__':
    main()
```

```python
#3(b)
import random

def outq(a,b):
    num1 = []
    num2 = []
    an = []
    m = []
    sc = 0
    for i in range(n):
        n1 = random.randint(a,b)
        num1.append(n1)
        
        n2 = random.randint(a,b)
        num2.append(n2)
    
        s = random.randint(1,5)
        m.append(s)
        if s == 1:
            print(n1,'+',n2,'= ?')
        elif s == 2:
            print(n1,'-',n2,'= ?')
        elif s == 3:
            print(n1,'*',n2,'= ?')
        else:
            print(n1,'/',n2,'= ?')
    anw = input('请依次写出答案，以逗号间隔，回车结束:')
    an.append(anw)
    da = an[0]
    for j in range(n):
        if m[j]==1:
            re = num1[j]+num2[j]
            if da[j] == re:
                sc += 5
        elif m[j]==2:
            re = num1[j]-num2[j]
            if da[j] == re:
                sc += 5
        elif m[j]==3:
            re = num1[j]*num2[j]
            if da[j] == re:
                sc += 10
        elif m[j]==4:
            re = num1[j]/num2[j]
            if da[j] == re:
                sc += 10
    print('最终分数为：',sc)
    
a=3
b=25
n=15
outq(a,b)
```

```python
#3(c)
def game():
    n = int(input('请输入题目个数：'))
    sc = 0
    qs = []
    for i in range(n):
        q = input('请出题：')
        qs.append(q)
        print('机器人的回答是：')
        re = int(input('请判断对错，加减法答对扣1，乘除法答对扣2，答错扣3'))
        
        if re == 1:
            sc += 5
        elif re == 2:
            sc += 10
        else:
            sc +=0
    print('最终得分是：',sc)

game()
```

```python
#4(b)
from collections import Counter
def count_words_freq_dict(filename):
    words_freq_dict = Counter()
   
    with open(filename) as f:
        text = f.read()
        words_freq_dict.update([word.split('/')[0] for word in text.split()])
    return words_freq_dict

def main():
    filename = r'd:\temp\成语.txt'
    table = count_words_freq_dict(filename)
    print(table.most_common(20))

if __name__ == '__main__':
    main()
```

```python
a=[]
for i in range(100000+1):
    for k in range(i):
        a.append(i)
        break
b = a[100000-1::-1]

import random, bisect, time
       
def loc(x):
    if x in b:
        for i in range(100000-1):
            while x == b[i]:
                print(x,'这个数在b中所处的位置是',i)
                break
    else:
        print(x,'这个数不在b中')
        
def bi_search(m):
    print(m,'这个数在a中的位置是',bisect.bisect_left(a, m))

start_time = time.time()   
for i in range(1000):
    number = random.randint(1,110001)
    loc(number)
    
    if i == 1000-1:
        end_time = time.time()
        print('——————顺序查找总共处理了{:.2f}秒...'.format(end_time - start_time))
                
for i in range(1000):
    number = random.randint(1,110001)
    bi_search(number)
    if i == 1000-1:
        end_time = time.time()
        print('——————二分查找总共处理了{:.2f}秒...'.format(end_time - start_time))
                
```

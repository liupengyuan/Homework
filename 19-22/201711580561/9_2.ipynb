```python
随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
```

```python
import random
numbers=[]
word_freq_pairs=[]
for i in range(10000):
    number=random.randint(1,100000)
    numbers.append(number)
for word in numbers:
    frequency=numbers.count(word)
    word_freq_pairs.append([word,frequency])
print(sorted(word_freq_pairs))
```

```python
from collections import Counter
def get_word_table(filename):
    word_table = []
    with open(filename) as f:
        text = f.read()
    words = [word.split('/')[0] for word in text.split()]
    for word in words:
        if word not in word_table:
            word_table.append(word)
    return word_table
def count_words_freq(filename, words):
    word_freq_pairs = []
    with open(filename) as f:
        text = f.read()
    for word in words:
        number = text.count(word)
        word_freq_pairs.append([word, number])
    return word_freq_pairs

filename = r'C:\temp\test.txt'

```

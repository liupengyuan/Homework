```python
def my_fac():
    import random,math
    n=int(input('请输入一个整数个数：'))
    m=int(input('请输入一个整数：'))
    k=int(input('请输入一个大于m的整数:'))
    i=0
    total_1=0
    total_2=0
    while i<m:
        i+=1
        number=random.randint(m,k)
        total_1+=math.log(number)
        total_2+=1/math.log(number)
    print(total_1)
    print(total_2)
    
my_fac()
```

```python
import random
def my_fac():
    n=int(input('请输入想要相加的个数，回车结束'))
    i=0
    m=0
    total=0
    a=random.randint(1,9)
    print('随机输出的数字为',a)
    while i<n:
        i+=1
        m=m+10**(i-1)*a
        total=total+m
    print(total)

my_fac()
```

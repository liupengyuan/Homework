```python
import random,time,bisect
A=[]
C=[]
for a in range(1,100001):
    A.append(a)
B=A
random.shuffle(B)
for i in range(1000):
    n=random.randint(1,110000)
    C.append(n)
start_time_1=time.time()
for number_1 in C:
    if number_1 in B:
        for i in range(1,100001):
            if number_1==B[i]:
                print('由顺序法查找得位置为:',i)
                break
    else:
        print('B中不存在该整数')
end_time_1=time.time()
print('顺序法查找耗时为：',end_time_1-start_time_1,'s')

start_time_2=time.time()
for number_2 in C:
    if number_2 in A:
        print('由顺序法查找得位置为:',bisect.bisect_left(A,number_2),'s')
    else:
        print('A中不存在该整数')
end_time_2=time.time()
print('二分法查找耗时为：',end_time_2-start_time_2,'s')
```

```python
import math,random
def sqrt(n,m,k):
    i=0
    j=0
    total=0
    sqrt=0
    while i<n:
        i=i+1
        j=random.randint(m,k)
        total=total+j
        print(j)
    sqrt=math.sqrt(total)
    return sqrt
n=int(input('请输入想要计算的整数个数。'))
m=int(input('请输入随机整数的下限。'))
k=int(input('请输入随机整数的上限。'))
print(sqrt(n,m,k))
```

```python
import math,random
def sum_1(n,m,k):
    i=0
    j=0
    total_1=0
    total_2=0
    while i<n:
        i=i+1
        j=random.randint(m,k)
        if j==1:
            j=random.randint(m,k)
        print(j)
        t=math.log(j)
        total_1=total_1+t
        total_2=total_2+1/t
    return total_1,total_2
n=int(input('请输入想要计算的整数个数。'))
m=int(input('请输入随机整数的下限。'))
k=int(input('请输入随机整数的上限。'))
print(sum_1(n,m,k))
```

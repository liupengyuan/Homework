2、随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。

```python
from collections import Counter
from operator import itemgetter
import random
num_freq_dict=Counter()
list_1=[]
for i in range(10000):
    list_1.append(random.randint(0,100000))
num_freq_dict.update(list_1)
nums=list(num_freq_dict.items())
nums=sorted(nums, key=itemgetter(0))
for i in range(len(nums)):
    print(nums[i][0],',',nums[i][1],'次')
```

4、挑战性任务：统计本任务中的语料.txt文件，得到每一个词对应不同词性的频次，可表示为：{词:[['n',1000], ['v', 98]...],...}，即在dict对象中，key为词，value为列表，输出词频最高的50个条目。

```python
from collections import Counter
from operator import itemgetter
with open(r'c:\temp\语料.txt') as fr:
    words=fr.read()
    words=words.replace('\n',' ')
    words=words.split(' ')
    words_1=[word.split('/')[0] for word in words]
    words_2=[word.split('/') for word in words]
word_freq_dict_1=Counter()
word_freq_dict_1.update(words_1)
word_freq_table_1=list(word_freq_dict_1.items())
word_freq_table_1=sorted(word_freq_table_1,key=itemgetter(1),reverse=True)
word_freq_table_2=[]
for i in range(50):
    word_freq_table_2.append(word_freq_table_1[i])
word_table_1=[]
word_table_2=[]
for i in word_freq_table_2:
    for j in words_2:
        if j[0]==i[0]:
            word_table_1.append(j[1])
    word_table_2.append(tuple(word_table_1))
    word_table_1=[]
word_freq_dict_2=Counter()
word_freq_table_3=[]
for i in word_table_2:
    word_freq_dict_2.update(i)
    word_freq_table_3.append(list(word_freq_dict_2.items()))
    word_freq_dict_2=Counter()
word_freq_table_4={word_freq_table_2[i][0]:word_freq_table_3[i] for i in range(len(word_freq_table_2))}
print(word_freq_table_4)
```

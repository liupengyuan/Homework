```python
#随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
import random
a=[]

for i in range(1,10001):
    number=random.randrange(1,100000)
    a.append(number)
print(a)
print('\n'*6)
number_freq_pairs=[]
for number in a:
    for item in number_freq_pairs:
        if number==number_freq_pairs[0]:
            number_freq_pairs[1]+=1
    else:
        number_freq_pairs.append([number,1])
print(sorted(number_freq_pairs))
    
    
```

```python
#挑战性任务：统计本任务中的语料.txt文件，得到每一个词对应不同词性的频次,输出词频最高的50个条目
from collections import Counter
from operator import itemgetter
def count_words_freq_dict(filename):
    words_freq_dict = Counter()
    
    with open(filename) as f:
        for line in f:
            words_freq_dict.update([word for word in line.split()]) 
    return words_freq_dict
def main():
    filename = r'c:\temp\语料.txt'
    table = count_words_freq_dict(filename)
    print(table.most_common(50))
    print(sorted(table,key=itemgetter(0)))

          
if __name__ == '__main__':
    main()
```

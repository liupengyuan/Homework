```python
import random,time,bisect

def time1():
    start_time1=time.time()
    for num1 in C:
        if num1 in str_B:
            print(num1)
    end_time1=time.time()
    print('顺序查找法的时间是:',end_time1-start_time1)
    print('-----------------------------------------')
    
def time2():
    start_time2=time.time()
    for num2 in C:
        if num2 in str_A:
            print(str_A[bisect.bisect(str_A,num2)])
    end_time2=time.time()
    print('二分查找法的时间是:',end_time2-start_time2)
    
        
str_A=[]
for i in range(1,100001):
    str_A.append(str(i))


str_B=[]
for n in range(1,100001):
    n=random.randint(1,100001)
    str_B.append(str(n))
C=random.sample(str_A,1000)    
time1()
time2()
    


```

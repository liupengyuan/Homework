```python
def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
            
    return ch_table


# 主程序
fh = open(r'C:\temp\idioms.txt')
text = fh.read()
chs = get_ch_table(text.replace('\n', ','))
import random

a=10
b=1
idioms = text.split()
while a<100:
    for i in range(a):
        ch = random.choice(chs)

        
        print(ch)
    a+=10
    b+=1
    print('txt',b)
```

```python
#2、随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
```

```python
import random
list = []
i = 0
while i < 10000:
    num = random.randint(0, 100000)
    list.append(num)
    i = i+1


def count_numbers_freq():
    numbers = []
    for num in list:
        if num not in numbers:
            numbers.append(num)
    A = [str(i) for i in list]
    B = [str(i) for i in numbers]
    
    numbers_n = []
    for number in B:
        n = A.count(number)
        t = [number,n]
        numbers_n.append(t)
    print(numbers_n)     
count_numbers_freq()


```

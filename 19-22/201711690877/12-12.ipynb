```python
def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table

# 主程序
fh = open(r'C:\Users\Cyy19990311\Desktop\idioms.txt ')
text = fh.read()
chs = get_ch_table(text.replace('\n', ''))

print(len(chs), chs)
fh.close()
```

```python
import random
n=int(input('文件序号为：'))
idioms=[]
def rewrite():
    for i in range(n*10):
        idiom= random.choice(chs)
        idioms.append(idiom)
        print(idiom)
        
    fh = open(r'C:\Users\Cyy19990311\Desktop\n.txt', 'w')
    fh.writelines(idioms)
    fh.close()
    
rewrite()

```

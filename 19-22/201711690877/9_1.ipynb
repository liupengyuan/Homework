```python
import random
import time
import bisect
list_a = list(range(1,100001))

list_b = random.sample(list_a,10000)


list_all = list(range(1,1100001))
find_list = random.sample(list_all,1000)
print(find_list)
start_time1 = time.time()
for n in find_list:
    if n in list_b:
        print(n)
end_time1 = time.time()
t1 = end_time1-start_time1
print('顺序查找法处理了',t1,'秒')
start_time2 = time.time()
for m in find_list:
    if bisect.bisect_left(list_a, m):
        print(m)
end_time2 = time.time()
t2 = end_time2-start_time2
print('二分查找法处理了',t2,'秒')
```

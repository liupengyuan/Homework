```python
#3.
#(a,b)
import random
i=0
n=15
a=3
b=25
score=0
while i<n:
    num_1=random.randint(a,b)
    num_2=random.randint(a,b)
    sym=random.randint(1,4)
    i+=1
    if sym==1:
        print(num_1+num_2=)
        answer=int(input('请用户输入答案，回车结束。'))
        if answer==num_1+num_2:
            score+=5

    elif sym==2:
        if num_1>num_2:
            print(num_1-num_2=)
            answer=int(input('请用户输入答案，回车结束。'))
            if answer==num_1-num_2:
                score+=5
            else:
                score+=0

        else:
            print(num_2-num_1=)
            answer=int(input('请用户输入答案，回车结束。'))
            if answer==num_2-num_1:
                score+=5
            else:
                score+=0


    elif sym==3:
        print(num_1*num_2=)
        answer=int(input('请用户输入答案，回车结束。'))
        if answer==num_1*num_2:
            score+=10
        else:
            score+=0


    else:
        print(num_1//num_2=)
        answer=int(input('请用户输入答案，回车结束。'))
        if answer==num_1//num_2:
            score+=10
        else:
            score+=0

print(score)
```

```python
import random,time,bisect

A=[]
B=[]
for i in range(1,100001):
    A.append(i)
B=A    
random.shuffle(B)

C=[]
for j in range(1000):
    C.append(random.randint(1,110001))

#顺序法
print('顺序法')
D=[]
start1=time.time()
for k in B:
    if k in C:
        D.append(k)
end1=time.time()
print('一共用时',end1-start1,'秒')
print(D)

#二分法  
print('二分法')
E=[]
start2=time.time()
for n in C:
    if n <=100000:
        m=bisect.bisect(A,n+1)
        E.append(m)
end2=time.time()
print('一共用时',end2-start2,'秒')
print(E)
```

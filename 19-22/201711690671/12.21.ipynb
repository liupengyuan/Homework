```python
import random
fh = open(r'c:\temp\idioms_correct.txt')
text = fh.read()

def get_n_table(line):
    n_table = []
    for n in line:
        if n not in n_table:
            n_table.append(n)
    return n_table

line=text.replace('\n','')
fhs=get_n_table(line)

a1=''
for i in range(10):
    b=random.choice(fhs)
    a1+=b
print(a1)
idioms=[a1]
fh_1 = open(r'c:\temp\1.txt', 'w')
fh_1.writelines(idioms)

a2=''
for i in range(20):
    b=random.choice(fhs)
    a2+=b
print(a2)
idioms=[a2]
fh_2 = open(r'c:\temp\2.txt', 'w')
fh_2.writelines(idioms)
                
a3=''
for i in range(30):
    b=random.choice(fhs)
    a3+=b
print(a3)
idioms=[a3]
fh_3 = open(r'c:\temp\3.txt', 'w')
fh_3.writelines(idioms)

a4=''
for i in range(40):
    b=random.choice(fhs)
    a4+=b
print(a4)
idioms=[a4]
fh_4 = open(r'c:\temp\4.txt', 'w')
fh_4.writelines(idioms)

a5=''
for i in range(50):
    b=random.choice(fhs)
    a5+=b
print(a5)
idioms=[a5]
fh_5 = open(r'c:\temp\5.txt', 'w')
fh_5.writelines(idioms)

a6=''
for i in range(60):
    b=random.choice(fhs)
    a6+=b
print(a6)
idioms=[a6]
fh_6 = open(r'c:\temp\6.txt', 'w')
fh_6.writelines(idioms)

a7=''
for i in range(70):
    b=random.choice(fhs)
    a7+=b
print(a7)
idioms=[a7]
fh_7 = open(r'c:\temp\7.txt', 'w')
fh_7.writelines(idioms)

a8=''
for i in range(80):
    b=random.choice(fhs)
    a8+=b
print(a8)
idioms=[a8]
fh_8 = open(r'c:\temp\8.txt', 'w')
fh_8.writelines(idioms)

a9=''
for i in range(90):
    b=random.choice(fhs)
    a9+=b
print(a9)
idioms=[a9]
fh_9 = open(r'c:\temp\9.txt', 'w')
fh_9.writelines(idioms)

a10=''
for i in range(100):
    b=random.choice(fhs)
    a10+=b
print(a10)
idioms=[a10]
fh_10 = open(r'c:\temp\10.txt', 'w')
fh_10.writelines(idioms)

fh.close()
```

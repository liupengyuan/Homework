```python
import random

def begin():
    print('welcome!')
def win():
    print('You are winner! ')
def lose():
    print('You are loser! ')
def over():
    print('Game Over.')
def made():
    print('made by (～￣▽￣)～')
def menu():
    print('''
              =====游戏菜单=====
                1. 开始游戏
                2. 退出游戏
                3. 制作团队
              =====游戏菜单=====''')
    


def game():
    fh=open(r'c:\temp\成语大全.txt')
    text = fh.read()
    idioms = text.split()
    
    grade_man = 0
    grade_com = 0
    a1=[]
    a=random.choice(idioms)
    a1.append(a)
    print(a)
    while grade_man>=0:  
        b=input('请输入一个首字与这个成语的尾字相同的成语')
        if a[3]!=b[0]:
            lose()
            break
        else:
            grade_man+=10
            aa=[]
            for idiom in idioms:
                if idiom[0]==b[3]:
                    aa.append(idiom)
            if len(aa)==0:
                win()
                break
            else:
                grade_com+=10
                a=random.choice(aa)
                print(a)

    print('你的分数是',grade_man)
    print('电脑的分数是',grade_com)
    over()
                   
def main():
    while True:
        menu()
        choice = int(input('请输入你的选择'))
        if choice == 1:
            game()
        elif choice == 2:
            over()
            break
        else:
            made() 
        
main()
```

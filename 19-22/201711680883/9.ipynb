```python
import random
import random, math
import time, bisect

A = []
i = 1
while i < 100001:
    A.append(i)
    i += 1
    
B = []
i = 1
while i < 100001:
    B.append(i)
    i += 1
random.shuffle(B)

C = []
m = 0
while m < 1000:
    n = random.randint(1,100000)
    C.append(n)
    m += 1
    
start_time = time.time()

for l in range(len(B)):
    if l in C:
        print (l)
end_time = time.time()
print(end_time - start_time)

def bi_search(number, numbers, low, high):
    if low > high:
        return -1
    else:
        mid = (low+high)//2
        if number == numbers[mid]:
            return mid
        elif number > numbers[mid]:
            return bi_search(number, numbers, mid+1, high)
        else:
            return bi_search(number, numbers, low, mid-1)
        
start_time = time.time()

for k in range(len(C)):
    print(bi_search(k, C, 0, 9))
end_time = time.time()
print(end_time - start_time)    
```

```python
import math
import random

n=int(input('pls input n: '))
m=int(input('pls input m: '))
k=int(input('pls input k: '))

if m>k:
    tmp=m
    m=k
    k=tmp

i=0
tot=0
while i<n:
    random.seed(1)
    tot+=random.randint(m,k)
    i+=1
avg=tot/n
print('The square root of the average of these integer is: ',math.sqrt(avg))
```

```python
import math
import random

n=int(input('pls input n: '))
m=int(input('pls input m: '))
k=int(input('pls input k: '))

if m>k:
    tmp=m
    m=k
    k=tmp

i=0
totlog=0
totdenolog=0

while i<n:
    random.seed(2)
    rlog=math.log(random.randint(m,k))
    while rlog==0:
        rlog=math.log(random.randint(m,k))

    totlog+=rlog
    totdenolog+=1/rlog
    i+=1
print('sigma of log is: ',totlog,'\n')
print('sigma of 1/log is: ',totdenolog,'\n')
```

```python
def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table

fh = open(r'c:\temp\idioms_correct.txt')
text = fh.read()
chs = get_ch_table(text.replace('\n', ''))
```

```python
import random
rch10=[]

for i in range(10):
    rch=random.choice(chs)
    print(rch)
    rch10.append(rch)
fh = open(r'c:\temp\1.txt', 'w')
fh.write(''.join(rch10))
fh.close()
```

```python
for n in range(1,11):
    rchlist=[]
    for i in range(10*n):
        rch=random.choice(chs)
        print(rch)
        rchlist.append(rch)
    print(len(rchlist))
    fh=open(r'c:\\temp\\'+str(n)+'.txt','w')
    fh.write(''.join(rchlist))
    fh.close()
```

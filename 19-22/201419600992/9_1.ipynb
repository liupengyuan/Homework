```python
import random
import bisect
import time

def RandomNumbers(st,end,iter):
    rnums=[]
    for i in range(iter):
        rnums.append(random.randint(st,end))
    return rnums
 
def CreateNumberList(stnum,endnum):
    numbers=[]
    for n in range(stnum,endnum+1):
        numbers.append(n)
    return numbers

def SeqSearch(wFind,tFind):
    foundNums=[]
    for i in tFind:
        stat=wFind.count(i)
        if stat!=0:
            foundNums.append(i)
    return foundNums

def BiSearch(w,t):
    fNums=[]
    for j in t:
        stat=bisect.bisect_left(w,j)
        if stat>0 and stat<len(w):
            fNums.append(j)
    return fNums

listO=CreateNumberList(1,100000)

listB=[]
random.shuffle(listO)
listB=listO
#print(listB)
listA=CreateNumberList(1,100000)

randomList=RandomNumbers(1,110000,1000)

seqStartTime=time.time()
seqFound=SeqSearch(listB,randomList)
seqEndTime=time.time()
seqTime=seqEndTime-seqStartTime

#print(randomList)
biStartTime=time.time()
biFound=BiSearch(listA,randomList)
biEndTime=time.time()
biTime=biEndTime-biStartTime
#print(biFound)

if len(biFound)==len(seqFound):
    print("The result of Sequence Search is:",seqFound,"\nTime used:",seqTime,"\n\nThe result of Bisect is:",biFound,"\nTime used:",biTime)
else:
    print("something error...")
```

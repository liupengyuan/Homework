```python
#2
import random

def CreateRandomNumbers(lr,hr,iter):
    rnums = []
    for i in range(iter):
        rnums.append(random.randint(lr,hr))
    return rnums

def CountNumberFreq(anylist):
    nFreqDict = {}
    for ele in anylist:
        if ele in nFreqDict:
            nFreqDict[ele] += 1
        else:
            nFreqDict[ele] = 1
    return nFreqDict

numberList = CreateRandomNumbers(0,100000,10000)
numberFreqDict = CountNumberFreq(numberList)

testn = 0
for number in sorted(numberFreqDict):
    print (number,numberFreqDict[number])
    testn += numberFreqDict[number]
if testn == 10000:
    print("done")
else:
    print(testn)
```

```python
#3
def GetWords(filename):
    wordTable = []
    with open(filename) as f:
        text = f.read()
    words = [word.split('/')[0] for word in text.split()]
    for word in words:
        if word not in wordTable:
            wordTable.append(word)
    return wordTable

def GetChns(anytable):
    mystr = ''.join(anytable)
    chns = set(mystr)
    return chns

filename = r'd:\\temp\\test.txt'
table = GetWords(filename)
chset = GetChns(table)
#print(chset)

filename1 = r'd:\\temp\\test1.txt'
table1 = GetWords(filename1)
chset1 = GetChns(table1)
#print(chset1)

commonTable = []
singleTable = []
for ch in chset1:
    if ch in chset:
        commonTable.append(ch)
    elif ch not in chset:
        singleTable.append(ch)

print(":\n",commonTable)
print(":\n",singleTable)
```

```python
#4
import operator
from operator import itemgetter
def CountWordAttris(filename):
    wordsDict = {}
    wordsCountDict = {}
    wordsChunk = []
    with open(filename) as f:
        text = f.read()
    for chunk in text.split():
        if chunk not in wordsChunk:
            wordsChunk.append(chunk)
    for wordChunk in wordsChunk:
        wordAttriNum = text.count(wordChunk)
        word = wordChunk.split('/')[0]
        if word not in wordsDict:
            wordsDict[word] = []
            wordsDict[word].append([wordChunk.split('/')[1],wordAttriNum])
            wordsCountDict[word] = wordAttriNum
        else:
            wordsDict[word].append([wordChunk.split('/')[1],wordAttriNum])
            wordsCountDict[word] += wordAttriNum
    return wordsDict,wordsCountDict
 
myfilename = r'd:\\temp\\test.txt' # '.txt'Memory Error,test.txt
mywordsDict = CountWordAttris(myfilename)[0]
mywordsCount = CountWordAttris(myfilename)[1]
sortedCount = sorted(mywordsCount.items(),key=operator.itemgetter(1),reverse=True)
i = 0
for pair in sortedCount:
    if i < 50:
        print(pair[0],mywordsDict[pair[0]])
        i += 1
    else:
        break
```

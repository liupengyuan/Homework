```python
#成语接龙，没有判断用户输入的成语是否存在于给定成语大全.txt中，对于答案正确性的检查只有长度和第一个汉字

import random
import sys

def win():
    print('YOU WIN!')
    
def lose():
    print('YOU LOSE')
     
def idiom_robot(file_name):
    with open(file_name) as fh:
        text=fh.read()
    text=text.replace('、','')
    idioms=text.split('\n')

    com_score=0
    user_score=0
    start_idiom=random.choice(idioms)
    print('GAME START(press q to exit during the game)')
    print(start_idiom)
    begin_ch=start_idiom[3:4]

    while True:
        
        user_idiom=input('Now it is your turn: ')
        last_ch=''
        
        if user_idiom=='q':
            print('Final scores are: Computer:',com_score,'; User:',user_score)
            break
        elif len(user_idiom)!=4:
            print('This is not a idiom!')
            print('Final scores are: Computer:',com_score,'; User:',user_score)
            lose()
            break
        elif user_idiom[0:1]!=begin_ch:
            print('This is not the correct idiom!')
            print('Final scores are: Computer:',com_score,'; User:',user_score)
            lose()
            break
        else:
            user_score+=10
            last_ch=user_idiom[3:4]

        com_idioms=[]

        for idiom in idioms:
            if idiom.startswith(last_ch):
                com_idioms.append(idiom)

        if len(com_idioms)==0:
            print('Cant find an idiom to answer.')
            print('Final scores are: Computer:',com_score,'; User:',user_score)
            break

        else:
            ans_idiom=random.choice(com_idioms)
            print(ans_idiom)
            com_score+=10
            begin_ch=ans_idiom[3:4]

def main():
    while True:
        f_name=r'c:\temp\成语大全.txt'
        idiom_robot(f_name)
        stat=input('Do you want to exit? (press q to exit, otherwise continue...)')
        if stat=='q':
            break

if __name__=='__main__':
    main()
```

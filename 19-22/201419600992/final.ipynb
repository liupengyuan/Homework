```python
def printStar(patt,stl):
    for row in range(stl):
        print(' '*(stl-row-1)*2,end='')
        i=0
        while i<row+1:
            print(patt,end='')
            print(' ',end='')
            i+=1
        print(patt*5,end='')
        i=0
        while i<row+1:
            print(' ',end='')
            print(patt,end='')
            i+=1
        print('')

mypatt='*'
myline=5
printStar(mypatt,myline)
```

```python
def Ackermann(m,n):
    if m==0:
        a=n+1
        return a
    elif m>0 and n==0:
        return Ackermann(m-1,1)
    else:
        return Ackermann(m-1,Ackermann(m,n-1))

print(Ackermann(3,4))
```

```python
def CountWords(filename):
    wordsChunk=[]
    attrisDict={}
    with open(filename) as f:
        text=f.read()
    for chunk in text.split():
        if chunk not in wordsChunk:
            wordsChunk.append(chunk)
        for wordChunk in wordsChunk:
            wordAttriNum=text.count(wordChunk)
            attri=wordChunk.split('/')[1]
            if attri not in attrisDict:
                attrisDict[attri]=wordAttriNum
            else:
                attrisDict[attri]+=wordAttriNum
    return attrisDict

myfilename = r'd:\\temp\\test_re_gbk.txt'
myAttrisDict=CountWords(myfilename)
for a in myAttrisDict:
    print(a,myAttrisDict[a])
```

```python
def get_word_table(filename):
   
    word_table = []
    with open(filename) as f:
        text = f.read()
    words = [word.split('/')[0] for word in text.split()]
    for word in words:
        if word not in word_table:
            word_table.append(word)
    return word_table
```

```python
#练习 1：写函数，求n个随机整数均值的平方根，整数范围在m与k之间（n,m,k由用户输入）
import random ,math

m=int(input('请输入一个整数，回车结束。'))
k=int(input('请输入一个整数，回车结束。'))
if m>k:
    m,k=k,m
n=int(input('请输入希望的整数间隔，回车结束。'))
i=1
sum=0
while i<=n:
    i+=1
    l=random.randint(m,k)
    sum=sum+l
    print(l)
j=int(sum/n)
print(math.sqrt(j))      
```

```python
#写函数，共n个随机整数，整数范围在m与k之间，（n,m,k由用户输入）。求1：西格玛log(随机整数)，2：西格玛1/log(随机整数)
import math, random
def sgm():
    n = int(input('请输入希望的整数的个数，回车结束。'))
    m = int(input('请输入整数，回车结束。'))
    k = int(input('请输入整数，回车结束。'))
    if m>k:
        m,k=k,m
    j = 0
    g = 0
    i = 0
    y = 0
    while i < n:
        i = i + 1
        j = random.randint(m,k)
        g = g + math.log(n,j)
        y = y +1/math.log(n,j)
    print(g)
    print(y)
        
sgm()
```

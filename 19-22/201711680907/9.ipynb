```python
A=[i for i in range(1,100001)]
```

```python
import random
B=[]
while True:
    if A==[]:
        break
    else :
        ch=random.choice(A)
        B.append(ch)
        A.remove(ch)
```

```python
import random
numbers=[]
for i in range(1000):
    n=random.randint(1,110000)
    numbers.append(n)
```

```python
import time
num_wz=[]
start_time=time.time()
for number in numbers:
    for i in range(len(B)):
        if number==B[i]:
            num_wz.append([number,i])
end_time=time.time()
print(num_wz)
n=end_time-start_time
print('共处理',n,'秒')
```

```python
import bisect,time
A=[i for i in range(1,100001)]
num_w=[]
start_time2=time.time()
for number in numbers:
    n=bisect.bisect_left(A,number)
    num_w.append([number,n])
end_time2=time.time()
m=end_time2-start_time2
print(num_w)
print('共用时',m,'秒')
```

```python
def que(a,b):
    import random
    list1=['+','-','*','/']
    question=[]
    num=0
    while True:
        x=random.randint(a,b+1)
        y=random.randint(a,b+1)
        z=random.choice(list1)
        if z=='/':
            if x%y==0:
                question.clear()
                question.append(x)
                question.append(z)
                question.append(y)
                question.append('=')
                print(question)
                break
        else:
            question.clear()
            question.append(x)
            question.append(z)
            question.append(y)
            question.append('=')
            print(question)
            break
```

```python
#(a)
a=3
b=25
n=15
questions=[]
for i in range(n):
    que(a,b)
    questions.append(que(a,b))
```

```python
#(b)
def result(questions):
    results=[]
    x=questions[1]
    for question in questions:
        if x=='+':
            m=question[0]
            n=question[2]
            re=m+n
        elif x=='-':
            m=question[0]
            n=question[2]
            re=m-n
        elif x=='*':
            m=question[0]
            n=question[2]
            re=m*n
        else:
            m=question[0]
            n=question[2]
            re=m/n
        results.append(re)
        return results


def grade(questions):
    x=questions[1]
    for question in questions:
        if x=='+' or '-':
            grade=5
        else:
            grade=10
        return grade
    
results=result(questions)
answers=input('一次输入所有的答案，逗号分隔')   
final_grade=0
for i in range(n):
    if results[i]==answers[i]:
        final_grade=final_grade+grade(questions)
print('最终得分为',final_grade)
```

```python
#(b)
with open(r'C:\Users\Administrator\Downloads\idioms.txt') as f:
    text=f.read()
text = text.replace('、', '')
idioms = text.split('\n')

def count_words_freq_dict(filename):
    words_freq_dict = {}
    with open(filename) as f:
        for line in f:
            words = [word.split('/')[0] for word in line.split()]
            for word in words:
                if word in words_freq_dict:
                    words_freq_dict[word] += 1
                else:   
                    words_freq_dict[word] = 1
    return words_freq_dict
filename= r'C:\Users\Administrator\Downloads\test_re_gbk.txt'
table=count_words_freq_dict(filename)

same=[]
for idiom in idioms:
    for word in table:
        if idiom==word:
            same.append()
```

```python
#(c)        
def count_cileis_freq_dict(filename):
    cileis_freq_dict = {}
    
    with open(filename) as f:
        for line in f:
            cileis = [word.split('/')[1] for word in line.split()]
            for cilei in cileis:
                if cilei in cileis_freq_dict:
                    cileis_freq_dict[cilei] += 1
                else:   
                    cileis_freq_dict[cilei] = 1
                    
    return cileis_freq_dict


filename = r'C:\Users\Administrator\Downloads\test_re_gbk.txt'
table = count_cileis_freq_dict(filename)
print(table)
```

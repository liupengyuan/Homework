```python
# 练习 1：写函数，求n个随机整数均值的平方根，整数范围在m与k之间（n,m,k由用户输入）。
import random,math
i=0
p=0
n=int(input('请输入想输入的整数个数，以回车键结束。'))
m=int(input('请输入想输入数字的左边界，以回车键结束。'))
k=int(input('请输入想输入数字的右边界，以回车键结束。'))
number=random.randint(m,k)
while i<n:
    i+=1
    p+=number
print(math.sqrt(p/n))
```

```python
import random

def win():
    print('congratulations,你赢了')
def lose():
    print('sorry,你输了')
def deuce():
    print('哈哈，平局了')
def game_over():
    print('game over,游戏结束')

def team():
    print('制作团队：sharon')
def instruction():
    print('用户和计算机初试分数均为0，由计算机出第一个成语，用户与计算机交替进行成语接龙，回答正确加10分，回答错误不扣分，但游戏直接结束，最终人和计算机分高者胜。')
def menu():
    print('''
             --------游戏菜单--------
                    1.游戏说明
                    2.开始游戏
                    3.退出游戏
                    4.制作团队
             --------游戏菜单--------''')
    
def game():
    fh=open(r'C:\temp\idiom.txt')
    text=fh.read()
    fh.close()
    idioms=text.split('\n')
    hum=0 
    com=0 
    i=1
    a=random.choice(idioms)
    print(a)
    memory = set()
    memory.add(a)
    while i>0:
        b=input('输入要接的成语，回车结束：')
    
        if b.strip() == '':
            break
        if b == False:
            print("游戏结束！该成语不存在")
            break
        if b in memory:
            print("游戏结束！该成语已被使用过")  
            break
        if b[0]==a[3]:
            hum=hum+10
            memory.add(b)
            for a in idioms:
                if a == False:
                    print("游戏结束！该成语不存在")
                    break
                if a[0]==b[3]:
                    print(a)
                    com=com+10
                    break
                
        
            i=i+1
        else:
            print('人类得分：',hum)
            print('计算机得分：',com)

        
    print('人类得分：',hum)
    print('计算机得分：',com)        
    if hum>com:
        win()  
    if hum==com:
        deuce()
    else:
        lose()

def main():
    while True:
        menu()
        choice = int(input('请输入你的选择'))
        if choice == 1:
            instruction()
        elif choice == 2:
            game()
            
        elif choice == 3:
            game_over()
            break
        else:
            team()


if __name__ == '__main__':
    main()
                
```

```python
#练习1：求n个随机整数均值的平方根，整数范围在m与k之间(n,m,k由用户输入)

import random 


n=int(input('请输入想得到的随机整数的个数，回车结束'))
m=int(input('请输入一个整数，回车结束'))
k=int(input('请输入一个大于m的整数，回车结束'))
def total():    
    total=0
    i=0
    number=0
    while i<n:
        i+=1
        number=random.randint(m,k)
        total=total+number
    return total
    
print((total()/n)**(1/2))
```

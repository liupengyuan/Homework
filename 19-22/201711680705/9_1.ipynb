```python
import random,bisect,time

def time_1(B,C):
    start_time1=time.time()
    for file_object1 in C:
        if file_object1 in B:
            print(file_object1)
    end_time1=time.time()
    print("顺序法的时间是:",end_time1-start_time1)

    
def time_2(A,C):
    start_time2=time.time()
    for file_object2 in C:
        print(A[bisect.bisect_left(A, file_object2)])
    end_time2 = time.time()
    print("二分法的时间是:",end_time2-start_time2)

    
B = [i for i in range(100001)]
random.shuffle(B)
A = [j for j in range(100001)]
C = random.sample(A,1000)
time_1(B,C)
time_2(A,C)
```

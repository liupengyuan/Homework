```python
def rev(m):
    line=m
    n=len(line)
    for i in range(n,0,-1):
        print(line[i-1],end='')

k=str(input('whatever you like:'))
rev(k)
```

```python
def lx(m,n):
    for i in range(1, n):
         for k in range(n-1-i,-1,-1):
            print(' ', end='')
         for j in range(1,i+1):
            print(m,'',end='')
         print()
    for i in range(1,n+1):
        for k in range(n+1-i,0,-1):
            print(m,'',end='')
        print()
        for j in range(0,i):
            print(' ',end='')
    print('第一种菱形')
    
    for i in range(1, n):
        for k in range(n-i,1,-1):
            print('  ', end='')
        for j in range(1,n):
            print('*',' ',end='')
        print()
    print('第二种菱形')
    
    for i in range(1, n):
         for k in range(3*n-i,2*i,-1):
            print(' ', end='')
         for j in range(0,3*(i)-2):
            print('*','',end='')
         print()
    
    for m in range(1,n+1):
        for k in range((3*n+1)-3*m,0,-1):
            print('*','',end='')
        print()
        for j in range(0,m*3):
            print(' ',end='')
    print('第三种菱形')

            
q=str(input('a word or a symbol'))
p=int(input('a number'))
lx(q,p)
```

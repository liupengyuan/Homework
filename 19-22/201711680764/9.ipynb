```python
import random,time,bisect
A=[]
for i in range(1,100001):
    A.append(i)
B=random.sample(A,100000)

    
def bi_search(number, numbers, low, high):
    if low > high:
        return -1
    else:
        mid = (low+high)//2
        if number == numbers[mid]:
            return mid
        elif number > numbers[mid]:
            return bi_search(number, numbers, mid+1, high)
        else:
            return bi_search(number, numbers, low, mid-1)
            

t1=time.clock()
for j in range(0,1000):
    k=random.randint(1,110000)
    nums =A
    nums.sort()
    print(bi_search(k, nums, 0, 99999))
t2=time.clock()
print(t2-t1)


tm3=time.clock()
for p in range(0,1000):
    m=random.randint(1,110001)
    if m in B:
        print(m)
tm4=time.clock()
print(tm4-tm3)
```

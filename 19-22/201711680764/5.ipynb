```python
#练习二
n=int(input('please input a number'))
m=int(input('please input a number'))
k=int(input('please input a number,bigger than the last number'))
def sum():
    import math ,random
    
    time=0
    total=0
    c=0
    while time <n:
        time=time+1
        total=total+math.log(random.randint(m,k))
        c=c+math.log(1/random.randint(m,k))
    return total,c
print(sum())   





```

```python
#练习三(助教，我打印出这个随机变量r，也是为了方便你们检查)
n=int(input('please input a number'))
def ab():
    import random 
    r=random.randint(1,9)
    b=0
    d=r
    time=0
    total=0
    while time<n:
        time=time+1
        b=10*b+d
        total=total+b
    return total,r
print('总和，随机变量',ab())
```

```python
def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table


fh = open(r'C:\Users\student\Desktop\idioms_correct.txt')
text = fh.read()
chs = get_ch_table(text.replace('\n', ''))

print(len(chs), chs)
```

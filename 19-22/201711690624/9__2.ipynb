```python
#1.随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
from random import randint
a=[randint(0,100000) for b in range(10000)]

def count_number(numbers):
    word_freq_pairs=[]
    for number in numbers:
        c=numbers.count(number)
        word_freq_pairs.append([number,c])
    return word_freq_pairs

numbers_freq=count_number(a)
print(numbers_freq)

print('----------分割-------------')
numbers_freq.sort()
print(numbers_freq)
```

```python
#对本任务中的语料.txt文件，抽取其5001-10000行存为test1.txt文件，写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。
with open(r'd:\temp\test1.txt','w') as fw,open(r'd:\temp\语料.txt') as fr:
    line_number=5001
    for line in fr:
        if 5000<line_number<=10000:
            fw.write(line)
            line_number+=1
        else:
            break

def a(b,c):
    e=set()
    f=set()
    with open(b) as f2:
        for line in f2:
            e |= {word.split('/')[0] for word in line.split()}
    with open(c) as fi:
        for line1 in fi:
            f |= {word.split('/')[0] for word in line1.split()}
    g=e&f
    h=f-e
    print('共用字',g)
    print('----------------------------------------------------------------------------------------------分隔线---------------------------------------------------------------------------------------------')
    print('独用字',h)
        
b=r'd:\temp\test1.txt'
c=r'd:\temp\语料.txt'
a(b,c)
```

```python
#1.10-10
from random import randint
a=[randint(0,100000) for b in range(10000)]

def count_number(numbers):
    word_freq_pairs=[]
    for number in numbers:
        c=numbers.count(number)
        word_freq_pairs.append([number,c])
    return word_freq_pairs

numbers_freq=count_number(a)
print(numbers_freq)

print('-----------------------')
numbers_freq.sort()
print(numbers_freq)
```

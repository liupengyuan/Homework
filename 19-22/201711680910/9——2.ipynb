```python
#10-10
import random
list1 = []
for i in range(10000):
    number=random.randrange(0,100000)
    list1.append(number)
def quick_sort(seq):
    left_seq =  []
    right_seq = []
    p=seq[0]
    for num in seq[1:]:
        if num <= p:
            left_seq.append(num)
        else:
            right_seq.append(num)
    if left_seq:
        left_seq = quick_sort(left_seq)
    if right_seq:
        right_seq = quick_sort(right_seq)

    return  left_seq + [p] + right_seq

sorted_numbers = quick_sort(list1)
print(sorted_numbers)
def count_numbers_freq_dict(numbers):
    numbers_freq_dict = {}
    
    for number in numbers:
        if number in numbers_freq_dict: 
            numbers_freq_dict[number] += 1
        else:                       
            numbers_freq_dict[number] = 1
                    
    return numbers_freq_dict
numbers_freq_dict = {}
count_numbers_freq_dict(list1)
            
        
        
```

```python
def mine(filename):
    a = []
    b = []
    with open(filename) as f:
        text = f.read()
    words = [word.split('/')[0] for word in text.split()]
    for word in words:
        for item in text:
            if word == item:
                a.append(word)
            else:
                b.append(word)
    print('',a)
    print('',b)
filename = r'e:\temp\.txt'
print(mine(filename))   
```

```python
#练习2：随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
import random
import collections
a = []
n=0
while n<10000:
    i=random.randint(0,100001)
    a.append(i)
    n += 1

def quick_sort(a):
    left_a =  []
    right_a = []
    p=a[0]
    for num in a[1:]:
        if num <= p:
            left_a.append(num)
        else:
            right_a.append(num)
    if left_a:
        left_a = quick_sort(left_a)
    if right_a:
        right_a = quick_sort(right_a)

    return  left_a + [p] + right_a

sorted_numbers = quick_sort(a)
b= collections.Counter(sorted_numbers)
for c in b:
    print (c,b[c])
```

```python

def count_words_freq_dict(filename):
    words_freq_dict = {}    
    with open(filename) as f:
        for line in f:
            words = [word.split('/')[1] for word in line.split()]
            for word in words:
                if word in words_freq_dict: 
                    words_freq_dict[word] += 1
                else:                    
                    words_freq_dict[word] = 1
                    
    return words_freq_dict
from operator import itemgetter
filename = r'd:\语料.txt'
table = count_words_freq_dict(filename)
sorted(table.items(),key = lambda x:x[1],reverse=True)[-50:]
```

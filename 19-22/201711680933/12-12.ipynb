```python
fh=open(r'c:\temp\idioms.txt')
text=fh.read()
text=text.replace('、','')
idioms=text.split('\n')
fh.close()

fha=open(r'c:\temp\idioms_correct.txt','w')
fha.writelines(idioms)
fha.close()

def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table

fh = open(r'c:\temp\idioms_correct.txt')
text = fh.read()
chs = get_ch_table(text.replace('\n', ''))
```

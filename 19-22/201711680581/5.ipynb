```python
#写函数，求n个随机整数均值的平方根，整数范围在m与k之间（n,m,k由用户输入）。
import random,math

i=0
total=0
m=int(input('请输入一个整数作为上界'))
k=int(input('请输入一个整数作为下界'))
n=int(input('请输入一个比上两数之差小的整数'))

while i<n:
    number=random.randint(m,k)
    total+=number/n
    i+=1

print(math.sqrt(total))
```

```python
#练习 2：写函数，共n个随机整数，整数范围在m与k之间，（n,m,k由用户输入）。求1：西格玛log(随机整数)，2：西格玛1/log(随机整数)
import random,math

i=0
he1=0
he2=0
m=int(input('请输入一个整数作为上界'))
k=int(input('请输入一个整数作为下界'))
n=int(input('请输入一个比上两数之差小的整数'))

while i<n:
    number1=random.randint(m,k)
    number2=random.randint(m,k)
    i+=1
x=math.log(number1)
y=math.log(number2)
z=1/y
he1+=x
he2+=z
print(he1,he2)
```

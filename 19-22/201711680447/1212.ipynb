```python
fh = open(r'C:\Users\lenov o\Desktop\idioms.txt')

text = fh.read()

fh.close()
print(len(text))
text = text.replace('、', '')
idioms = text.split('\n')
fh = open(r'C:\Users\lenov o\Desktop\idioms_correct.txt', 'w')
fh.writelines(idioms)
fh.close()
def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table

fh = open(r'C:\Users\lenov o\Desktop\idioms_correct.txt')
text = fh.read()
chs = get_ch_table(text.replace('\n', ''))
a=10
b=0
while b<a:
    b=b+1
    p=b*10
    i=[]
    import random
    for x in range(p):
        a=random.randint(1,len(chs))
        i.append(chs[a])
    print(i)
```

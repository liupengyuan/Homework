```python
import random, math


def win():
    print('我是不是很聪明')
    
def lose():
    print('啊啊啊我实在不知道了')

def game_over():
    print('游戏结束，喜欢本游戏不如赏个鲜花')

def show_team():
    print('不用知道了，程序员已祭天')

def show_instruction():
    print('您选择一个数字，规定我能猜几次，并告诉我我猜的大了还是小了，看看我能不能看透您的心^-^')

def menu():
    print('''=====游戏菜单=====
                1. 游戏说明
                2. 开始游戏
                3. 退出游戏
                4. 制作团队
             =====游戏菜单=====''')
def main():
    while True:
        menu()
        choice = int(input('请输入你的选择'))

        if choice == 1:
            show_instruction()
        elif choice == 2:
            n = int(input('请输入一个大于0的整数，作为神秘整数的上界，回车结束。'))
            m = int(input('请输入一个大于0的整数，作为神秘整数的下界，回车结束。'))
            k = int(input('请输入一个大于0的整数，这将会决定我能猜几次，回车结束。'))

            max_times = math.ceil(math.log2(k))
            guess_times = 0
            import random
            number = random.randint(n, m)
            print(number)
            while guess_times < max_times:
                guess = number
                guess_times += 1
                print('我一共可以猜', max_times, '次')
                print('我已经猜了', guess_times, '次')

                y = int(input('我猜得对吗，比您想的数大请call1，比您想的数小请call2，和您想的一样请call3。回车结束。'))
                if y==1:
                    import random
                    number = random.randint(n, guess-1)
                    print(number)
                elif y==2:
                    import random
                    number = random.randint(guess+1,m)
                    print(number)
                elif y==3:
                    win()
                    break
            else:
                lose()
            
    

        elif choice == 3:
            game_over()
            break

        else:
            show_team()
if __name__ == '__main__':    
    main()

```

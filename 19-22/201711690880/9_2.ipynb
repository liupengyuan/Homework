```python
#2

import random

A=[]
i=0
while i<10000:
    a=random.randint(1,100001)
    A.append(a)
    i+=1

A=sorted(A)
words_freq_dict={}
for num in A:
    if num in words_freq_dict: 
        words_freq_dict[num] += 1
    else:                       
        words_freq_dict[num] = 1
        
print(words_freq_dict)
```

```python
#33.txt5001-10000test1.txttest.txt
from collections import defaultdict

with open(r'd:\temp\test1.txt', 'w') as fw, open(r'd:\temp\.txt') as fr:
    line_number = 0
    for line in fr:
        if line_number <5001:
            line_number += 1
        elif line_number >=5001 and line_number <=10000:
            fw.write(line)
            line_number += 1
        else:
            break
            
def pick_words(filename):
    letter=set()
    with open(filename) as f:
        text = f.read() 
    words = [word.split('/')[0] for word in text.split()]
    for word in words:
        for ch in word:
            if ch not in letter:
                letter |= {ch}
    return letter


def compare_words(m,n):
    public = m & n
    single = m.symmetric_difference(n)
    print('',public)
    print('*'*30,'','*'*30)
    print('',single)

filename1 = r'd:\temp\test1.txt'
filename2 = r'd:\temp\test.txt'

letter1=pick_words(filename1)
letter2=pick_words(filename2)
compare_words(letter1,letter2)
```

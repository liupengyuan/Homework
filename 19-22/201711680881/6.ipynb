```python
#写函数，返回某个元素/对象在一个list中的位置，如果不在，则返回-1.
x=input('请问你要找的是？')
food=[1,5,'fish',66,'apple','pt',5,77,'pie','dumpling',]
i=0
for n in food:
    i+=1
    if n==x:
        print(x,'在第',i,'个')
        break
```

```python
#写函数，可将两个相同长度的list，间隔插入，生成新的list。例如：给两个list，a=[1,2,3,4], b=[5,6,7,8]。则可以生成：[1,5,2,6,3,7,4,8]。
a=[1,2,3,4]
b=[5,6,7,8]
list_ab=[]
i=0

while i<len(a):
        i+=1
        list_ab.append(a[i-1])
        list_ab.append(b[i-1])
print(list_ab)
```

```python
a=[int(input('x1=?  ')),int(input('y1=?  '))]
b=[int(input('x2=?  ')),int(input('y2=?  '))]
import math
aa=math.sqrt(a[0]*a[0]+a[1]*a[1])
bb=math.sqrt(b[0]*b[0]+b[1]*b[1])
ab=a[0]*b[0]+a[1]*b[1]

cos=ab/(aa+bb)
print(cos)
```

```python
#挑战性习题：python语言老师为了激励学生学python，自费买了100个完全相同的Macbook Pro，分给三个班级，每个班级至少分5个，用穷举法计算共有多少种分法？
a=range(5,85)
listlist=[]
for aa in a:
    b=range(5,95-aa)
    for bb in b:
        cc=100-aa-bb
        n=(aa,bb,cc)
        listlist.append(n)
        
print(listlist)
print('一共有',len(listlist),'种分法')
```

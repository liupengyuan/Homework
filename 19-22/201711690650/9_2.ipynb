随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。

```python
import random
from collections import Counter


i=0
X=[]
while i<10000:
    i=i+1
    a=random.choice(range(0,100000))
    X.append(a)

m=Counter(X)
dic=dict(m)
dict=sorted(dic.items(), key=lambda e:e[0], reverse=True)
print(dict)
```

挑战性任务：统计本任务中的语料.txt文件，得到每一个词对应不同词性的频次，可表示为：{词:[['n',1000], ['v', 98]...],...}，即在dict对象中，key为词，value为列表，输出词频最高的50个条目。

```python
def count_words_freq_dict(filename):
    words_freq_dict = {}
    
    with open(filename) as f:
        for line in f:
            words = [word.split('/')[1] for word in line.split()]
            for word in words:
                if word in words_freq_dict: #如果词典中有键为word，则该word的value+1
                    words_freq_dict[word] += 1
                else:                       #否则，词典中加入键word，其value为1
                    words_freq_dict[word] = 1
                    
    return words_freq_dict
from operator import itemgetter
filename = r'd:\temp\语料.txt'
table = count_words_freq_dict(filename)
sorted(table.items(),key=lambda x:x[1],reverse=True)[-50:]
```

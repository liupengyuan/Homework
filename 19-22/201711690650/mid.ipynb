```python
def compute(m):
    i=0
    x=0
    total=0
    while i<m:
        i=i+1
        x=x+i
        total=total+x
    return total

m=int(input('请输入一个整数'))
print(compute(m))

        
```

练习4

```python
n=[]
i=int(input('请输入列表中的元素，若已结束输入000'))
while i!=000:
    n.append(i)
    i=int(input('请输入列表中的元素，若已结束输入000'))
print(n)
total=0
for number in n:
    total=total+number
    average=total/len(n)
print(average)
for number in n:
    if number>average:
        print(number)
```

```python
#写函数，求n个随机整数均值的平方根，整数范围在m与k之间（n,m,k由用户输入）。
import random
def main(n,m,k):
    i = 0
    total = 0
    while i < n:
        number = random.randint(m,k)
        i += 1
        total += number
        result = (total/n)**(1/2)
    print('n个随机整数均值的平方根=',result)
    
main(5,1,10)
```

```python
#写函数，共n个随机整数，整数范围在m与k之间，（n,m,k由用户输入）。求1：西格玛log(随机整数)，2：西格玛1/log(随机整数)
import random,math

def main(n,m,k):
    i = 0
    total_1 = 0
    total_2 = 0
    
    while i < n:
        number = random.randint(m,k) 
        i += 1
        number = (math.log2(number))
        total_1 += number
        total_2 += (1/number)
    print('sigma log=',total_1)
    print('sigma 1/log=',total_2)
    
main(5,1,10)




```

```python
#写函数，求s=a+aa+aaa+aaaa+aa...a的值，其中a是[1,9]之间的随机整数。例如2+22+222+2222+22222(此时共有5个数相加)，几个数相加由键盘输入。

import random

def main():
    i=0
    s=0
    c=0
    end=int(input('请输入一个正整数，以回车结束'))
    while i<end:
        a=random.randint(1,9)
        i+=1
        while c<i:
            s+=a*((10)**c)
            c+=1
    print(s)
    
    
main()



```

```python
#仿照task5，将猜数游戏改成由用户随便选择一个整数，让计算机来猜测的猜数游戏，要求和task5中人猜测的方法类似，但是人机角色对换，由人来判断猜测是大、小还是相等，请写出完整的猜数游戏。
import random, math
def welcome():
    print(
        '''
           ======欢迎来到猜数游戏=======
        
                     /  \     ,    ,
           _._     _ |oo| _  / \__/ \
    
          _||||   ((/ () \))   /   \
    
          |||||/|  ( ==== )    |oo|    
           \____/  _`\  /'_    /   \
    
           /   /.-' /\<>/\ `\.( () )_._      
           |    `  /  \/  \  /`'--'////)
            \__,-'`|  |.  |\/ |/\/\|"\"` 
                   |  |.  | \___/\___/  
                   |  |.  |   |    |
          
          ======欢迎来到猜数游戏=======
        '''
    )

def win():
    print(
        '''
           ======恭喜你，你赢了=======
        
    
                ."".    ."",
                |  |   /  /
                |  |  /  /
                |  | /  /
                |  |/  ;-._ 
                }  ` _/  / ;
                |  /` ) /  /
                | /  /_/\_/\
                |/  /      |
                (  ' \ '-  |
                 \    `.  /
                  |      |
                  |      |
          
          ======恭喜你，你赢了=======
        '''
    )
    
def lose():
    print(
        '''
           ======YOU LOSE=======
        
    
                

                   .-"      "-.
                  /            \
                 |              |
                 |,  .-.  .-.  ,|
                 | )(__/  \__)( |
                 |/     /\     \|
       (@_       (_     ^^     _)
  _     ) \_______\__|IIIIII|__/__________________________
 (_)@8@8{}<________|-\IIIIII/-|___________________________>
        )_/        \          /
       (@           `--------`
       
       
       
          ======YOU LOSE=======
        '''
    )
    
def game_over():
    print(
        '''
           ======GAME OVER=======
        
             _________ 
            / ======= \ 
           / __________\ 
          | ___________ | 
          | | -       | | 
          | |         | | 
          | |_________| |________________ 
          \=____________/                ) 
          / """"""""""" \               / 
         / ::::::::::::: \          =D-' 
        (_________________) 

       
          ======GAME OVER=======
        '''
    )
    
def show_team():
    print('Alice,Bob,Wendy,Tom')
def show_instruction():
    print('first step',
         'second step')

    
def menu():
    print('''=====游戏菜单=====
                1. 游戏说明
                2. 开始游戏
                3. 退出游戏
                4. 制作团队
             =====游戏菜单=====''')
welcome()       
def guess_game():
    number = int(input('请输入一个大于0的整数，以回车结束。'))
    max_times = math.ceil(math.log2(number))
    guess_times = 0
    
    while guess_times < max_times:
        guess = random.randint(1,number**2)
        guess_times += 1
        print('一共可以猜', max_times, '次')
        print('你已经猜了', guess_times, '次')
        
        if guess == number:
            win()
            print('数字是：', guess)
            print('你比标准次数少', max_times-guess_times, '次')
            break
        elif guess > number:
            print('抱歉，你猜大了')
        else:
            print('抱歉，你猜小了')
            
    else:
        print('数字是：', number)
        lose()

# 主函数
def main():
    while True:
        menu()
        choice = int(input('请输入你的选择'))
        if choice == 1:
            show_instruction()
        elif choice == 2:
            guess_game()
        elif choice == 3:
            game_over()
            break
        else:
            show_team()


#主程序
if __name__ == '__main__':
    main()




```

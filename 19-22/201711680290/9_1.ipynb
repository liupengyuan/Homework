```python
import random,time,bisect
A=[]
B=[]
V=[]
C=[]
D=[]
for i in range(1,100001):
    A.append(i)
B=A
random.shuffle(B)

for j in range(1000):
    v = random.randint(1,110001)
    V.append(v)
start_time1 = time.time()
for x in B:
    if x in V:
        C.append(x)
end_time1 = time.time()
print('顺序法',C)
start_time2 = time.time()
for k in V:
    if k<=100000:
        d=bisect.bisect(A,k+1)
        D.append(d)
end_time2= time.time()
print('二分法',D)        
print('顺序法时间',end_time1 - start_time1)
print('二分法时间',end_time2 - start_time2)        
```

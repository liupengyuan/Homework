### 写函数，返回某个元素/对象在一个list中的位置，如果不在，则返回-1.

```python

numbers=[]
n=int(input('请输入一个整数，表示将要输入元素的个数'))
i=0
while i<n:
    number=input('请输入一个元素')
    numbers.append(number)
    i+=1
m=input('请输入一个元素')
t=0
while t<len(numbers):
    a=numbers[t]
    if a==m:
        print(t)
        break
    t+=1
    
```

### 写函数，可将两个相同长度的list，间隔插入，生成新的list。例如：给两个list，a=[1,2,3,4], b=[5,6,7,8]。则可以生成：[1,5,2,6,3,7,4,8]。


```python
numbers1=[]
numbers2=[]
numbers=[]
n = int(input('请输入一个整数，表示将要输入list1和list2的元素数，回车结束。'))
i = 0
while i < n:
    number1 = input('请向list1中输入一个元素，回车结束。')
    numbers1.append(number1)
    number2 = input('请向list2中输入一个元素，回车结束。')
    numbers2.append(number2)
    i += 1
for i in numbers1:
    if i not in numbers:
        numbers.append(i)
    for j in numbers2:
        if j not in numbers:
            numbers.append(j)
            break
print(numbers)
```

### 挑战性习题：python语言老师为了激励学生学python，自费买了100个完全相同的Macbook Pro，分给三个班级，每个班级至少分5个，用穷举法计算共有多少种分法？

```python
n=[]
i=0
a=0
while a<=85:
    for b in range(a,86-a):
        c=85-a-b
        n.append(a)
        n.append(b)
        n.append(c)
    print(n)
    a=a+1
    i=i+1
print(i)
```

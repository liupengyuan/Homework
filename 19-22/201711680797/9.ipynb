```python
#生成列表A：[1，2，3，4，......,100000]
A=list(range(1,100001))
#生成列表B：[A的乱序]
import random
B=list(range(1,100001))
random.shuffle(B)
#生成乱序的包含1~110000的整数的列表
C=list(range(1,110001))
random.shuffle(C)
#定义一个顺序查找算法的函数(此函数的缺点是不能显示出不存在于列表B中的元素)
c=0
def shun_search(c,B):
    for i in range(1000):
        c=C[i]
        for k in range(len(B)):     
            if c==B[k]:          
                print(c,'存在于','B中') 
#用顺序查找算法在B中找上述整数并计算时间(不知道此处是求程序的执行时间还是CPU的执行时间)            
import time
start=time.time()    
print(shun_search(c,B))
end=time.time()
print('总时间为',end-start)
#定义一个二分查找算法的函数
def er_search(c, A, low, high):
    low=1
    high=100000
    for i in range(1000):
        c=C[i]
        if low > high:
            return -1
        else:
            mid = (low+high)//2
            if c == A[mid]:
                return mid
            elif c > A[mid]:
                return bi_search(c, A, mid+1, high)
            else:
                return bi_search(c, A, low, mid-1)
#用二分查找算法在A中找上述整数并计算时间
import time
start=time.time()    
print(er_search(c,A))
end=time.time()
print('总时间为',end-start)
```

```python
def my_sum(n):
    i=0
    total_n=0
    while i<n:
        i=i+1
        total_n=total_n+i
    return total_n

n=int(input('请输入一个正整数'))
print(my_sum(n))
```

```python
def my_mul(n):
    i=0
    total_n=1
    while i<n:
        i=i+1
        total_n=total_n*i
    return total_n

n=int(input('请输入一个正整数'))
print(my_mul(n))
```

```python
def com_jia(end):
    i=0
    total_n=0
    while i<end:
        i=i+1
        total_n=total_n+i
    return total_n

def com_cheng(end):
    i=0
    total_n=1
    while i<end:
        i=i+1
        total_n=total_n*i
    return total_n

m=int(input('请输入一个正整数'))
n=int(input('请输入一个正整数'))
if m>n:
    print(com_jia(m)+com_jia(n))
else:
    print(com_cheng(m)+com_cheng(n))
```

练习2：随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数

```python
import random
nums=[]
for i in range(10000):
    num=random.randint(0,100001)
    nums.append(num)
    
nums_freq_dict={}
for num in nums:
    if num in nums_freq_dict:
        nums_freq_dict[num]+=1
    else:
        nums_freq_dict[num]=1
sorted_dict=sorted(nums_freq_dict.items())
print(sorted_dict)
```

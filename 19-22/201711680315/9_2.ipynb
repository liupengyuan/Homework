```python
#1.随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出
#整数及出现次数。
```

```python
import random
from operator import itemgetter
a = [random.randint(0,100000) for i in range(10000)]
b=set(a)
c=[]
for j in b:
    d=j,a.count(j)
    c.append(d)
x = sorted(c, key=itemgetter(0))
for i in range(len(x)):
    print(x[i])
```

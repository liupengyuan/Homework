```python
#练习 1：写函数，求n个随机整数均值的平方根，整数范围在m与k之间（n,m,k由用户输入）。
```

```python
import random
def main_():
    m=int(input('请输入一个整数'))
    k=int(input('请输入一个整数'))
    n=int(input('请输入一个整数'))
    i=0
    total=0
    if m>k:
        m,k=k,m
    while i<n:
        number=random.randint(m,k)
        i=i+1
        total=total+number
        result=(total/n)**(1/2)
        print(number)
    return result
print(main_())
```

```python
#练习 2：写函数，共n个随机整数，整数范围在m与k之间，（n,m,k由用户输入）。求1：西格玛log
#(随机整数)，2：西格玛1/log(随机整数)
```

```python
import random,math
def main():
    m=int(input('请输入一个整数'))
    k=int(input('请输入一个整数'))
    n=int(input('请输入一个整数'))
    i=0
    result1=0
    result2=0
    if m>k:
        m,k=k,m
    while i<n:
        number=random.randint(m,k)
        print(number)
        i=i+1
        result1+=math.log(number)
        result2+=(math.log(number))**(-1)
    return result1,result2
print('西格玛log(随机整数),西格玛1/log(随机整数)分别为:',main())
    
    
```

```python
#2 随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
import random
from operator import itemgetter
numlist = []
numdict = {}
for i in range(10000):
    num = random.randint(0,100001)
    numlist.append(num)

for number in numlist:
    if number in numdict:
        numdict[number] += 1
    else:
        numdict[number] = 1

sorted(numdict.items(),key = lambda item:item[0],reverse = True)   
```

```python
#3 对本任务中的语料.txt文件，随机抽取其5001-10000行存为test1.txt文件，写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。
with open(r'd:\temp\test1.txt', 'w') as fw, open(r'd:\temp\语料.txt') as fr:
    line_number = 1
    for line in fr:
        if  line_number <= 5000:
            line_number += 1
        elif 5001 <= line_number <=10000:
            fw.write(line)
            line_number += 1
        else:
            break
def cut(filename):
    table = set()
    with open(filename) as f:
        text = f.read()
        words = [line.split('/')[0]for line in text.split()]
    for word in words:
        for ch in word:
            if ch not in table:
                table |= {ch}
    return table

def compare(table,table1):
    homo = table&table1
    heter = table.difference(table1)
    heter1 = table1.difference(table)
    print('共用字为',homo)
    print('test独用字为',heter)
    print('test1独用字为',heter1)
    
table = cut(r'd:\temp\test.txt')
table1 = cut(r'd:\temp\test1.txt')
compare(table,table1)           
```

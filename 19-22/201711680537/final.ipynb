```python
#1
def pic(sym,line):
    for i in range(1,line+1):
        print(' '*((line-i)*2)+(sym+' ')*i + sym*line + (' '+sym)*i)
        
sym = input('请输入符号')
line = int(input('请输入行数'))
pic(sym,line)
```

```python
#2
def A(m,n):
    if m == 0:
        return n+1
    elif m > 0 and n == 0:
        return A(m-1,1)
    elif m > 0 and n > 0:
        return A(m-1,A(m,n-1))
    
print(A(3,4))
```

```python
#3(a)
import random

key =[]
i = 0
n = int(input('请输入题目个数'))
a = int(input('请输入下界'))
b = int(input('请输入上界'))
while i < n:
    int1 = random.randint(a,b)
    int2 = random.randint(a,b)
    choice = [int1+int2,int1-int2,int1*int2,int1/int2]
    question = random.choice(choice)

    if question == int1+int2:
        answer1 = question
        key.append(answer1)
        print(int1,'+',int2,'=?')
        i += 1
    elif question == int1-int2 and type(question) == int and question > 0:
        answer2 = question
        key.append(answer2)
        print(int1,'-',int2,'=?')
        i += 1
    elif question == int1-int2 and type(question) == int and question < 0:
        answer3 = -1*question
        key.append(answer3)
        print(int2,'-',int1,'=?')
        i += 1
    elif question == int1*int2:
        answer4 = question
        key.append(answer4)
        print(int1,'*',int2,'=?')
        i += 1
    elif question == int1/int2 and type(question) == int:
        answer5 = question
        key.append(answer5)
        print(int1,'/',int2,'=?')
        i += 1
 
```

```python
#3(c)
i = 0
score = 0
n = int(input('请输入题目个数'))
while i < n:
    quesstr = input('请输入题目(规则我就不写了hhh)')

    if quesstr[1] == '+':
        queslist = quesstr.split('+')
    
        int3 = int(queslist[0])
        int4 = int(queslist[1][0])
        print('答案是',int3+int4)
        score += 5
        i += 1
    
    elif quesstr[1] == '-':
        queslist = quesstr.split('-')
        int3 = int(queslist[0])
        int4 = int(queslist[1][0])
        if type(int3-int4) == int and int3 - int4 > 0:
            print('答案是',int3-int4)
            score += 5
            i += 1
        else:
            print('题目无效，请重新输入')
            
    elif quesstr[1] == '*':
        queslist = quesstr.split('*')
        int3 = int(queslist[0])
        int4 = int(queslist[1][0])
        print('答案是',int3*int4)
        score += 10
        i += 1
    elif quesstr[1] == '/':
        queslist = quesstr.split('/')
        int3 = int(queslist[0])
        int4 = int(queslist[1][0])
        if type(int3/int4) == int:
            print('答案是',int3/int4)
            score += 10
            i += 1
        else:
            print('题目无效，请重新输入')


        
print('分数',score)
```

```python
#4(c)
from collections import Counter

def count(filename):
    words_freq_dict = Counter()
    
    with open(filename) as f:
        for line in f:
            words_freq_dict.update([word.split('/')[1] for word in line.split()])
    return words_freq_dict


filename = r'd:\temp\test_re_gbk.txt'
table = count(filename)
print(table)
```

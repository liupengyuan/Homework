```python
#成语接龙（呜呜呜我觉得我写得好乱qaq麻烦助教了 谢谢你
import random
    
def team():
    print('hhhhh')
    
def direction():
    print('假装打了说明嘻嘻嘻')    
    
def menu():
    print('''
         ----------------------
                 1.direction
                 2.game
                 3.quit
                 4.team
          ----------------------
          ''')
    
def game():
    fh = open(r'c:\temp\成语大全.txt')
    text = fh.read()
    idioms = text.split()
    fh.close()
    
    cscore = 0
    hscore = 0
    i = 1

    cans = random.choice(idioms)
    print(cans)

    while True:
    
        if i == 1:            #用户给出成语
            ans = str(input('plz type an idiom: (if u want to quit,type q)    '))
            
            if len(ans) == 4 and ans[0] == cans[3]:       #用户给对成语
                hscore += 10
                print('u r right!') 
                i = 0
            elif len(ans) != 4 or  ans[0] != cans[3]:     #用户给错成语
                print('88,ur score is',hscore,'my score is ',cscore )    
                break
            elif ans == 'q':                              #用户退出
                print('88,ur score is',hscore,'my score is ',cscore )    
                break

        elif i == 0:          #计算机给出成语
            comans = []
            for idiom in idioms:
                if len(idiom) == 4 and idiom[0] == ans[3]:
                    comans.append(idiom)
            
            if len(comans) == 0:         #成语大全里没有对应成语
                print('but I can\'t find the idiom ;(   ')
                choi = input('if u want to quit,type q; if u want to continue,type c:  ')
                
                if choi == 'q':          #用户退出
                    print('88,ur score is',hscore,'my score is ',cscore )    
                    break
                elif choi == 'c':        #计算机给出另一个成语 游戏继续
                    cans = random.choice(idioms)
                    print(cans)
            
            elif len(comans) > 0:        #成语大全里有对应成语
                cans = random.choice(comans)
                cscore += 10
                print(cans)
            i = 1

        
            
menu()
choice = int(input('plz type ur choice:'))
if choice == 1:
    direction()
elif choice == 2:
    game()
elif choice == 3:
    print('88')
elif choice ==4:
    team()               
```

```python
import time,bisect,random
A = [i for i in range(1,100001)]

B = [i for i in range(1,100001)]
random.shuffle(B)

randnum = []
for i in range(1,1001):
    num = random.randint(1,110001)
    randnum.append(num)

start = time.time()
amount = 0
for item in randnum:
    if item > 100000:
        print(item,'不在B中')
        amount += 1
    for number in B:
        if item == number:
            print('%d是B中的第%d位'%(item,B.index(number)))
            amount += 1
    if amount == 1000:
        end = time.time()
        print('用时: ',end-start,'秒')
        break
       

newstart = time.time()
newamount = 0
for item in randnum:
    if item > 100000:
        print(item,'不在A中')
        newamount += 1
    if item <= 100000:
        print('%d是A中的第%d位'%(item,bisect.bisect_left(A,item)+1))
        newamount += 1
    if newamount == 1000:
        newend = time.time()
        print('用时:',newend-newstart,'秒')
        break
#麻烦助教啦 谢谢你(❁´ω`❁)
```

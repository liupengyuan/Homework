```python
import random, math
def sum_1(n,m,k):
    i = 1
    total = 0
    while i <= n:
        number = random.randint(m, k)
        x = math.log( number)
        total = total + x
        i = i + 1
    return total
def sum_2(n,m,k):
    i = 1
    total = 0
    while i <= n:
        number = random.randint(m, k)
        y = 1/math.log(number)
        total = total + y
        i = i + 1
    return total

n = int(input('请输入大于零一个整数'))
m = int(input('请输入一个整数'))
k = int(input('请输入一个不等于上一个数的整数'))
if m > k:
    m,k = k,m
else:   
    print (sum_1(n,m,k), sum_2(n,m,k))


    
```

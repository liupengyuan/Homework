```python
#练习2
import random
numbers=[random.randint(0,100001) for i in range(10000)]
print(numbers)
def count_numbers_freq(numbers):
    number_freq_pairs=[]
    with numbers as f:
        for line in f:
            numbers = [number.split('/')[0] for number in line.split()]
            for number in numbers:
                for item in number_freq_pairs:
                    if number==item[0]:
                        item[1]+=1
                        break
                else:
                    number_freq_pairs.append([number,1])
                
    return number_freq_pairs
numbers=numbers
table=count_numbers_freq(numbers)
print(table[:10000])
                
```

```python
#挑战性任务
def count_words_freq_dict(filename):
    words_freq_dict = {}
    
    with open(filename) as f:
        for line in f:
            words = [word.split('/')[0] for word in line.split()]
            for word in words:
                if word in words_freq_dict: 
                    words_freq_dict[word] += 1
                else:                       
                    words_freq_dict[word] = 1
                    
    return words_freq_dict

#测试用

filename = r'd:\temp\语料.txt'
table = count_words_freq_dict(filename)

for i, n in enumerate(table):
    print(n, table[n])
for i, v in enumerate(table):
    print(v, table[v])
for i, m in enumerate(table):
    print(m, table[m])
    
        
```

```python
#1
def f(symbol,n):
    a=symbol
    b=symbol*n
    for i in range(n):
        print((2*n-i)*' '+a*(i+1)+' '+b+' '+a*(i+1))
        
symbol=input('输入塔的符号')
n=int(input('输入层数'))
f(symbol,n)
```

```python
#2
def Ackermann(m,n):
    if m==0:
        A=n+1
    elif m>0 and n==0:
        A=Ackermann(m-1,1)
    elif m>0 and n>0:
        A=Ackermann(m-1,Ackermann(m,n-1))
    return A
m=int(input('m的值'))
n=int(input('n的值'))
Ackermann(m,n)
        
        
```

```python
#3.(a) 3.(b)
import random
def questions(a,b,n):
    e=[]
    i=1
    points=0
    while i>0:
        m=random.randint(a,b)
        k=random.randint(a,b)
        
        c=['+','-','*','/']
        d=random.choice(c)
        if d=='-' :
            if m-k>0:
                e.append(m)
                print(m,d,k,'=',sep=' ')
                
        elif d=='/':
            if (m/k)//1==m/k:
                e.append(m)
                print(m,d,k,'=',sep=' ')
        else:
            e.append(m)
            print(m,d,k,'=',sep=' ')
            
        if len(e)>n:
            break
a=int(input('a的值'))
b=int(input('b的值'))
n=int(input('题目数目'))
questions(a,b,n)
```

```python
#3(c)
n=int(input('题目数目'))
point=0
for i in range(n):
    a=input('题目格式为m+n=  ')
    
    if a[1]=='+':
        print(int(a[0])+int(a[2]))
        point+=5
    elif a[1]=='-':
        print(int(a[0])-int(a[2]))
        point+=5
    elif a[1]=='*':
        print(int(a[0])*int(a[2]))
        point+=10
    elif a[1]=='/':
        print(int(a[0])/int(a[2]))
        point+=10
print('总分为',point)
    
```

```python
def get_words(filename):
    words=set()
    texts=set()
    a=''
    b=''
    c=set()
    d=set()
    with open(filename) as f:
        for line in f:
            c |=set(word.split('/')[0] for word in line.split())
        for i,m in enumerate(c):
            a=a+m
        for j in a:
            words.add(j)
    with open(r'D:\test.txt') as fh:
        for line in fh:
            d |=set(text.split('/')[0] for text in line.split())
        for i,n in enumerate(d):
            b=b+n
        for k in b:
            texts.add(k)
    print('',words&texts)
    print('',words-texts)

#

get_words(r'D:\test1.txt')
```

```python
from collections import Counter

a=[]
b=Counter()

with open(r'D:\.txt') as f:
    for line in f:
        for word in line.split():
            a.append(word)
    for i in a:
        if i in b:
            b[i]=b[i]+1
        else:
            b[i]=1
c=b.most_common(50)
print(c)






```

```python
#练习2
import random

res = []
sin = []
tur = []
ans = []

for i in range(0,100000):
    x = random.choice(range(0,100000))
    res.append(x)
    i += 1
    if i == 10000:
        break
        
for m in res:
    if m not in sin:
        t = res.count(m)
        tur.append(t)
        sin.append(m)
        sin.sort()
        
print('抽出的数字从小到大依次是:',sin)
print('它们被抽中的次数是',tur)

for a in zip(sin,tur):ans.extend(a)
print (ans)

```

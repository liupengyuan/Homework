```python
#写函数，返回某个元素/对象在一个list中的位置，如果不在，则返回-1.
def number():
    one= [1, 2, 3, 4, 5]
    n = int(input('请输入一个整数，回车结束'))
    if n in one:
        print(one.index(n))
    else:
        print(-1)

#主程序
number()
```

```python
#写函数，可将两个相同长度的list，间隔插入，生成新的list。例如：给两个list，a=[1,2,3,4], b=[5,6,7,8]。则可以生成：[1,5,2,6,3,7,4,8]。
def new():
    n=int(input('请输入数字，确定列表中元素的数目'))
    a=[]
    for i in range(n):
        number = int(input('请输入一个a中数字，回车结束。'))
        a.append(number)
    
    b=[]
    for i in range(n):
        number = int(input('请输入一个b中数字，回车结束。'))
        b.append(number)
       
    numbers=[]
    i=0
    while i<n:
        numbers.append(a[i])
        numbers.append(b[i])
        i+=1
    print(numbers)

#主程序
new()
        
```

```python
#写函数，可求两个向量的夹角余弦值，向量可放在list中。主程序调用该函数。
def cos():  
    a=[]
    for i in range(2):
        number = int(input('请输入a的坐标'))
        a.append(number)
    b=[]
    for i in range(2):
        number = int(input('请输入b的坐标'))
        b.append(number)
    product=a[0]*b[0]+a[1]*b[1]
    distant_a=(a[0]**2+a[1]**2)**(1/2)
    distant_b=(b[0]**2+b[1]**2)**(1/2)
    end=product/(distant_a*distant_b)
    print('两向量夹角余弦值为:',end)
#主程序
cos()
```

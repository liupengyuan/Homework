```python
#随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
import random

numbers=[]
for i in range(10000):
    numbers.append( random.randint(0, 100000))
    
numbers_orderly = sorted(numbers)

number_freq_pairs = []
for number in numbers_orderly:
    for item in number_freq_pairs:
        if number == item[0]:
            item[1] += 1
            break
    else:
        number_freq_pairs.append([number, 1])
print(number_freq_pairs)
```

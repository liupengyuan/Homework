```python
练习 2：写函数，共n个随机整数，整数范围在m与k之间，（n,m,k由用户输入）。
求1：西格玛log(随机整数)，2：西格玛1/log(随机整数)
```

```python
import random,math

def word():
    n = int(input('请输入整数个数:'))
    m = int(input('请输入范围内的下界:'))
    k = int(input('请输入范围内的上界:'))
   
    
    i = 0
    total_1 = 0
    total_2 = 0
    
    while i < n:
        number = random.randint(m,k) 
        i += 1
        total_1 = total_1 +math.log(number)
        total_2 = total_2 +math.log(number)**(-1)
        
        print(number)
        
    return total_1,total_2


print(word())
    
```

```python
创建一个空的列表用来防止生成的所有数字 
创建一个词对应词和对应词频的列表

将数字放入空列表1
遍历出所有数字
返回遍历出的每个数字对应的位置,并将该词词频插入到第二个列表中
判断在该位置是否有这个数字，如果有，在对应的词频上加1
如果没有，就将这个词加入到第二个列表中，并频数为一
import random,bisect

list1 = []
word_freq_pairs = []
for i in range(10001):
    list1.append(random.randint(1,100000))
for word in list1:
    i = bisect.bisect_left(word_freq_pairs, [word])
    if i != len(word_freq_pairs) and word_freq_pairs[i][0] == word:
        word_freq_pairs[i][1] += 1
    else:
        word_freq_pairs.insert(i, [word,1])
print(word_freq_pairs)
```

```python
#第一题
def xing(shape,n):

    for i in range(1,n):
        print(' '*(n-i+1) +(shape)*i+' '+shape*n +(' '+ shape)*i)
shape = input('请输入形状')
n = int(input('行数'))
xing(shape,n)
    
```

```python
#第二题
def a(m,n):
    if m ==0:
        print(n+1)
        
    elif m>0 and n ==0:
        m = m-1
        n = 1
        a(m,n)
        
    elif m>0 and n>0:
        m = m-1
        n = a(m,n-1)
        
a(3,4)
```

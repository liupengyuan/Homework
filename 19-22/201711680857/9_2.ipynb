###  练习二 随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。

```python
import random
list1=[]
list2=[]
for i in range(0,10000):
    a=random.randint(0,100001)
    list1.append(a)
list1.sort()
for j in list1:
    num=list1.count(j)
    list2.append([j,num])
print(list2)
```

### 成语接龙游戏
### 初始计算机得分为0，人类得分为0
### 计算机随机出一个成语，人类进行接龙，然后交替。程序自动判断接龙是否正确。每接对一个成语加十分，接龙接错则游戏结束。最后给出人类得分和计算机得分。

```python
import random


def win():
    print('哈哈，成功了')
def lose():
    print('哈哈，失败了')
def game_over():
    print('游戏结束，你已经成为了一个新晋大佬，恭喜你！')
def team():
    print('制作团队：7k7k')
def instruction():
    print('游戏说明：成语接龙')
def menu():
    print('''
              =====游戏菜单=====
         **这是一个大佬才会玩的游戏**
                1. 游戏说明
                2. 开始游戏
                3. 退出游戏
                4. 制作团队
              =====游戏菜单=====''')   
    
def game():
    score_com=0
    score_man=0
    fh=open(r'c:\temp\成语大全.txt')
    text=fh.read()
    idioms = text.split('\n')
    idiom1=random.choice(idioms)
    print('计算机：',idiom1)
    while score_man>=0:
        idiom2=input('人类：')
        if idiom1[-1]!=idiom2[0]:
            game_over()
            break
        elif idiom1[-1]==idiom2[0]:
            score_man+=10 
            print('现在你的分数是：',score_man)
            print('现在计算机的分数是：',score_com)
            idioms_new=[]
            for idiom in idioms:
                
                if idiom[0]==idiom2[-1]:
                    idioms_new.append(idiom)
            if len(idioms_new)==0:
                game_over()
                break
            elif len(idioms_new)!=0:
                
                idiom1=random.choice(idioms_new)  
                score_com+=10
                print('计算机：',idiom1)
                    
                print('现在你的分数是：',score_man)
                print('现在计算机的分数是：',score_com)
            
                    
                
    print('你的分数是：',score_man)
    print('计算机的分数是：',score_com)
def main():
    while True :
        menu()
        choice = int(input('请输入你的选择'))
        if choice == 1:
            instruction()
        
        elif choice == 2:
            game()
        
        elif choice == 3:
            game_over()
            break
        else :
            team()
        
main()
```

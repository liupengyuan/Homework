```python

def A(m,n):
    if m==0:
        return n+1
    elif m>0:
        if n==0:
            return A(m-1,1)
        elif n>0:
            return A(m-1,A(m,n-1))
m=int(input('m:'))
n=int(input('n:'))
print(A(m,n))
```

```python
n=int(input('题数'))
score=0
for i in range(n):
    a=input('题目')
    c1=a[0]
    c3=a[1]
    c2=a[2]
    if c3== '+':
        c=int(c1)+int(c2)
        print(c)
        score+=5
    if c3== '-':
        c=int(c1)-int(c2)
        print(c)
        score+=5
    if c3== '*':
        c=int(c1)*int(c2)
        print(c)
        score+=10
    if c3== '/':
        c=int(c1)/int(c2)
        print(c)
        score+=10
print(score)

```

```python
with open(r'd:\temp\成语大全(含成语解释)_utf8.txt',encoding='utf-8') as f:
    a=f.read()
    A=[]
    a1 = [word.split(' ')[0] for word in a.split()]
    for word in a1:
        if word not in A:
            A.append(word)

    num_list=[]
    l=len(A)
    for i in range(l):
        num=a1.count(A[i])
        num_list.append(num)
        nums=sorted(num_list)
        for j in range(20):
            a=num_list.index(nums[j])
            print(A[a],nums[j])
```

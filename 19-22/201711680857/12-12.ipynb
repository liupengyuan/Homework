```python
def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table

import random
fh = open(r'c:\temp\idioms_correct.txt')
text = fh.read()
chs = get_ch_table(text.replace('\n', ''))



a1=''   
for i in range(10): 
    
    ch=random.choice(chs)
    a1=a1+ch
print(a1)
list=[a1]
fh=open(r'c:\temp\1.txt','w')
fh.writelines(list)

a2=''   
for i in range(20): 
    
    ch=random.choice(chs)
    a2=a2+ch
print(a2)
list=[a2]
fh=open(r'c:\temp\2.txt','w')
fh.writelines(list)

a3=''   
for i in range(30): 
    
    ch=random.choice(chs)
    a3=a3+ch
print(a3)
list=[a3]
fh=open(r'c:\temp\3.txt','w')
fh.writelines(list)

a4=''   
for i in range(40): 
    
    ch=random.choice(chs)
    a4=a4+ch
print(a4)
list=[a4]
fh=open(r'c:\temp\4.txt','w')
fh.writelines(list)

a5=''   
for i in range(50): 
    
    ch=random.choice(chs)
    a5=a5+ch
print(a5)
list=[a5]
fh=open(r'c:\temp\5.txt','w')
fh.writelines(list)

a6=''   
for i in range(60): 
    
    ch=random.choice(chs)
    a6=a6+ch
print(a6)
list=[a6]
fh=open(r'c:\temp\6.txt','w')
fh.writelines(list)

a7=''   
for i in range(70): 
    
    ch=random.choice(chs)
    a7=a7+ch
print(a7)
list=[a7]
fh=open(r'c:\temp\7.txt','w')
fh.writelines(list)

a8=''   
for i in range(80): 
    
    ch=random.choice(chs)
    a8=a8+ch
print(a8)
list=[a8]
fh=open(r'c:\temp\8.txt','w')
fh.writelines(list)

a9=''   
for i in range(90): 
    
    ch=random.choice(chs)
    a9=a9+ch
print(a9)
list=[a9]
fh=open(r'c:\temp\9.txt','w')
fh.writelines(list)

a10=''   
for i in range(100): 
    
    ch=random.choice(chs)
    a10=a10+ch
print(a10)
list=[a10]
fh=open(r'c:\temp\10.txt','w')
fh.writelines(list)


fh.close()
```

```python
#随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。

import random,bisect
numbers=[]
for i in range(10001):
    a=random.randint(0,100000)
    v=bisect.bisect_left(numbers,[a])
    if v!=len(numbers) and a==numbers[v][0]:
        numbers[v][1]+=1
    else:
        numbers.insert(v,[a,1])
print(numbers)


    

        
    
    
```

```python
#挑战性任务：统计本任务中的语料.txt文件，得到每一个词对应不同词性的频次，
#可表示为：{词:[['n',1000], ['v', 98]...],...}，即在dict对象中，key为词，value为列表，输出词频最高的50个条目。
def count_words_freq_dict(filename):
    words_freq_dict = {}
    
    with open(filename,encoding='gbk') as f:
        for line in f:
            words = [word.split('/')[0] for word in line.split()]
            for word in words:
                if word in words_freq_dict: 
                    words_freq_dict[word] += 1
                else:                       
                    words_freq_dict[word] = 1
    final=sorted(words_freq_dict.items(),key=lambda item:item[1])

                    
    return final
filename=r'/Users/lwd/Desktop/py/语料.txt'
table=count_words_freq_dict(filename)
for i, item in enumerate(table):
    print(item, table[item])
    if i==50:
        break
```

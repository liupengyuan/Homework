```python

def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table

import random
def idiom_robot(i):
    
    idioms=[]
    fh = open(r'C:\temp\成语大全.txt')
    text = fh.read()
    chs = get_ch_table(text.replace('\n', ''))
    guess_ch_table = [ch for ch in idioms]
    
    while len(guess_ch_table) < i*10:
        ch = random.choice(chs)
        if ch not in guess_ch_table:
            guess_ch_table.append(ch)
        
        fh = open(r'C:\temp\n.txt', 'w')
        fh.writelines(guess_ch_table)
        fh.close()
    print(guess_ch_table)
    
for j in range(11):
    idiom_robot(j)
```

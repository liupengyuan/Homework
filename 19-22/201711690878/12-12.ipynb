```python
import random
fh=open(r'c:\temp\idioms.txt')
text=fh.read()
text=text.replace('\n','')
text=text.replace('`','')
line=[]
for m in range(1,11):
    for i in range(1,10*m+1):
        x=random.choice(text)
        line.append(x)
    n=str(m)
    fh=open('c:/temp/'+n+'.text','w')
    fh.write('\n'.join(line))
    fh.close()
```

```python
with open(r'd:\temp\语料.txt') as f:
    line=f.read()
words=[word.split('/') for word in line.split()]
words2=[word2 for word2 in line.split()]
colkey=[]
for word in words:
    colkey.append(word[0])
from collections import Counter
listkeys=sorted(dict(Counter(colkey)).items(),key=lambda item:item[1])
keys_50=listkeys[:-51:-1]
final=[]
for exkey in keys_50:
    key=list(exkey)
    final.append(key[0])
    for word2 in words2:
        if key[0] in word2[:len(key[0])]:
            temp1=word2[len(word2)-len(key[0]):len(word2)]
            n=final.index(key[0])
            if temp1 in final[n+1]:
                mark=final[n+1].index(temp1)
                final[n+1][mark+1]+=1
            elif temp1 not in final[n+1]:
                final[n+1].append(temp1)
                final[n+1].append(1)

list1=[]
for i in range(50):
    temp1=(final[2*i],final[2*i+1])
    list1.append(temp1)
final2=dict(list1)
print(final2)
```

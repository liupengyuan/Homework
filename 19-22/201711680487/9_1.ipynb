```python
import random,time

#A
a=[i for i in range(1,100001)]
print('列表A为：',a)
print('-'*40 + '分隔线' + '-' *40)

#B
b=[]
random.shuffle(a)
for i in range (0,100000):
    b.append(a[i])
print('列表B为：',b)
print('-'*40 + '分隔线' + '-' *40)

#随机抽1000个整数
c=[]
for i in range(0,1000):
    num=random.randint(1,110000)
    c.append(num)
print(c)
print('-'*40 + '分隔线' + '-' *40)

#用顺序查找法在B中找
print('用顺序查找法在B中查找上述整数：')
start_time = time.time()
for i in range (0,1000):
    if  c[i] in b:
        print('找到',c[i])
    else:
        print(c[i],'不在序列中')
end_time = time.time()
print('用时：',end_time-start_time,'秒')
print('-'*40 + '分隔线' + '-' *40)

#用二分查找法在A中查找
print('用二分查找法在A中查找上述整数：')
def bi_search(number, numbers, low, high):
    if low > high:
        return c[i],'不在序列中'
    else:
        mid = (low+high)//2
        if number == mid:
            return '找到',c[i]
        elif number > mid:
            return bi_search(number, numbers, mid+1, high)
        else:
            return bi_search(number, numbers, low, mid-1)
start = time.time()
for i in range (0,1000):
    print(bi_search(c[i],a,0,99999))
end = time.time()
print('用时：',end-start,'秒')
    
```

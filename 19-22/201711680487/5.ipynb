```python
#练习 2：写函数，共n个随机整数，整数范围在m与k之间，（n,m,k由用户输入）。求1：西格玛log(随机整数)，
#2：西格玛1/log(随机整数)
import random , math
def total1(n,m,k):
    total=0
    i=1
    while i<=n:
        number = random.randint(m,k)
        num = math.log(number)
        i=i+1
        total=total+num
    return total

def total2(n,m,k):
    total=0
    i=1
    while i<=n:
        number = random.randint(m,k)
        num = 1/(math.log(number))
        i=i+1
        total=total+num
    return total

m=int(input('输入一个较小的数'))
k=int(input('输入一个较大的数'))
n=int(input('输入随机整数的个数'))

print(total1(n,m,k))
print(total2(n,m,k))
      
```

```python
#练习 3：写函数，求s=a+aa+aaa+aaaa+aa...a的值，其中a是[1,9]之间的随机整数。例如2+22+222+2222+22222(此时共
#有5个数相加)，几个数相加由键盘输入。
import random
def total(n):
    a=random.randint(1,9)
    i=1
    total=0
    while i<=n:
        m=(10**i-1)*(a/9)
        total=total+m
        i=i+1
    return total

n=int(input('输入相加整数的个数'))

print(total(n))
    
```

```python
#挑战性练习：仿照task5，将猜数游戏改成由用户随便选择一个整数，让计算机来猜测的猜数游戏，要求和task5中人猜测
#的方法类似，但是人机角色对换，由人来判断猜测是大、小还是相等，请写出完整的猜数游戏。

'''
猜数函数流程：
请玩家设想一个整数n
取一个最多可猜测次数，计算机已猜测次数设为0，
一直做如下步骤直到已猜测次数大于最多可猜测次数：
    让计算机猜测一次
    计算机猜测次数加一
    告诉计算机目前猜测次数
    如果计算机猜中了：
        恭喜猜中
        显示数字及猜测次数
        退出循环
    如果计算机猜测数字大于number：
        告诉及计算机猜大了
    否则：
        告诉计算机猜小了
    告诉计算机还能猜几次
否则：
    显示超过次数，猜测失败。
'''
import random, math

def welcome():
    print(
        '''
           ======欢迎来到猜数游戏=======
        
                     /  \     ,    ,
           _._     _ |oo| _  / \__/ \
    
          _||||   ((/ () \))   /   \
    
          |||||/|  ( ==== )    |oo|    
           \____/  _`\  /'_    /   \
    
           /   /.-' /\<>/\ `\.( () )_._      
           |    `  /  \/  \  /`'--'////)
            \__,-'`|  |.  |\/ |/\/\|"\"` 
                   |  |.  | \___/\___/  
                   |  |.  |   |    |
          
          ======欢迎来到猜数游戏=======
        '''
    )

def win():
    print(
        '''
           ======成功=======
        
    
                ."".    ."",
                |  |   /  /
                |  |  /  /
                |  | /  /
                |  |/  ;-._ 
                }  ` _/  / ;
                |  /` ) /  /
                | /  /_/\_/\
                |/  /      |
                (  ' \ '-  |
                 \    `.  /
                  |      |
                  |      |
          
          ======WIN=======
        '''
    )
    
def lose():
    print(
        '''
           ======失败=======
        
    
                

                   .-"      "-.
                  /            \
                 |              |
                 |,  .-.  .-.  ,|
                 | )(__/  \__)( |
                 |/     /\     \|
       (@_       (_     ^^     _)
  _     ) \_______\__|IIIIII|__/__________________________
 (_)@8@8{}<________|-\IIIIII/-|___________________________>
        )_/        \          /
       (@           `--------`
       
       
       
          ======LOSE=======
        '''
    )

def game_over():
    print(
        '''
           ======游戏结束=======
        
             _________ 
            / ======= \ 
           / __________\ 
          | ___________ | 
          | | -       | | 
          | |         | | 
          | |_________| |________________ 
          \=____________/                ) 
          / """"""""""" \               / 
         / ::::::::::::: \          =D-' 
        (_________________) 

       
          ======GAME OVER=======
        '''
    )

def show_team():
    print('王丹')

def show_instruction():
    print('这是一个炒鸡好玩的游戏')

def menu():
    print('''
             =====游戏菜单=====
                1. 游戏说明
                2. 开始游戏
                3. 退出游戏
                4. 制作团队
             =====游戏菜单=====''')
def guess_game():
    n = int(input('请输入一个大于0的整数，作为神秘整数的上界。'))
    number = int(input('在规定范围内选取一个神秘整数'))
    max_times = math.ceil(math.log(n, 2))
    guess_times = 0
    guess = random.randint(1, n)
    max_=n
    min_=1
    
    while guess_times < max_times:
        print('计算机猜测为：',guess)
        guess_times = guess_times + 1
        m = int(input('猜大了输入1，猜小了输入2，猜测正确输入3'))
        print('一共可以猜', max_times, '次')
        print('已经猜了', guess_times, '次')
        
        if m==3:
            win()
            print('神秘数字是：', guess)
            print('比标准次数少', max_times-guess_times, '次')
            break
        elif m==1:
            max_=guess
            guess = random.randint(min_,guess)
        elif m==2:
            min_=guess
            guess = random.randint(guess,max_)
    else:
        print('神秘数字是：', number)
        lose()

# 主函数
welcome()
def main():
    while True:
        menu()
        choice = int(input('请输入你的选择'))
        if choice == 1:
            show_instruction()
        elif choice == 2:
            guess_game()
        elif choice == 3:
            game_over()
            break
        else:
            show_team()
            
#主程序
if __name__ == '__main__':
    main()
                
```

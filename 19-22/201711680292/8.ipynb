```python
# 成语接龙游戏

import random

def win():
    print('恭喜，你赢了')
    
def lose():
    print('很遗憾你输了')
    

def g(idiom):
    line=[]
    for ch in idiom:
        line.append(ch)
    return line
def f(idioms):
    m=[]
    for ch in idioms:
        m.append(ch)
    return m
        
    

def idiom_robot():
    with open(r'c:\temp\成语大全.txt') as fh:
        text = fh.read()
    idioms = text.split()
    idiom = random.choice(idioms)
    i=1
    a=0 
    r=0
    co=0
    while i>0:
        print(idiom)
        u=[]
        n=int(input('请输入成语字数'))
        ans=input('请输入成语：')
        for an in ans:
            u.append(an)
        if u[0]==g(idiom)[len(g(idiom))-1]:
            r+=10
            for k in range(len(f(idioms))-1):
                if g(f(idioms)[k])[0]==u[n-1]:
                    idiom=f(idioms)[k]
                    co+=10
                    break
    
        else:
            print(lose())
            co+=10
            break
    print('您的得分',r)
    print('计算机的得分',co)
        
    
    
idiom_robot()
                        

   


    
```

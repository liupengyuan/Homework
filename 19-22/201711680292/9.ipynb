```python
import random,time,bisect
def f():
    a=[]
    b=[]
    for i in range(1,100001):
        a.append(i)
        b.append(i)
    
    random.shuffle(b)
    c=[]
    for j in range(1000):
        n=random.randint(1,110000)
        c.append(n)
    
    start_time1=time.time()
    for word in c:
        if word in b:
            for i in range(len(b)):
                if b[i]==word:
                    print('在b中的位置是',i) 
                    break
        else:
            print(word,'不在序列b中')
    end_time1=time.time()
    print('------------------------------------------------------------------------------------------------------------------------------------')
    start_time2=time.time()
    for word in c:
        if word in a:   
            print('在a中的位置是',bisect.bisect_left(a, word))
        else:
            print(word,'不在序列a中')
    end_time2=time.time()
    print('-----------------------------------------------------------------------------------------------------------------------------------')
    print('算出a中位置的时间',end_time2-start_time2)
    print('算出b中位置的时间',end_time1-start_time1)

f()
```

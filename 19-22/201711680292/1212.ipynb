```python
import random
def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table

# 主程序
fh = open(r'E:\idioms.txt')
text = fh.read()
chs = get_ch_table(text.replace('\n', ''))
print(len(chs))

texta=[]
i=0
a=0
for j in range(10,110,10):
    while i<j:
        sh=random.choice(chs)
        texta.append(sh)
        fh = open(r'E:\texta.txt', 'a')
        fh.write(''.join(texta))
        fh.close()
        i=i+1
    a+=1
    print('text',a,texta)
    


```

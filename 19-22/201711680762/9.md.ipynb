```python
#2随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
```

```python
import random
A=[]
for i in range(10000):
    A.append(random.randint(0,100000))

def con(m):
    n=1
    for i in range(10000):
        if A[m]==A[m+1]:
            n+=1
        else:
            break
    print(A[m],n)

for m in range(10000):
    con(m)
    
```

```python
#4挑战性任务：统计本任务中的语料.txt文件，得到每一个词对应不同词性的频次，可表示为：{词:[('n',1000), ('v', 98)...],...}，即在dict对象中，key为词，value为列表，输出词频最高的50个条目。
```

```python
with open(r"E:\语料.txt") as ph:
    sing = ph.read()
    for line in sing:
        words = [word.split('/')[0] for word in sing.split()]
        for word in words:
            number = ph.count(word)
    print([word,number])
    
```

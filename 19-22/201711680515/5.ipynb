```python
import random,math
def pkq():
    total=0
    sump=0
    m=int(input('请输入一个数'))
    k=int(input('请输入一个更大的数'))
    n=int(input('请输入一个比两数之差小的数'))
    while sump<n:
        sump+=1
        total+=random.randint(m,k)
        q=math.sqrt(total/n)
    return q
pkq()
```

```python
import random,math
def love():
    s=0
    p=0
    t=0
    m=int(input('请输入一个数'))
    k=int(input('请输入一个更大的数'))
    n=int(input('请输入一个比两数之差小的数'))
    while t<n:
        number=random.randint(m,k)
        s+=math.log(number)
        p+=1/math.log(number)
        t+=1
    return s,p
love()
```

```python
import random
def i():
    s=0
    t=1
    total=0
    n=int(input('请输入数目'))
    while t <=n:
        number=random.randint(1,9)
        p=10**(t-1)*number
        s+=p
        total+=s
        t+=1
    return total
i()
```

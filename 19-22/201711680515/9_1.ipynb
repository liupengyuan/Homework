```python
import random,time,bisect
A=[]
B=[]
C=[]
D=[]
i=0
j=0
for i in range(1,100001):
    A.append(i)

for i in range(1,100001):
    B.append(i)
random.shuffle(B)

for j in range(1,10001):
    a=random.randint(1,110001)
    C.append(a)

end1=0
start1 = time.time()   
for number in B:
    if number in C:
        D.append(number)
        number+=1

        
end1 = time.time()
print(D)

start2 = time.time()
end2=0
for number in C:
    print(bisect.bisect_left(A, number))
    
end2 = time.time()

print('顺序法已经处理了{:.2f}秒...'.format(end1 - start1))
print('二分法已经处理了{:.2f}秒...'.format(end2 - start2))
```

```python
def demo(a,b):
    i=1
    while 1<=i<=b:
        print((b-i)*2*' ',end='')
        for j in range(i):
            print(a,' ',sep='',end='')
        print(b*a,end='')
        for j in range(i):
            print(' ',a,sep='',end='')
        print('\n')
        i=i+1
demo('*',5)
```

```python
def A(m,n):
    if m==0:
        print(n+1)
    if m>0:
        if n==0:
            m=m-1
            n=1
        if n>0:
            m=m-1
            n=A(m,n-1)
A(3,4)
```

```python
import random
def demo(n,a,b):
    i=1
    m=0
    while 1<=i<=n:
        x=random.randint(a,b)
        y=random.randint(a,b)
        g=random.choice('+-*/')
        if g=='+':
            z=x+y
            if z>0:
                print(x,g,y,'=',sep='')
                i=i+1
                o=int(input("请输入你的答案："))
                if o==z:
                    m=m+5
        if g=='-':
            z=x-y
            if z>0:
                print(x,g,y,'=',sep='')
                i=i+1
                o=int(input("请输入你的答案："))
                if o==z:
                    m=m+5
        if g=='*':
            z=x*y
            if z>0:
                print(x,g,y,'=',sep='')
                i=i+1
                o=int(input("请输入你的答案："))
                if o==z:
                    m=m+10
        if g=='/':
            z=x/y
            if x%y==0 and z>0:
                print(x,g,y,'=',sep='')
                i=i+1
                o=int(input("请输入你的答案："))
                if o==z:
                    m=m+10
    print('你的总分为：',m)
demo(15,3,25)
```

```python
n=int(input("你想让我答几道题："))
m=0
for i in range(n):
    a=str(input("请输入你的算式："))
    a=a.rstrip('=')
    if a.find('+')>0:
        x,y=a.split('+')
        print("答案为",int(x)+int(y))
        m=m+5
    if a.find('-')>0:
        x,y=a.split('-')
        print("答案为",int(x)-int(y))
        m=m+5
    if a.find('*')>0:
        x,y=a.split('*')
        print("答案为",int(x)*int(y))
        m=m+10
    if a.find('/')>0:
        x,y=a.split('/')
        print("答案为",int(x)/int(y))
        m=m+10
print('我的总分为：',m)
```

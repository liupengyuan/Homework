```python
import random
import time
import bisect
a=[]
b=[]
c=[]
for i in range(100001):
    a.append(i)
    b.append(i)
random.shuffle(b)
for i in range(1000):
    c.append(random.randint(1,110000))
start=time.time()
for n in c:
    for m in b:
        if n==m:
            print(m)
end=time.time()
print(end-start)
start=time.time()
for n in c:
    k=bisect.bisect_left(a,n)
    if 0<=k<=100000:
        print(a[k])
end=time.time()
print(end-start)
```

```python
import random
with open(r'c:\temp\成语大全.txt') as fl:
    text=fl.read()
    idioms = text.split()
def select_idiom(peo_idiom):
    list_idiom=[]
    for idiom in idioms:
        
        if idiom[0]==peo_idiom[-1]:
            list_idiom.append(idiom)
    return list_idiom
u=0
c=0  
a=random.choice(idioms)
print(a)

b=input('请输入一个成语')
if b[0]==a[-1]and b in idioms:
    select_idiom(b)
    u+=10
    if select_idiom(b)==[]:
        print('over')
        
        
    else:
        com_idiom=random.choice(select_idiom(b))
        print(com_idiom)
        c+=10
        while 1<2:
       
     


            b=input('请输入一个成语')
            if b[0]==com_idiom[-1]and b in idioms:
                u+=10
                select_idiom(b)
                if select_idiom(b)==[]:
                    print('over')
                    break
                
                else:    
                    com_idiom=random.choice(select_idiom(b))
                    print(com_idiom)
                    c+=10
            else:
                print('over')
                break
else:
    print('over')
    
print('你的得分为',u)
print('电脑得分为',c)
```

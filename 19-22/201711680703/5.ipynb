```python
#写函数，求n个随机整数均值的平方根，整数范围在m与k之间（n,m,k由用户输入）。

import random,math
def pfg():
    m=int(input('请输入一个较小的整数'))
    k=int(input('请输入一个较大的整数'))
    n=int(input('请输入一个小于前两个数之差的数'))
    total=0
    times=0
    while times<n:
        times+=1
        total+=random.randint(m,k)
        s=math.sqrt(total/n)
    return s
print(pfg())
```

```python
#写函数，共n个随机整数，整数范围在m与k之间，（n,m,k由用户输入）。求1：西格玛log(随机整数)，2：西格玛1/log(随机整数)

import random,math
m=int(input('请输入一个较小的整数'))
k=int(input('请输入一个较大的整数'))
n=int(input('请输入一个小于前两个数之差的数'))
def xgm_1():
    times=0
    a=0
    b=0
    while times<n:
        s=random.randint(m,k)
        times+=1
        a+=math.log(s)
        b+=1/math.log(s)
    return a,b
print(xgm_1())
```

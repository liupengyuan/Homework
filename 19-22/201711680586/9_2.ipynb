```python
#练习2
#随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
import random
numbers_=[]
i=0
while i<10000:
    number=random.randint(1,100000)
    numbers_.append(number)
    i+=1


from operator import itemgetter
numbers=sorted(numbers_,reverse=False)

def count_numbers_freq_dict():
    numbers_freq_dict={}
    for number_ in numbers:
        if number_ in numbers_freq_dict: #如果词典中有键为word，则该word的value+1
            numbers_freq_dict[number_] += 1
        else:                       #否则，词典中加入键word，其value为1
            numbers.append(number_)
            numbers_freq_dict[number_] = 1
                    
    return numbers_freq_dict
    print(numbers,numbers_freq_dict)

table=count_numbers_freq_dict()
for i,item in enumerate(table):
    print(item,table[item])
```

```python
#练习4
#挑战性任务：统计本任务中的语料.txt文件，得到每一个词对应不同词性的频次，
#可表示为：{词:[['n',1000], ['v', 98]...],...}，即在dict对象中，key为词，value为列表，输出词频最高的50个条目。
fh=open(r'd:\temp\语料.txt')

word_table=[]
words=[word.split('/')[0] for word in fh.split()]
for word in words:
    if word not in word_table:
        word_table.append(word)
return word_table

characteristic_table=[]
characteristics=[characteristic.split('/')[:] for characteristic in fh.split()]
for characteristic in characteristics:
    if characteristic in characteristic_table:
        characteristic_table.append(characteristic)
return characteristic_table

freq=Counter(words)
print({word:[[characteristic,freq]]}[:50])
```

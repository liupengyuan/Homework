```python
#第一题
def tuxing():
    symble=input('请输入符号')
    line=int(input('请输入行数'))
    n=1
    while n<=line:
        print((line-n)*'  ',(symble+' ')*n,symble*line,(' '+symble)*n,(line-n)*'  ',sep='')
        n=n+1
```

```python
tuxing()
```

```python
#第二题
def Ackermann():
    m=int(input('请输入数m'))
    n=int(input('请输入数n'))
    
    if m==0:
        print(n+1)
    elif m>0:
        if n==0:
            print(n-m)
        elif n>0:
            print(n-m+1)
    else:
        print('wrong')
```

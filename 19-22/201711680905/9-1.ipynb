```python
# A1,2,3,4100000  BA
# 10001-110000BA
# 


import random

A = []
i = 1
while i <=100000:
    a = i
    A.append(a)
    i+=1
print(A)
```

```python
import random

B = random.shuffle(A)
B = A
print(B)
```

```python
import random,time

i = 0
xs = []
while i < 1000:
    x = random.randrange(1,110001)
    i+=1
    xs.append(x)
print(xs)
    
n = 0
start_time1 = time.time()
for x in xs:
    if x in B:
        n+=1
    else:
        n=n
    end_time1 = time.time()
time.B = end_time1 - start_time1
print('B',time.B)
    

def bi_search(num,lst,low,high):
    x1 = sorted(xs)
    start_time2 = time.time()
    for x in x1:
        if x > A[-1]:
            return -1
        else:
            mid = (low+high)//2
            if x == lst[mid]:
                return mid
            elif x > lst[mid]:
                return bi_search(num,lst, mid+1, high)
            else:
                return bi_search(num, lst, low, mid-1)
        end_time2 = time.time()
    time.A = end_time2 - start_time2
    print('A,',time.A)

bi_search(x in x1,A,A[0],A[-1])
```

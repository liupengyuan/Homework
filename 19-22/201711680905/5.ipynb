```python
# 练习 1：写函数，求n个随机整数均值的平方根，整数范围在m与k之间（n,m,k由用户输入）。
import random
def num():
    m=int(input())
    k=int(input())
    n=int(input('输入整数个数：'))
    i=1
    total=0
    if m>k:
        a=m
        m=k
        k=a
    while i<=n:
        s=random.randint(m,k)
        total+=s
        i+=1
    return total
    print((total/n)**(1/2))
# 主程序
num()
```

```python
# 练习 2：写函数，共n个随机整数，整数范围在m与k之间，（n,m,k由用户输入）。求1：西格玛log(随机整数)，2：西格玛1/log(随机整数)
import random,math
def num():
    m=int(input())
    k=int(input())
    n=int(input())
    i=1
    total_1=0
    total_2=0
    if m>k:
        a=m
        m=k
        k=a
    while i<=n:
        s=random.randint(m,k)
        total_1+=math.log(s)
        total_2+=1/math.log(s)
        i+=1
    return total_1,total_2
# 主程序
num()
    
    
```

```python
# 练习 3：写函数，求s=a+aa+aaa+aaaa+aa...a的值，其中a是[1,9]之间的随机整数。例如2+22+222+2222+22222(此时共有5个数相加)，几个数相加由键盘输入。
import random
def su():
    n=int(input('输入相加数个数：'))
    i=1
    s=0
    total=0
    a=random.randint(1,9)
    while i<=n:
        s=a*(10**(i-1))+s
        i+=1
        total+=s
    return total
# 主程序
su()
```

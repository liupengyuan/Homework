```python
# 2、随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。

import random
i = 0
nums = []
while i <=100000:
    num = random.randint(0,100000)
    i+=1
    nums.append(num)
    
nums.sort()

for num in nums:
    times = nums.count(num)
    print(num,times)
```

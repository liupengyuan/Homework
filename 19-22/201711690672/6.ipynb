```python
#写函数，返回某个元素/对象在一个list中的位置，如果不在，则返回-1.
def test(numbers):
    n = int(input('请输入一个整数，回车结束'))
    if n in numbers:
        a=numbers.index(n)
        print(a)
    else:
        print(-1)
numbers = [1, 2, 3, 4, 5]
test(numbers)
```

```python
#写函数，可将两个相同长度的list，间隔插入，生成新的list。例如：给两个list，a=[1,2,3,4], b=[5,6,7,8]。则可以生成：[1,5,2,6,3,7,4,8]。
def test(a,b):
    numbers=[]
    i=0
    while i<len(a):
        x=a[i]
        numbers.append(x)
        y=b[i]
        numbers.append(y)
        i+=1
    print('生成新的list: ',numbers)
a=[1,2,3,4]
b=[5,6,7,8]
test(a,b)
```

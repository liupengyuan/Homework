```python
def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table
#主程序
import random
fh = open(r'C:\temp\idioms.txt')
text = fh.read()
chs = get_ch_table(text.replace('\n', ''))

x1=''   
for i in range(10): 
    ch=random.choice(chs)
    x1=x1+ch
print(x1)
list=[x1]
fh=open(r'c:\temp\1.txt','w')
fh.writelines(list)

x2=''   
for i in range(20): 
    ch=random.choice(chs)
    x2=x2+ch
print(x2)
list=[x2]
fh=open(r'c:\temp\2.txt','w')
fh.writelines(list)

x3=''   
for i in range(30): 
    ch=random.choice(chs)
    x3=x3+ch
print(x3)
list=[x3]
fh=open(r'c:\temp\3.txt','w')
fh.writelines(list)

x4=''   
for i in range(40): 
    ch=random.choice(chs)
    x4=x4+ch
print(x4)
list=[x4]
fh=open(r'c:\temp\4.txt','w')
fh.writelines(list)

x5=''   
for i in range(50): 
    ch=random.choice(chs)
    x5=x5+ch
print(x5)
list=[x5]
fh=open(r'c:\temp\5.txt','w')
fh.writelines(list)

x6=''   
for i in range(60): 
    ch=random.choice(chs)
    x6=x6+ch
print(x6)
list=[x6]
fh=open(r'c:\temp\6.txt','w')
fh.writelines(list)

x7=''   
for i in range(70): 
    ch=random.choice(chs)
    x7=x7+ch
print(x7)
list=[x7]
fh=open(r'c:\temp\7.txt','w')
fh.writelines(list)

x8=''   
for i in range(80): 
    ch=random.choice(chs)
    x8=x8+ch
print(x8)
list=[x8]
fh=open(r'c:\temp\8.txt','w')
fh.writelines(list)

x9=''   
for i in range(90): 
    ch=random.choice(chs)
    x9=x9+ch
print(x9)
list=[x9]
fh=open(r'c:\temp\9.txt','w')
fh.writelines(list)

x10=''   
for i in range(100): 
    ch=random.choice(chs)
    x10=x10+ch
print(x10)
list=[x10]
fh=open(r'c:\temp\10.txt','w')
fh.writelines(list)


fh.close()
```

```python
#10-10
import random
def sort_simple_selection(A):
    for n in range(len(A)-1):
        min = n
        for m in range(n+1, len(A)):
            if A[m] < A[min]:
                min = m
        A[n], A[min] = A[min], A[n]

A=[]
for i in range(10000):
    integer=random.randint(0,100000)
    A.append(integer)
sort_simple_selection(A)

for a in A:
    print(a,': ',A.count(a))
```

```python
#随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
import random
def sort_simple_selection(A):
    for n in range(len(A)-1):
        min = n
        for m in range(n+1, len(A)):
            if A[m] < A[min]:
                min = m
        A[n], A[min] = A[min], A[n]

A=[]
for i in range(10000):
    integer=random.randint(0,100000)
    A.append(integer)
sort_simple_selection(A)

for a in A:
    print(a,': ',A.count(a))
```

```python
#对本任务中的语料.txt文件，抽取其5001-10000行存为test1.txt文件，写函数
#可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）
with open(r'C:\temp\test1.txt','w') as fw,\
open(r'C:\temp\语料.txt') as fr:
    line_num = 0
    for line in fr:
        if line_num < 5000:
            line_num += 1
        elif 5000 <= line_num < 10000:
            fw.write(line)
            line_num += 1
        else:
            break
            

def c_table(filename):
    words_table = set()
    c_table = set()
    with open(filename) as f:
        for line in f:
            words_table |= {word.split('/')[0] for word in line.split()} 
            for word in words_table:
                c_table |= {c for c in word}
        return c_table
                                        
word_1 = c_table(r'C:\temp\test.txt')
word_2 = c_table(r'C:\temp\test1.txt')

print('共用字是',word_1 & word_2)
print('独用字是',word_1 - word_2)
```

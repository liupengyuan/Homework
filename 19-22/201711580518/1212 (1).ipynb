```python

def get_ch_table(line):
    ch_table = []
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table


fh = open(r'c:\Users\asus\Downloads\idioms_correct.txt')
text = fh.read()
chs = get_ch_table(text.replace('\n', ''))
def www():
    for w in range(10,101,10):
        import random
        
        
       
        n=str(w/10)
        fh = open('c:/temp/'+n+'.txt', 'w')
        for i in range(w):    
            idiom = random.choice(chs)
            print(idiom)
            h=str(idiom)
            fh.write(h + '\n')
        
www()              
























```

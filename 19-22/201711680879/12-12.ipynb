```python
fh = open(r'c:\temp\成语大全.txt')
text = fh.read()
fh.close()

text = text.replace('、', '')
idioms = text.split('\n')

fh = open(r'c:\temp\成语大全2.txt', 'w')
fh.writelines(idioms)
fh.close()

import random
def get_ch_table(line):
    ch_table=[]
    for ch in line:
        if ch not in ch_table:
            ch_table.append(ch)
    return ch_table

fh = open(r'c:\temp\成语大全2.txt')
text = fh.read()
chs = get_ch_table(text.replace('\n', ''))

for i in range(1,11):
    for j in range(1,10*i+1):
        a=random.choice(chs)
        print(i,a)
```

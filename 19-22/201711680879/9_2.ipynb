```python
#随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。
```

```python
import random
def count_words_freq_dict(filename):
    words_freq_dict = {}
    for word in filename:
        if word in words_freq_dict:
            words_freq_dict[word] += 1
        else:
            words_freq_dict[word] = 1
                    
    return words_freq_dict

filename = [random.randint(1,100001) for i in range(10000)]
sorted_filename = sorted(filename)
table = count_words_freq_dict(sorted_filename)

for i, item in enumerate(table):
    print(item, table[item])
```

```python
#对本任务中的语料.txt文件，抽取其5001-10000行存为test1.txt文件，写函数，可得到其与本任务中test.txt文件的共用字以及独用字（相关概念自行百度）。
```

```python
A_ip=[]
B_ip=[]
filename = r'c:\temp\语料.txt'
with open(filename) as rf:
    for line in rf:
        ip=line.split('\n').pop(0)
        A_ip.append(ip)

with open(r'c:\temp\test1.txt', 'w') as fw:
    line_number = 0
    for line in filename:
        if 5000 < line_number < 10000:
            fw.write(line)
            line_number += 1
        else:
            break

with open(r'c:\temp\test1.txt', 'w') as tf:
    for line in tf:
        ip=line.split('\n').pop(0)
        B_ip.append(ip)
        
same_ip=[]
dif_ip=[]
for each_t in A_ip:
    if each_t in B_ip:
        same_ip.append(each_t)
    else:
        dif_ip.append(each_t)
print(same_ip)
```

```python
listA=[]
for i in range(100000):
    listA.append(list[i])

import random
random.shuffle(list)
listB=list
```

```python
S=[]
while len(S)<1000:
    a=random.randint(1,110000)
    if a not in S:
        S.append(a)
```

```python
import time
start=time.time()

for i in range(len(listB)):
    if listB[i] in S:
        print([listB[i],i])
end=time.time()
print(end-start)
```

```python
start=time.time()
for i in range(len(S)):
    L=0
    R=len(listA)-1
    while L<R:
        M=(L+R)//2
        if listA[M] == S[i]:
            print([listA[M],M])
            break
        
        if S[i]>listA[M]:
            L=M
        if S[i]<listA[M]:
            R=M
        elif S[i]>listA[len(listA)-1]:
            break 
        
        
end=time.time()
print(end-start)
```

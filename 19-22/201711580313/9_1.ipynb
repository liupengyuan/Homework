```python
A = []
for i in range(1,100001):
    A.append(i)
a = A[:]
import random,time,bisect
random.shuffle(a)
B = a
C = []
for j in range(1,1001):
    C.append(random.randint(1,11001))
D = []
start_time = time.time()
for k in C:
    if k in B:
        D.append(k)
end_time = time.time()
print(D,'\n'"用了{}秒".format(end_time - start_time))
E = []
start_time1 = time.time()
for n in C:
    E.append(bisect.bisect_left(A,n))
end_time1 = time.time()    
print(E,'\n''用了{}秒'.format(end_time1 - start_time1))
```

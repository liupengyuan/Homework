```python
import random,time,bisect

A = [i for i in range(100001)]
for n in range(1,1001):
    if n==1:
        start_time1 = time.time()
    elif n==1000:
        end_time1 = time.time()
    x=random.randint(1,110001)
    print(x,'在A中的索引为',bisect.bisect_left(A, x))
print('所用总时间为',end_time1-start_time1)
```

```python
random.shuffle(A)
B=A
for n in range(1001):
    if n==1:
        start_time2 = time.time()
    elif n==1000:
        end_time2 = time.time()
    x=random.randint(1,110001)
    if x in B:
        print(x,'在列表中的索引为',B.index(x))
    else:
        print(x,'不在B中')
print('所用总时间为',end_time2-start_time2)
```

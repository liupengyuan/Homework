```python
#1
def triangle():
    x=str(input('请给定符号'))
    n=int(input('请给定行数'))
    for i in range(n+1):
        for k in range(n*2-i,0,-1):
            print('',end='  ')
        for j in range(0,i):
            print(x,end=' ')
        print()
triangle()
```

```python
#2
def A(m,n):
    if m == 0:
        a=n+1
        return a
    elif m>0 and n == 0:
        m=m-1
        n=1
        return A(m,n)
    elif m>0 and n>0:
        n=A(m,n-1)
        m=m-1
        return A(m,n)

A(3,4)
```

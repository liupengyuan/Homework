```python
#练习 1：写函数，求n个随机整数均值的平方根，整数范围在m与k之间（n,m,k由用户输入）。
import random

def f():
    n=int(input('请输入随机整数的个数，以回车结束'))
    m=int(input('请输入随机整数的下界，以回车结束'))
    k=int(input('请输入随机整数的上界，以回车结束'))
    total=0
    i=0
    while i<n:
        num=random.randint(m,k)
        total=total+num
        i=i+1
        x=total/n
    print(x**(1/2))
    
f()
```

```python
#练习 2：写函数，共n个随机整数，整数范围在m与k之间，（n,m,k由用户输入）。求1：西格玛log(随机整数)，2：西格玛1/log(随机整数)
import random,math

def g(content):
    n=int(input('请输入随机整数的个数，以回车结束'))
    m=int(input('请输入随机整数的下界，以回车结束'))
    k=int(input('请输入随机整数的上界，以回车结束'))
    i=0
    total=0
    if content==1:
        while i<n:
            num=random.randint(m,k)
            a=math.log(num)
            total=total+a
            i=i+1
        print(total)
    if content==2:
        while i<n:
            num=random.randint(m,k)
            a=math.log(num)
            total=total+1/a
            i=i+1
        print(total)
g(1)
g(2)
```

```python
#练习 3：写函数，求s=a+aa+aaa+aaaa+aa...a的值，其中a是[1,9]之间的随机整数。例如2+22+222+2222+22222(此时共有5个数相加)，
#几个数相加由键盘输入。
import random

def s():
    n=int(input("请输入相加整数的个数，以回车结束"))
    a=random.randint(1,9)
    i=0
    total=0
    while i<n:
        total=total+a
        i=i+1
        a=a*(10**i)+a
    print(total)
s()
```

```python
# 210-10
import random
my_list = []
for i in range(10000):
    a = random.randint(0,100000)
    my_list.append(a)
# print(my_list)
words_freq_dict = {}
for word in my_list:
    if word in words_freq_dict: #wordwordvalue+1
        words_freq_dict[word] += 1
    else:                       #wordvalue1
        words_freq_dict[word] = 1
# print(words_freq_dict)
print (sorted(words_freq_dict.items(), key=lambda v: v[0]))
```
